5. The following is a definition of the class Point that takes in the coordinates x, y. Add a method plot_point that plots the position of a point.



```python
import matplotlib.pyplot as plt

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def plot_point(self):
        plt.scatter(self.x, self.y)
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title('Plot of Point')
        plt.show()
```

```python
p = Point(3, 5)
p.plot_point()
```

6. Use the class from Problem 5 and add a method calculate_dist which takes in x and y from another point, and returns the distance calculated between the two points.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def calculate_dist(self, other_x, other_y):
        dx = other_x - self.x
        dy = other_y - self.y
        distance = (dx**2 + dy**2)**0.5
        return distance
```

```python
p1 = Point(3, 5)
p2 = Point(7, 2)
distance = p1.calculate_dist(p2.x, p2.y)
print(distance)
```

```python
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species
    
    def eat(self):
        print(f"{self.name} is eating.")
    
    def sleep(self):
        print(f"{self.name} is sleeping.")

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name, "Dog")
        self.breed = breed
    
    def bark(self):
        print(f"{self.name} is barking.")

class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name, "Cat")
        self.color = color
    
    def meow(self):
        print(f"{self.name} is meowing.")
```

```python
# Creating instances and using methods
dog1 = Dog("Max", "Labrador")
dog1.eat()
dog1.sleep()
dog1.bark()

cat1 = Cat("Whiskers", "Gray")
cat1.eat()
cat1.sleep()
cat1.meow()
```
