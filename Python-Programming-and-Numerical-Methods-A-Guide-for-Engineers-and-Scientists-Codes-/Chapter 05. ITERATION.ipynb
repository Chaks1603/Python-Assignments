```python
def my_max(x):
    if len(x) == 0:     raise ValueError("Empty list has no maximum value.")
    
    maximum = x[0]
    for element in x:
        if element > maximum:    maximum = element
    return maximum

my_list = [3, 7, 2, 9, 5]
my_max(my_list)
```

```python
def my_n_max(x, n):
    if n <= 0:
        raise ValueError("n must be a positive integer.")
    if n > len(x):
        raise ValueError("n cannot be greater than the length of x.")
    
    sorted_x = sorted(x, reverse=True)
    return sorted_x[:n]

x = [7, 9, 10, 5, 8, 3, 4, 6, 2, 1]
n = 3
my_n_max(x, n)
```

```python
def my_trig_odd_even(m):
    q = np.zeros_like(m, dtype=float)
    
    for i in range(m.shape[0]):
        for j in range(m.shape[1]):
            if m[i, j] % 2 == 0:
                q[i, j] = np.sin(m[i, j])
            else:
                q[i, j] = np.cos(m[i, j])
    return q

m = np.array([[2, 3,  4],
              [5, 6,  7],
              [8, 9, 10]])

my_trig_odd_even(m)
```

```python
def my_mat_mult(P, Q):
    m, p = P.shape
    p, n = Q.shape
    M = np.zeros((m, n))
    
    for i in range(m):
        for j in range(n):
            for k in range(p):
                M[i, j] += P[i, k] * Q[k, j]
    return M
```

```python
P = np.ones((3, 3))
my_mat_mult(P, P)
```

```python
P = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
Q = np.array([[1, 1, 1], [2, 2, 2], [3, 3, 3], [4, 4, 4]])
my_mat_mult(P, Q)
```

```python
def my_saving_plan(P0, i, goal):
    years = 0
    balance = P0
    
    while balance < goal:
        balance *= (1 + i)
        years += 1
    return years
```

```python
my_saving_plan(1000, 0.05, 2000)
```

```python
def my_find(M):
    indices = []
    
    for i in range(len(M)):
        if M[i] == 1:
            indices.append(i)
   
    return indices
```

```python
def my_monopoly_dice():
    total = 0
    same_roll = False
    
    while True:
        dice1 = np.random.randint(1, 6)
        dice2 = np.random.randint(1, 6)
        roll_sum = dice1 + dice2
       
        total += roll_sum
        
        if dice1 == dice2:
            same_roll = True
        else:
            if same_roll:
                return total
            same_roll = False
```

```python
def my_is_prime(n):
    if n < 2:
        return 0
    
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return 0
    return 1
```

```python
my_is_prime(17)
```

```python
my_is_prime(12)
```

```python
def my_n_primes(n):
    primes = []
    num = 2  # Start checking for primes from 2
    
    while len(primes) < n:
        if my_is_prime(num):
            primes.append(num)
        num += 1
    
    return primes
```

```python
my_n_primes(5)
```

```python
my_n_primes(10)
```

```python
def my_is_fibonacci(n):
    if n < 0:
        return 0

    a, b = 0, 1
    while b < n:
        a, b = b, a + b

    return n == b
```

```python
def my_n_fib_primes(n):
    fib_primes = []
    count = 0
    num = 3

    while count < n:
        if my_is_prime(num) and my_is_fibonacci(num):
            fib_primes.append(num)
            count += 1
        num += 1
    
    return fib_primes
```

```python
my_n_fib_primes(8)
```

```python
my_n_fib_primes(3)
```

```python
def my_trig_odd_even(M):
    Q = np.zeros_like(M, dtype=float)

    for i in range(M.shape[0]):
        for j in range(M.shape[1]):
            if M[i, j] % 2 == 0:
                Q[i, j] = np.cos(np.pi / M[i, j])
            else:
                Q[i, j] = np.sin(np.pi / M[i, j])
    return Q
```

```python
M = np.array([[3, 4], 
              [6, 7]])

my_trig_odd_even(M)
```

```python
def my_connectivity_mat_2_dict(C, names):
    node = {}
    
    for i in range(len(C)):
        connected_points = []
        
        for j in range(len(C[i])):
            if C[i][j] == 1:
                connected_points.append(j+1)
        
        node[names[i]] = connected_points
    return node
```

```python
C = [[0, 1, 0, 1], [1, 0, 0, 1], [0, 0, 0, 1], [1, 1, 1, 0]]
names = ["Los Angeles", "New York", "Miami", "Dallas"]

my_connectivity_mat_2_dict(C, names)
```
