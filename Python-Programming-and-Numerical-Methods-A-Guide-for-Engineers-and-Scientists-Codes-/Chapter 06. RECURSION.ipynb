```python
def my_chebyshev_poly1(n, x):
    
    if n == 0:
        return [1] * len(x)
    elif n == 1:
        return x
    else:
        Tn_minus_2 = [1] * len(x)
        Tn_minus_1 = x
        Tn = []
        for _ in range(2, n + 1):
            Tn = [2 * xi * Tn_minus_1[i] - Tn_minus_2[i] for i, xi in enumerate(x)]
            Tn_minus_2, Tn_minus_1 = Tn_minus_1, Tn
        return Tn
```

```python
def my_ackermann(m, n):
    if m == 0:
        return n + 1
    elif m > 0 and n == 0:
        return my_ackermann(m - 1, 1)
    elif m > 0 and n > 0:
        return my_ackermann(m - 1, my_ackermann(m, n - 1))
```

```python
def my_n_choose_k(n, k):
    if n == k or k == 0:
        return 1
    else:
        return my_n_choose_k(n - 1, k) + my_n_choose_k(n - 1, k - 1)
```

```python
def my_pascal_row(m):
    if m == 1:
        return [1]
    else:
        prev_row = my_pascal_row(m - 1)
        row = [1]
        for i in range(1, m - 1):
            row.append(prev_row[i - 1] + prev_row[i])
        row.append(1)
        return row
```

```python
def my_spiral_ones(n):
    if n == 1:
        return 1
    A = np.zeros((n,n))
    B = np.zeros((n-1,n-1))
    B = my_spiral_ones(n-1)
    
    if n % 4 == 1:
        A[0,:] = 1
        A[1:n,0:n - 1] = np.logical_or(A[1:n,0:n - 1], B)
    elif n % 4 == 2:
        A[:,n - 1] = 1
        A[0:n - 1, 0:n - 1] = np.logical_or(A[0:n - 1,0:n - 1], B)
    elif n % 4 == 3:
        A[n - 1,:] = 1
        A[0:n - 1,1:n] = np.logical_or(A[0:n - 1,1:n], B)
    else :
        A[:,0] = 1
        A[1:n,1:n] = np.logical_or(A[1:n,1:n], B)
    return A
```

12. Rewrite the Tower of Hanoi function in this chapter without using recursion.

```python
def my_towers(N, from_tower, to_tower, alt_tower):
    stack = [(N, from_tower, to_tower, alt_tower)]

    while stack:
        N, from_tower, to_tower, alt_tower = stack.pop()

        if N != 0:
            stack.append((N - 1, alt_tower, to_tower, from_tower))
            print("Move disk %d from tower %d to tower %d." % (N, from_tower, to_tower))
            stack.append((N - 1, from_tower, alt_tower, to_tower))
```

13. Draw the recursion tree for my_quicksort([5 4 6 2 9 1 7 3]).

```python
def my_quicksort(lst):
    if len(lst) <= 1:
        sorted_list = lst
    else:
        pivot = lst[0]
        bigger = []
        smaller = []
        same = []
        for item in lst:
            if item > pivot:
                bigger.append(item)
            elif item < pivot:
                smaller.append(item)
            else:
                same.append(item)
        sorted_list = my_quicksort(smaller) + same + my_quicksort(bigger)
    return sorted_list
```

```python
my_quicksort([2, 1, 3, 5, 6, 3, 8, 10])
```

```python
my_quicksort([5, 4, 6, 2, 9, 1, 7, 3])
```

14. Rewrite the quicksort function in this chapter without using recursion

```python
def my_quicksort(lst):
    stack = [(0, len(lst) - 1)]  # Create a stack to track subarray boundaries

    while stack:
        start, end = stack.pop()  # Retrieve subarray boundaries from the stack
        if start >= end:
            continue  # Skip if subarray has 1 or fewer elements

        pivot = lst[start]  # Select the pivot as the first element
        left = start + 1
        right = end

        while True:
            while left <= right and lst[left] <= pivot:
                left += 1
            while left <= right and lst[right] >= pivot:
                right -= 1

            if left > right:
                break  # Elements have been partitioned

            lst[left], lst[right] = lst[right], lst[left]  # Swap misplaced elements

        lst[start], lst[right] = lst[right], lst[start]  # Move pivot to its sorted position

        # Push the boundaries of the remaining subarrays onto the stack
        if right - 1 > start:
            stack.append((start, right - 1))
        if right + 1 < end:
            stack.append((right + 1, end))

    return lst
```
