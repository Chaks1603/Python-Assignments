Q1. A cycloid is the curve traced by a point located on the edge of a wheel rolling along a flat surface.                           
The (x, y) coordinates of a cycloid generated from a wheel with radius, $r$, can be described by the parametric equations:       
        $\qquad\qquad x = r(φ −sin φ) $                                                                                                         
        $\qquad\qquad y = r(1 −cos φ) $                                                                                                            
where $φ$ is the number of radians that the wheel has rolled through.                                                               
 Generate a plot of the cycloid for $0 ≤ φ ≤ 2π$ using 1000 increments and $r = 3$.                                                  
 Give your plot a title and labels.                                                                                             
 Turn on the grid and modify the axis limits to make the plot neat and attractive.

```python
r=3
phi=np.arange(0,2*np.pi,0.001)

x=r*(phi-np.sin(phi))
y=r*(1-np.cos(phi))

plt.plot(phi,x, phi,y)
plt.grid(which="both")
plt.title(" a plot of the cycloid",fontsize = 30)
plt.legend(['x','y'],fontsize = 18)
plt.xlabel(r"$\phi$",fontsize = 18)
plt.ylabel("Functions",fontsize = 18)
plt.show()
```

Q2.Consider the following function:

$\qquad\qquad \displaystyle{ y(x)=\sqrt{\frac{100(1-0.01x^2)^2 + 0.02x^2}{(1-x^2+0.1x^2)}}}$                      
Generate a 2×2 subplot of y(x) for 0 ≤ x ≤ 100 using plot, semilogx, semilogy, and loglog.                  
Use a fine enough discretization in x to make the plot appear smooth.                                
Give each plot axis labels and a title. Turn on the grid.                                              
Which plot seems to convey the most information?.

```python
def f(x):
    num=100*(1-(0.01*x*x))**2+(0.02*x*x)
    dnum=(1-x*x)**2+(0.1*x*x)
    sr=(num/dnum)**0.5
    return sr

x=np.arange(0,100,0.01)
y=f(x)

plt.figure(figsize = (20, 8))
plt.subplot(1, 4, 1)
plt.plot(x,y)
plt.title("Plot",fontsize = 18)
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.subplot(1, 4, 2)
plt.loglog(x,y)
plt.title("Loglog",fontsize = 18)
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(which="both")
plt.subplot(1, 4, 3)
plt.semilogx(x,y)
plt.title("Semilogx",fontsize = 18)
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(which="both")
plt.subplot(1, 4, 4)
plt.semilogy(x,y)
plt.title("Semilogy",fontsize = 18)
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.tight_layout()
plt.show()
```

Q7. Write a function `my_polygon(n)` that plots a regular polygon with n sides and radius 1.                                 
Recall that the radius of a regular polygon is the distance from its centroid to the vertices.         
Use plt.axis("equal") to make the polygon look regular.                                                              
Remember to give the axes a label and a title. Use plt.title to title the plot according to the number of sides.          
Hint: This problem is significantly easier if you think in polar coordinates.                             
Recall that a complete revolution around the unit circle is $2\pi$ radians.                                                    
Note that the first and last point on the polygon should be the point associated with the polar coordinate angles,                                                                                 
$0$ and $2\pi$, respectively.

```python
def  my_polygon(n):
    r=1
    t=np.arange(0,2*np.pi+(2*np.pi/n),(2*np.pi/n))
    x=r*np.sin(t)
    y=r*np.cos(t)  
    plt.plot(x,y)
    plt.title('%i side polygon'%n,fontsize=18)
    plt.axis("equal")
    return None

plt.figure(figsize = (20,16))
for n in range(3,10,1):
    plt.subplot(3,3,n)
    my_polygon(n)
```

Q10. Assume you have three points at the corner of an equilateral triangle, $P_1= (0, 0),P_2 = \bigg(0.5,\frac{\sqrt{2}}{2}\bigg),and P_3 = (1, 0)$.   
Generate another set of points $p_i = (x_i, y_i )$ such that $p_1 = (0, 0)$ and $p_{i+1}$ is the midpoint between $p_i$ and $P_1$ with $33\%$ probability,                                                                                                
the midpoint between $p_i$ and $P_2$ with $33\%$ probability, and the midpoint between $p_i$ and $P_3$ with $33\%$ probability.                                                                                                                             
Write a function `my_sierpinski(n)` that generates the points pi for $i = 1, . . . , n$.                                      
The function should make a plot of the points using blue dots (i.e., "b." as the third argument to the plt.plot function).

```python
def my_sierpinski(n):
    # Define the corner points of the equilateral triangle
    P1 = np.array([0, 0])
    P2 = np.array([0.5, np.sqrt(2)/2])
    P3 = np.array([1, 0])

    # Initialize the list to store the generated points
    points = [P1]

    # Generate n points based on the probability distribution
    for _ in range(1,n):
        # Choose the next point based on the probabilities
        r = np.random.random()
        if r < 1/3:
            next_point = (points[-1] + P1) / 2
        elif r < 2/3:
            next_point = (points[-1] + P2) / 2
        else:
            next_point = (points[-1] + P3) / 2

        points.append(next_point)
    points = np.array(points)

    # Plot the points using blue dots
    plt.figure(figsize=(4,4))
    plt.plot(points[:, 0], points[:, 1], 'b,')
    plt.title('my_sierpinski with {} iterations'.format(n))
    plt.show()
```

```python
my_sierpinski(100)
```

Q11. Assume you are generating a set of points $(x_i, y_i )$ where $x_1 = 0$ and $y_1 = 0.$         
The points $(x_i, y_i )$ for $i = 2, . . . , n$ are generated according to the following probabilistic relationship:
    
With $1\%$ probability :                                                                                                             
$\qquad x_i = 0,$                                                                                                                   
$\qquad y_i = 0.16y_i−1;$   
With $7\%$ probability:                                                                                                            
$\qquad x_i = 0.2x_i−1 − 0.26y_i−1,  $                                                                                                          
$\qquad y_i = 0.23x_i−1 + 0.22y_i−1 + 1.6;      $                                                                                                      
With $7\%$ probability:                                                                                                            
$\qquad x_i =−0.15x_i−1 +0.28y_i−1,  $                                                                                                          
$\qquad y_i = 0.26x_i−1 + 0.24y_i−1 + 0.44;  $                                                                                                          
With $85\%$ probability:                                                                                                            
$\qquad x_i = 0.85x_i−1 + 0.04y_i−1,   $                                                                                                                                                                                                                     
$\qquad y_i =−0.04x_i−1 +0.85y_i−1 +1.6.  $                                                                                                          
Write a function `my_fern(n)` that generates the points $(x_i, y_i )$ for $i = 1, . . . , n$ and plots them using blue dots.  
 Also use `plt.axis("equal")` and `plt.axis("off")` to make the plot more attractive.

```python
def my_fern(n):
    # Initialize the lists to store the x and y coordinates
    x = [0]
    y = [0]

    # Generate n points based on the probabilistic relationship
    for _ in range(n):
        r = np.random.random()

        if r < 0.01:
            xi = 0
            yi = 0.16 * y[-1]
        elif r < 0.08:
            xi = 0.2 * x[-1] - 0.26 * y[-1]
            yi = 0.23 * x[-1] + 0.22 * y[-1] + 1.6
        elif r < 0.15:
            xi = -0.15 * x[-1] + 0.28 * y[-1]
            yi = 0.26 * x[-1] + 0.24 * y[-1] + 0.44
        else:
            xi = 0.85 * x[-1] + 0.04 * y[-1]
            yi = -0.04 * x[-1] + 0.85 * y[-1] + 1.6

        x.append(xi)
        y.append(yi)

    # Plot the points using blue dots
    plt.plot(x, y, 'b,')
    plt.title('my_fern with {} iterations'.format(n))
    plt.axis("equal")
    plt.axis("off")
    plt.show()
```

Q13. Write a function `my_surface_plotter(f, x, y, option)` where f is a function object f(x,y).                              
The function my_surface_plotter should produce a 3D surface plot of f(x,y) using plot_surface
if the option used is the string "surface".                                                                                 
It should produce a contour plot of f(x,y) if the option used is the string "contour".            
Assume that x and y are one-dimensional arrays or lists.                                                                                               
Remember to give the plot a title and label your axes.

```python
def my_surface_plotter(f, x, y, option) :

    x1 = np.arange(-1, 1.1, 0.2)
    y1 = np.arange(-2, 2.1, 0.2)
    X, Y = np.meshgrid(x1, y1)
    Z = f(X,Y)
    if option=="surface":
        fig = plt.figure(figsize = (12,10))
        ax = plt.axes(projection="3d")
        ax.plot_surface(X, Y, Z, cmap = plt.cm.cividis)
        ax.set_title("Surface plot")
        ax.set_xlabel("x", labelpad=20)
        ax.set_ylabel("y", labelpad=20)
        ax.set_zlabel("z", labelpad=20)
        plt.show()
    elif option=="contour":
        fig, ax = plt.subplots(figsize = (12,10))
        ax.contour(X, Y, Z,  cmap=plt.cm.coolwarm)
        ax.set_title("Contour plot")
        ax.set_xlabel("x", labelpad=20)
        ax.set_ylabel("y", labelpad=20)
        plt.show()
    else :
        print ('stop')
    return None

f = lambda x, y: np.cos(y)*np.sin(np.exp(x))
x = np.linspace(-1, 1, 20)
y = np.linspace(-2, 2, 40)

my_surface_plotter(f, x,y, "surface")
my_surface_plotter(f, x,y, "contour")
```

Q15. We can make maps in cartopy using the online program Web Map Tile Service (WMTS) online.                                       
Plot an Earth-night map shown below for the main part of North America, with the extent latitude from 19.50139 to 64.85694 and longitude from −128.75583 to −68.01197. Hint: Check out the gallery on cartopy website.


```python
import datetime
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.feature.nightshade import Nightshade

plt.figure(figsize = (20, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.set_extent([-128.75583, -68.01197, 19.50139, 64.85694])
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.STATES, linestyle=":")
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.LAKES, alpha=0.5)
ax.add_feature(cfeature.RIVERS)
ax.add_feature(cfeature.COASTLINE)
date = datetime.datetime(2022, 11, 30, 11)
ax.add_feature(Nightshade(date, alpha=0.2))
plt.show()
```
