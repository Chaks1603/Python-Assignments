```python
def my_tip_calc(bill, party):
    if party < 6:
        tip_percent = 0.15
    elif party < 8:
        tip_percent = 0.18
    elif party < 11:
        tip_percent = 0.20
    else:
        tip_percent = 0.25
    
    tips = bill * tip_percent
    return tips
```

```python
def my_mult_operation(a, b, operation):
    if operation == "plus":
        out = np.add(a, b)
    elif operation == "minus":
        out = np.subtract(a, b)
    elif operation == "mult":
        out = np.multiply(a, b)
    elif operation == "div":
        out = np.divide(a, b)
    elif operation == "pow":
        out = np.power(a, b)
    else:
        raise ValueError("Invalid operation. Please choose 'plus', 'minus', 'mult', 'div', or 'pow'.")
    return out
```

```python
def my_inside_triangle(x, y):
    # Triangle vertices
    v0 = (0, 0)
    v1 = (1, 0)
    v2 = (0, 1)

    # Compute vectors
    v0v1 = (v1[0] - v0[0], v1[1] - v0[1])
    v0v2 = (v2[0] - v0[0], v2[1] - v0[1])
    v0p = (x - v0[0], y - v0[1])
    # Compute dot products
    dot00 = v0v0 = v0v0 = v0v0 = v0v1[0] * v0v1[0] + v0v1[1] * v0v1[1]
    dot01 = v0v1[0] * v0v2[0] + v0v1[1] * v0v2[1]
    dot02 = v0v1[0] * v0p[0] + v0v1[1] * v0p[1]
    dot11 = v0v2[0] * v0v2[0] + v0v2[1] * v0v2[1]
    dot12 = v0v2[0] * v0p[0] + v0v2[1] * v0p[1]
    # Compute barycentric coordinates
    inv_denom = 1 / (dot00 * dot11 - dot01 * dot01)
    u = (dot11 * dot02 - dot01 * dot12) * inv_denom
    v = (dot00 * dot12 - dot01 * dot02) * inv_denom
    # Check position
    if u >= 0 and v >= 0 and u + v < 1:
        position = "inside"
    elif (u == 0 or v == 0 or u + v == 1.0) and (0 <= u <= 1.0 or 0 <= v <= 1.0 or 0 <= u + v <= 1.0):
        position = "border"
    else:
        position = "outside"
    return position
```

```python
my_inside_triangle(0.25,0.25)
```

```python
my_inside_triangle(5,5)
```

```python
def my_make_size10(x):
    if len(x) < 10:
        x=np.append(x,x[-1]+1)
        size10 = list(x[:]) + [0] * (10 - len(x))
        return size10
    else:
        size10 = list(x[:10])
        return size10
    return size10
```

```python
def my_letter_grader(percent):
    if percent > 97:         grade = "A+"
    elif percent > 93:       grade = "A"
    elif percent > 90:       grade = "A-"
    elif percent > 87:       grade = "B+"
    elif percent > 83:       grade = "B"
    elif percent > 80:       grade = "B-"
    elif percent > 77:       grade = "C+"
    elif percent > 73:       grade = "C"
    elif percent > 70:       grade = "C-"
    elif percent > 67:       grade = "D+"
    elif percent > 63:       grade = "D"
    elif percent > 60:       grade = "D-"
    else:                    grade = "F"
    return grade
```

```python
my_letter_grader(97)
```

```python
my_letter_grader(84)
```

```python
def my_nuke_alarm(s1, s2, s3):
    if abs(s1 - s2) > 10 or abs(s1 - s3) > 10 or abs(s2 - s3) > 10:
        response = "alarm!"
    else:
        response = "normal"
    return response
```

```python
def my_n_roots(a, b, c):
    import cmath
    discriminant = b**2 - 4*a*c
    
    if discriminant > 0:
        n_roots = 2
        r1 = (-b + cmath.sqrt(discriminant)) / (2*a)
        r2 = (-b - cmath.sqrt(discriminant)) / (2*a)
        r = [r1, r2]
    elif discriminant == 0:
        n_roots = 1
        r = [-b / (2*a)]
    else:
        n_roots = -2
        r1 = (-b + cmath.sqrt(discriminant)) / (2*a)
        r2 = (-b - cmath.sqrt(discriminant)) / (2*a)
        r = [r1, r2]
    return n_roots, r
```

```python
n_roots, r = my_n_roots(1,0,-9)
print('n_roots:',n_roots, '\nr:', r)
```

```python
n_roots, r = my_n_roots(3,4,5)
print('n_roots:',n_roots, '\nr:', r)
```

```python
def my_split_function(f, g, a, b, x):
    if x <= a:
        return f(x)
    elif x >= b:
        return g(x)
    else:
        return 0
```

```python
my_split_function(np.exp, np.sin, 2, 4, 1)
```

```python
my_split_function(np.exp, np.sin, 2, 4, 3)
```
