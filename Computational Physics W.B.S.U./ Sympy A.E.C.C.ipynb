$\text{Q2.  Fibonacci Polynomials are defined as :}$

$\qquad F_1(x) = x,\quad F_2(x) = x ,\quad F_n(x) = xF_{n-1} + F_{n-2}(x),\quad n>2 $

$\qquad\text{Use the function Fibbonacci(n,x) to generate the library function.}$

$\text{Q3. Use the function besselj(n,x) to generate }J_n(x).$

$\text{Superpose }J_1(x),J_2(x),J_3(x)\text{ using different colours.}$

$\text{Bessel functions of the first kind of integer orders : }$

$\qquad\qquad\qquad\qquad\qquad\qquad \boxed{\displaystyle{ J_{p} (x) = \sum_{n=0}^\infty \frac{x^{2n+p}{(-1)^n}}{{2^{2n+p}}{n!}\Gamma{(p+n+1)}}}}$

```python
from sympy import*
from sympy.plotting import plot
x,n, p=var('x,n, p')
def besselj(p,x):
    return summation(((-1)**n*x**(2*n+p))/(factorial(n)*gamma(n+p+1)*2**(2*n+p)),[n,0,oo])
st="J_{p}(x)"
p1=plot(besselj(0,x),(x,-20,20),line_color='b',title=' $'+st+ '$',show=False)    
p2=plot(besselj(1,x),(x,-20,20),line_color='g',show=False)  
p3=plot(besselj(2,x),(x,-20,20),line_color='r',show=False)
p4=plot(besselj(3,x),(x,-20,20),line_color='c',show=False)
p1.extend(p2)
p1.extend(p3)
p1.extend(p4)
p1.show()
```

$\text{Q6. Calculate the matrices }J_x,J_y,J_z\text{ for an arbitray }j.$

$\text{Use the formula for matrix elements of }J_+\ \& \ J_-\ \text{ and generate the matrix.}$ 

$\text{Find out the eigenvalues and eigenvectors of the matrices.}$

## $$\boxed{J_{+}=J_x+iJ_y }$$

## $$\boxed{J_{-}=J_x-iJ_y }$$

$$\boxed{J_x= \frac{1}{2}(J_{+}+J_{-}) }$$

$$\boxed{J_y= \frac{1}{2i}(J_{+}-J_{-}) }$$

$$\ \quad\qquad\qquad\boxed{J_z=i[J_x,J_y]=i(J_xJ_y-J_yJ_x)}$$

```python
def J_plus_entry(ii, jj, j):
    m1 = j - ii
    m2 = j - jj
    if (m1-m2 == 1):   
        return sqrt(Rational(j*(j+1)-m2*(m2+1)))
    else:              
        return 0
    
def J_minus_entry(ii, jj, j):
    m1 = j - ii
    m2 = j - jj
    if (m2-m1 == 1):   
        return sqrt(Rational(j*(j+1)-m2*(m2-1)))
    else:              
        return 0
    
def J_z_entry(ii, jj, j):
    m1 = j - ii
    m2 = j - jj
    if (m1 == m2):     
        return Rational(m1)
    else:             
        return 0   
    
def compute_matrix(j, entry):
    dim = int(2*j+1)
    entries = []
    for ii in range(dim):
        row = []
        for jj in range(dim):
            row.append(entry(ii, jj, j))
        entries.append(row)
    return entries 

def compute_Js(j):
    J_plus = Matrix(compute_matrix(j, J_plus_entry))
    J_minus =  Matrix(compute_matrix(j, J_minus_entry))
    J_x = (J_plus+J_minus)/2
    J_y = (J_plus-J_minus)/2j
    J_z = Matrix(compute_matrix(j, J_z_entry))
    print('total angular momentum quantum number for the system , i.e. :-')
    display(Math(r'j = '+latex(Rational(j))))
    display(Math(r'J_x = '+latex(J_x)),
            Math(r'Eigenvals : J_x = '+latex(J_x.eigenvects())))
    display(Math(r'J_y = '+latex(J_y)),
            Math(r'Eigenvals : J_y = '+latex(J_y.eigenvects())))
    display(Math(r'J_z = '+latex(J_z)),
            Math(r'Eigenvals : J_z = '+latex(J_z.eigenvects())))    
    return j

for j in [1/2,3/2,5/2]:
    compute_Js(j)
    print('\n\n')
```

$$\boxed{\displaystyle{[\sigma_x,\sigma_y] = 2i\sigma_z} } $$

$$\displaystyle\boxed{{ \left\{ \sigma_x,\sigma_y\right\} = 0}} $$

$ \text{Anti-Commutator rules : } \boxed{\displaystyle{\left\{\sigma_x,\sigma_y\right\} = \sigma_x \sigma_y + \sigma_y \sigma_x }} $

Q1. $ \displaystyle{\frac{d x}{d t} + 5xt =0 \qquad x(0)=\pi} $

```python
t=symbols('t',real=True)
x=Function('x')(t)

dxdt=x.diff(t)
eq1=Eq(dxdt,-5*x*t)
display(eq1)

s11=dsolve(eq1)
print('\n general soln. :')
display(s11)

s12=dsolve(eq1, ics={x.subs(t,0):pi})
print('\n particular soln. :\n')
display(s12)
plot(s12.rhs,(t,-2,2))
```

### $ \qquad\qquad c) \text{plot (for t = 0 to 10) x,y as functions of t also the x-y phase track} $
### Q1. $ \displaystyle{\frac{dx}{dt}=x-2y \qquad \frac{dy}{dt}=3x+2y \qquad : x(0)=4,y(0)=-3.9}$      

```python
t=symbols('t',real=True)
x=Function('x')(t)
y=Function('y')(t)
dxdt=x.diff(t)
dydt=y.diff(t)

eq11,eq12=Eq(dxdt,x-2*y),Eq(dydt,3*x+2*y)
print('\n given equation :\n')
display(eq11,eq12)

s1=dsolve([eq11,eq12])
print('\n general soln. :\n')
display(*s1)

s2=dsolve([eq11,eq12], ics={x.subs(t,0):4, y.subs(t,0):-3.9})
print('\n particular soln. :\n')
display(*s2)

S01=plot((s2[0].rhs),(t,0,10),legend=True,label=('x'),
         show=False,title='plot of X,Y as function of t')
S02=plot((s2[1].rhs),(t,0,10),legend=True,label=('y'),show=False)
S01.extend(S02)
S1=plot_parametric((s2[0].rhs,s2[1].rhs),(t,0,10),show=False,
                   axis=True,title='the x-y phase track',
                   xlabel='x',ylabel='y')
PlotGrid(1,2, S01, S1, size=(16,8))
```
