#### 4. Write a program which inputs (x,a,c) and calculates confluenet hypergeometric function,$F(a; c; x)$. 
      Make default values of a and b to be 1 and 2.
 
  The confluent hypergeometric function of the first kind,also known as Kummerâ€™s function and sometimes denoted by M(a,b,z).
#### $$\displaystyle{{}^{}{\phantom{1}_1}\text{F}_1(a,b,z)=\sum_{k=0}^\infty \frac{(a)_k}{(b)_k} \frac{z^k}{k!}}$$

```python
def f1(x,a,c):
    sumi=1
    t=1
    n=1
    while(abs(t)>0.000000000001):
        t=t*(a/c)*(x/n)
        sumi=sumi+t
        n=n+1
        a=a+1
        c=c+1
    return sumi

A=[i/10 for i in range (0,101)]
B=[f1(x,1,2) for x in A]
C=[f1(x,2,3) for x in A]
D=[f1(x,3,4) for x in A]
plt.figure(figsize=(18,8))
plt.plot(A,B,'-')
plt.plot(A,C,'--')
plt.plot(A,D,'--')
plt.legend(['exact','approx1','approx2'],fontsize=20)
plt.show()
```

### 5. Write a function which returns the mean, median and mode of an array. 
    You may use python ufunc and methods like sort(), count() etc.

##### Mean : $\displaystyle{{\bar{x}}=\frac{x_1 + x_2 + ... + x_n}{n}=\sum_{i=1}^n \frac{x_i}{n}}$
##### Median = Size of $\frac{(n+1)}{2}$th item for odd term of n and Median = average of $\frac{n}{2}$th & $\frac{(n+2)}{2}$th item for even term of n.
 
##### Mode is that value in a series of observation which occurs with greatest frequency.

```python
def mmm(n_num):
    n = len(n_num)
    n_num.sort()
    #mean
    print("Mean is: " + str(sum(n_num) / n))
    #median
    if n % 2 == 0:
        median1 = n_num[n//2]
        median2 = n_num[n//2 - 1]
        median = (median1 + median2)/2
    else:
        median = n_num[n//2]
    print("Median is: " + str(median))
    #mode
    L1=[]
    i = 0
    while i < len(n_num) :
        L1.append(n_num.count(n_num[i]))
        i += 1
    d1 = dict(zip(n_num, L1))
    d2=[k for (k,v) in d1.items() if v == max(L1)]
    print("Mode is :" + str(d2))
    return(mmm)

s=[1, 2, 3, 2, 4, 2, 4, 8, 5, 7, 9]
mmm(s)
```

##### Find the histogram of the array Y. Find mean, median from the above histogram. Compare them with direct statistics calculation.  

```python
def Histogram(y,n):
    mx,mi=max(y),min(y)
    mx=mx+mx*0.00001
    h=(mx-mi)/n
    p=[(mi+i*h) for i in range(n+1)]
    pr=[0 for i in range(n)]
    
    for i in y:
        k=int((i-mi)/h)
        pr[k]+=1  
    plt.bar([(i) for i in range(n)], pr)
    plt.title('For n=%i'%n)
    plt.show()
    
    r=list(zip(p,p[1:]))
    mn=sum([pr[i]*(r[i][0]+r[i][1])/2 for i in range(len(pr))])/sum(pr)
    print('calculated mean :',mn,'\tactual mean :',np.mean(y))

    c,l=0,0
    rp=sum(pr)/2
    while c<rp:
        if c+pr[l]>rp:
            break
        c+=pr[l]
        l+=1
    mdn=r[l][0]+((rp-c)*h)/pr[l]
    print('calculated median :',mdn,'\tactual median :',np.median(y))
    return None
```

### Q1. Use d1.txt to fit with the function $ y = mx + c $

## $$m = \frac{N\sum x_i y_i - \sum x_i \sum y_i}{N\sum x_i^2 - (\sum x_i)^2} \qquad c = \frac{\sum y_i \sum x_i^2 - \sum x_i \sum x_i y_i}{N\sum x_i^2 - (\sum x_i)^2} $$

```python
def fit(x,y):
    n = len(x)
    sx = sum(x)
    sx2 = sum(x**2)
    sy = sum(y)
    sxy = sum(x*y)
    b = (n*sx2-sx**2)
    m = (n*sxy-sx*sy)/b
    c = (sy*sx2-sx*sxy)/b
    return m,c
```

```python
A1=np.loadtxt('d1.txt')
x1=A1[:,0]
y1=A1[:,1]
plt.plot(x1,y1,'b8')

m1,c1=fit(x1,y1)
Y1=m1*x1+c1
print('m :', m1 ,'\t\t c :',c1)
plt.plot(x1,Y1, 'r',linewidth=4)
plt.xlabel(r'$ x \rightarrow $')
plt.ylabel(r'$ y \rightarrow $')
plt.suptitle(r'$y=mx+c$',y=1)
plt.title('m : %0.4f  and c : %0.4f '%(m1,c1))
#plt.grid(True)
plt.legend(['data plot','best fit'],fontsize=40)
plt.show()
```
