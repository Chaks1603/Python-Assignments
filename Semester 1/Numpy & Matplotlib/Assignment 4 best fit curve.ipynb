### 1. Use d1.txt to fit with the function $ y = mx + c $

## $m = \frac{N\sum x_i y_i - \sum x_i \sum y_i}{N\sum x_i^2 - (\sum x_i)^2} \qquad c = \frac{\sum y_i \sum x_i^2 - \sum x_i \sum x_i y_i}{N\sum x_i^2 - (\sum x_i)^2} $

```python
import numpy as np
import matplotlib.pyplot as plt
A1=np.loadtxt('d1.txt')
x1=A1[:,0]
y1=A1[:,1]
n1=len(x1)
sx1=sum(x1)
sx2=sum(x1**2)
sy1=sum(y1)
sxy=sum(x1*y1)
b1=(n1*sx2-sx1**2)
m1=(n1*sxy-sx1*sy1)/b1
c1=(sy1*sx2-sx1*sxy)/b1
Y1=m1*x1+c1
print('m :', m1 ,'\t\t c :',c1)
fig, ax = plt.subplots(figsize=(20,12))
plt.plot(x1,y1, 'c8', x1,Y1, 'r',linewidth=4)
plt.xlabel(r'$ x \rightarrow $',fontsize=40)
plt.ylabel(r'$ y \rightarrow $',fontsize=40)
plt.suptitle(r'$y=mx+c$',fontsize=50,y=1)
plt.title('m : %0.4f  | c : %0.4f '%(m1,c1),fontsize=44)
#plt.grid(True)
plt.legend(['data plot','best fit'],fontsize=40)
plt.show()
```

### 2. Use d2.txt to fit with the function $ y = mx^2 + c$

## $m = \frac{N\sum x_i^2 y_i - \sum x_i^2 \sum y_i}{N\sum x_i^4 - (\sum x_i)^4} \qquad c = \frac{\sum y_i \sum x_i^4 - \sum x_i^2 \sum x_i^2 y_i}{N\sum x_i^4 - (\sum x_i)^4} $

```python
import numpy as np
import matplotlib.pyplot as plt
A2=np.loadtxt('d2.txt')
X2=A2[:,0]
x2=X2**2
y2=A2[:,1]
n2=len(X2)
sx1=sum(x2)
sx2=sum(x2**2)
sy1=sum(y2)
sxy=sum(x2*y2)
b2=(n2*sx2-sx1**2)
m2=(n2*sxy-sx1*sy1)/b2
c2=(sy1*sx2-sx1*sxy)/b2
Y2=m2*X2*X2+c2
print('m :', m2 ,'\t\t c :',c2)
fig, ax = plt.subplots(figsize=(20,12))
plt.plot(X2,y2, 'c*', X2,Y2, 'r',linewidth=4)
plt.xlabel(r'$ x \rightarrow $',fontsize=40)
plt.ylabel(r'$ y \rightarrow $',fontsize=40)
plt.suptitle(r'$y=mx^2+c$',fontsize=50,y=1)
plt.title('m : %0.4f  | c : %0.4f '%(m2,c2),fontsize=40)
#plt.grid(True)
plt.legend(['data plot','best fit'],fontsize=40)
plt.show()
```

### 3. Use d3.txt to fit with the function $ y = \frac{a}{x^2} + b$

## $a = \frac{N\sum {\frac{1}{x_i^2}} y_i - \sum {\frac{1}{x_i^2}} \sum y_i}{N\sum {\frac{1}{x_i^4}} - (\sum {\frac{1}{x_i^2}})^2} \qquad b = \frac{\sum y_i \sum {\frac{1}{x_i^4}} - \sum {\frac{1}{x_i^2}} \sum {\frac{1}{x_i^2}} y_i}{N\sum {\frac{1}{x_i^4}} - (\sum {\frac{1}{x_i^2}})^2} $

```python
import numpy as np
import matplotlib.pyplot as plt
A3=np.loadtxt('d3.txt')
X3=A3[:,0]
x3=1/(X3**2)
y3=A3[:,1]
n3=len(x3)
sx1=sum(x3)
sx2=sum(x3**2)
sy1=sum(y3)
sxy=sum(x3*y3)
b3=(n3*sx2-sx1**2)
m3=(n3*sxy-sx1*sy1)/b3
c3=(sy1*sx2-sx1*sxy)/b3
Y3=(m3/(X3*X3))+c3
print('a :', m3 ,'\t\t b :',c3)
fig, ax = plt.subplots(figsize=(20,12))
plt.plot(X3,y3, 'cd', X3,Y3, 'r',linewidth=4)
plt.xlabel(r'$ x \rightarrow $',fontsize=40)
plt.ylabel(r'$ y \rightarrow $',fontsize=40)
plt.suptitle(r'$y=\frac{a}{x^2}+b$',fontsize=50,y=1)
plt.title('a : %0.4f  | b : %0.4f '%(m3,c3),fontsize=40)
#plt.grid(True)
plt.legend(['data plot','best fit'],fontsize=40)
plt.show()
```

### 4. Use d4.txt to fit with the function $y = a \exp(-\frac{x}{x_0}) $

### $ \log y = -\frac{x}{x_0} + \log a => Y = xm + c $
### $\text{where} \qquad Y=\log y , \qquad m=-\frac{1}{x_0}, \qquad c=\log a =>  a=\exp(c)$
   

## $m = \frac{N\sum x_i y_i - \sum x_i \sum y_i}{N\sum x_i^2 - (\sum x_i)^2} \qquad c = \frac{\sum y_i \sum x_i^2 - \sum x_i \sum x_i y_i}{N\sum x_i^2 - (\sum x_i)^2} $   

```python
import numpy as np
import matplotlib.pyplot as plt
A=np.loadtxt('d4.txt')
X4=A[:,0]
Y4=A[:,1]
B=np.array([i for i in A if i[0]>0 and i[1]>0])
x4=B[:,0]
y4=np.log(B[:,1])
n4=len(x4)
sx1=sum(x4)
sx2=sum(x4**2)
sy1=sum(y4)
sxy=sum(x4*y4)
b4=(n4*sx2-sx1**2)
m4=(n4*sxy-sx1*sy1)/b4
c4=(sy1*sx2-sx1*sxy)/b4
a4=np.exp(c4)
x04=(-1/m4)
Y44=a4*np.exp(-X4/x04)
print('x0 :', x04 ,'\t\t a :',a4)
fig, ax = plt.subplots(figsize=(20,12))
plt.plot(X4,Y4, 'b^', X4,Y44,'g',linewidth=4)
plt.xlabel(r'$ x \rightarrow $',fontsize=40)
plt.ylabel(r'$ y \rightarrow $',fontsize=40)
plt.title('x0 : %0.4f  | a : %0.4f '%(x04,a4),fontsize=40)
plt.suptitle(r'$y=a.e^{(-\frac{x}{x_0})}$',fontsize=50,y=1)
#plt.grid(True)
plt.legend(['data plot','best fit'],fontsize=40)
plt.show()
```

### 5. Use d5.txt to fit with the function $y = ax^k$

### $\log y =  k\log x + \log a => Y = mX + c $
### $\text{where} \qquad Y = \log y ,\qquad m=k ,\qquad X = \log x ,\qquad c = \log a=>a=\exp(c)$

## $m = \frac{N\sum x_i y_i - \sum x_i \sum y_i}{N\sum x_i^2 - (\sum x_i)^2} \qquad c = \frac{\sum y_i \sum x_i^2 - \sum x_i \sum x_i y_i}{N\sum x_i^2 - (\sum x_i)^2} $

```python
import numpy as np
import matplotlib.pyplot as plt
A5=np.loadtxt('d5.txt')
X5=A5[:,0]
Y5=A5[:,1]
x5=np.log(A5[16:,0])
y5=np.log(A5[16:,1])
n5=len(x5)
sx1=sum(x5)
sx2=sum(x5**2)
sy1=sum(y5)
sxy=sum(x5*y5)
b5=(n5*sx2-sx1**2)
k5=(n5*sxy-sx1*sy1)/b5
c5=(sy1*sx2-sx1*sxy)/b5
a5=np.exp(c5)
Y55=a5*(X5**k5)
print('k :', k5 ,'\t\t a :',a5)
fig, ax = plt.subplots(figsize=(20,12))
plt.plot(X5,Y5, 'bh', X5,Y55, 'r',linewidth=4)
plt.xlabel(r'$x \rightarrow $',fontsize=40)
plt.ylabel(r'$y \rightarrow $',fontsize=40)
plt.title('k : %0.4f  | a : %0.4f '%(k5,a5),fontsize=40)
plt.suptitle(r'$y=ax^k $',fontsize=50,y=1)
#plt.grid(True)
plt.legend(['data plot','best fit'],fontsize=40)
plt.show()
```
