### 1. Use d1.txt to fit with the function $ y = mx + c $

## $m = \frac{N\sum x_i y_i - \sum x_i \sum y_i}{N\sum x_i^2 - (\sum x_i)^2} \qquad c = \frac{\sum y_i \sum x_i^2 - \sum x_i \sum x_i y_i}{N\sum x_i^2 - (\sum x_i)^2} $

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
plt.rcParams["figure.figsize"] = (20,12)
plt.rcParams['font.size'] = 40
plt.rcParams['legend.fontsize'] = 40

def fit(x,y):
    n = len(x)
    sx = sum(x)
    sx2 = sum(x**2)
    sy = sum(y)
    sxy = sum(x*y)
    b = (n*sx2-sx**2)
    m = (n*sxy-sx*sy)/b
    c = (sy*sx2-sx*sxy)/b
    return m,c
```

```python
A1=np.loadtxt('d1.txt')
x1=A1[:,0]
y1=A1[:,1]
plt.plot(x1,y1,'c8')

m1,c1=fit(x1,y1)
Y1=m1*x1+c1
print('m :', m1 ,'\t\t c :',c1)
plt.plot(x1,Y1, 'r',linewidth=4)
plt.xlabel(r'$ x \rightarrow $')
plt.ylabel(r'$ y \rightarrow $')
plt.suptitle(r'$y=mx+c$',y=1)
plt.title('m : %0.4f  and c : %0.4f '%(m1,c1))
#plt.grid(True)
plt.legend(['data plot','best fit'],fontsize=40)
plt.show()
```

### 2. Use d2.txt to fit with the function $ y = mx^2 + c$

## $m = \frac{N\sum x_i^2 y_i - \sum x_i^2 \sum y_i}{N\sum x_i^4 - (\sum x_i)^4} \qquad c = \frac{\sum y_i \sum x_i^4 - \sum x_i^2 \sum x_i^2 y_i}{N\sum x_i^4 - (\sum x_i)^4} $

```python
A2=np.loadtxt('d2.txt')
X2=A2[:,0]
y2=A2[:,1]
plt.plot(X2,y2, 'c*')

x2=X2**2
m2,c2=fit(x2,y2)
Y2=m2*X2*X2+c2
print('m :', m2 ,'\t\t c :',c2)
plt.plot(X2,Y2, 'r',linewidth=4)
plt.xlabel(r'$ x \rightarrow $')
plt.ylabel(r'$ y \rightarrow $')
plt.suptitle(r'$y=mx^2+c$',y=1)
plt.title('m : %0.4f  | c : %0.4f '%(m2,c2))
#plt.grid(True)
plt.legend(['data plot','best fit'])
plt.show()
```

### 3. Use d3.txt to fit with the function $ y = \frac{a}{x^2} + b$

## $a = \frac{N\sum {\frac{1}{x_i^2}} y_i - \sum {\frac{1}{x_i^2}} \sum y_i}{N\sum {\frac{1}{x_i^4}} - (\sum {\frac{1}{x_i^2}})^2} \qquad b = \frac{\sum y_i \sum {\frac{1}{x_i^4}} - \sum {\frac{1}{x_i^2}} \sum {\frac{1}{x_i^2}} y_i}{N\sum {\frac{1}{x_i^4}} - (\sum {\frac{1}{x_i^2}})^2} $

```python
A3=np.loadtxt('d3.txt')
X3=A3[:,0]
y3=A3[:,1]
plt.plot(X3,y3, 'cd')

x3=1/(X3**2)
m3,c3=fit(x3,y3)
Y3=(m3/(X3*X3))+c3
print('a :', m3 ,'\t\t b :',c3)
plt.plot(X3,Y3, 'r',linewidth=4)
plt.xlabel(r'$ x \rightarrow $')
plt.ylabel(r'$ y \rightarrow $')
plt.suptitle(r'$y=\frac{a}{x^2}+b$',y=1)
plt.title('a : %0.4f  | b : %0.4f '%(m3,c3))
#plt.grid(True)
plt.legend(['data plot','best fit'])
plt.show()
```

### 4. Use d4.txt to fit with the function $y = a \exp(-\frac{x}{x_0}) $

### $ \log y = -\frac{x}{x_0} + \log a => Y = xm + c $
### $\text{where} \qquad Y=\log y , \qquad m=-\frac{1}{x_0}, \qquad c=\log a =>  a=\exp(c)$
   

## $m = \frac{N\sum x_i y_i - \sum x_i \sum y_i}{N\sum x_i^2 - (\sum x_i)^2} \qquad c = \frac{\sum y_i \sum x_i^2 - \sum x_i \sum x_i y_i}{N\sum x_i^2 - (\sum x_i)^2} $   

```python
A=np.loadtxt('d4.txt')
X4=A[:,0]
Y4=A[:,1]
plt.plot(X4,Y4, 'b^')

B=np.array([i for i in A if i[0]>0 and i[1]>0])
x4=B[:,0]
y4=np.log(B[:,1])
m4,c4=fit(x4,y4)
a4=np.exp(c4)
x04=(-1/m4)
Y44=a4*np.exp(-X4/x04)
print('x0 :', x04 ,'\t\t a :',a4)
plt.plot(X4,Y44,'g',linewidth=4)
plt.xlabel(r'$ x \rightarrow $')
plt.ylabel(r'$ y \rightarrow $')
plt.title('x0 : %0.4f  | a : %0.4f '%(x04,a4))
plt.suptitle(r'$y=a.e^{(-\frac{x}{x_0})}$',y=1)
#plt.grid(True)
plt.legend(['data plot','best fit'])
plt.show()
```

### 5. Use d5.txt to fit with the function $y = ax^k$

### $\log y =  k\log x + \log a => Y = mX + c $
### $\text{where} \qquad Y = \log y ,\qquad m=k ,\qquad X = \log x ,\qquad c = \log a=>a=\exp(c)$

## $m = \frac{N\sum x_i y_i - \sum x_i \sum y_i}{N\sum x_i^2 - (\sum x_i)^2} \qquad c = \frac{\sum y_i \sum x_i^2 - \sum x_i \sum x_i y_i}{N\sum x_i^2 - (\sum x_i)^2} $

```python
A5=np.loadtxt('d5.txt')
X5=A5[:,0]
Y5=A5[:,1]
plt.plot(X5,Y5, 'bh')

x5=np.log(A5[16:,0])
y5=np.log(A5[16:,1])
m5,c5=fit(x5,y5)
a5=np.exp(c5)
Y55=a5*(X5**m5)
print('k :', m5 ,'\t\t a :',a5)
plt.plot(X5,Y55, 'r',linewidth=4)
plt.xlabel(r'$x \rightarrow $')
plt.ylabel(r'$y \rightarrow $')
plt.title('k : %0.4f  | a : %0.4f '%(m5,a5))
plt.suptitle(r'$y=ax^k $',y=1)
#plt.grid(True)
plt.legend(['data plot','best fit'])
plt.show()
```
