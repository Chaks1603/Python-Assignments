### find the a) general solution, b) particular solution corresponding to the given initial condition
 
         Plot( for t=0 to 10) x,y as functions of t also the x-y phase track
## 1. $ \frac{dx}{dt}=x-2y \qquad \frac{dy}{dt}=3x+2y \qquad x(0)=4,y(0)=-3.9$      

```python
from sympy.plotting import plot,plot_parametric,PlotGrid
from sympy import *
t=symbols('t',real=True)
x=Function('x')(t)
y=Function('y')(t)
dxdt=x.diff(t)
dydt=y.diff(t)
eq11,eq12=Eq(dxdt,x-2*y),Eq(dydt,3*x+2*y)
print('\n given equation :\n')
display(eq11,eq12)
s1=dsolve([eq11,eq12])
print('\n general soln. :\n')
display(*s1)
s2=dsolve([eq11,eq12], ics={x.subs(t,0):4, y.subs(t,0):-3.9})
print('\n particular soln. :\n')
display(*s2)
S01=plot((s2[0].rhs),(t,0,10),legend=True,label=('x'),show=False,title='plot of X,Y as function of t')
S02=plot((s2[1].rhs),(t,0,10),legend=True,label=('y'),show=False)
S01.extend(S02)
S1=plot_parametric((s2[0].rhs,s2[1].rhs),(t,0,10),show=False,axis=True,title='the x-y phase track',xlabel='x',ylabel='y')
PlotGrid(1,2, S01, S1, size=(20,8))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
t=symbols('t',real=True)
x=Function('x')(t)
y=Function('y')(t)
dxdt=x.diff(t)
dydt=y.diff(t)
eq11,eq12=Eq(dxdt,x-2*y),Eq(dydt,3*x+2*y)
display(eq11,eq12)
s1=dsolve([eq11,eq12])
print('\n general soln. :\n')
display(*s1)
s2=dsolve([eq11,eq12], ics={x.subs(t,0):4, y.subs(t,0):-3.9})
print('\n particular soln. :\n')
display(*s2)
T=np.linspace(0,10,1000)
sp1=utilities.lambdify(t,s2[0].rhs,'numpy')
sp2=utilities.lambdify(t,s2[1].rhs,'numpy')
X=sp1(T)
Y=sp2(T)
figure, (ax1, ax2) = plt.subplots(1, 2,figsize=(20,8))
ax1.plot(T,X, T,Y)
ax1.set_title('plot of X,Y as function of t')
ax1.legend(['X','Y'])
ax2.plot(X,Y)
ax2.set_title('the x-y phase track')
plt.show()
```
