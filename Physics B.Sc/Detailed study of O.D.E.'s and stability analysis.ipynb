## Trapezoidal 

$ \begin{align}  x_{n+1} &=x_n + \left( \frac{f(x_n,t_n) + f(x_{n+1},t_{n+1})}{2} \right) \ dt \end{align}$

$ \begin{align} \text{ Let }, &\dot{x}=\alpha x \\ &{\therefore}\quad x_{n+1} = x_n + \left( \frac{f(x_n,t_n) + f(x_{n+1},t_{n+1})}{2} \right) \ dt \\ &{or,} \quad x_{n+1} = x_n + \frac{\Delta t}{2}\left( \alpha x_n + \alpha x_{n+1} \right) \\ &{or,} \quad x_{n+1} - \frac{\Delta t}{2} \alpha x_{n+1} = x_n + \frac{\Delta t}{2} \alpha x_n \\ &{or,} \quad \left(1 - \frac{\Delta t}{2} \alpha\right) x_{n+1} = \left(1 + \frac{\Delta t}{2} \alpha\right) x_n \\  &{or,} \quad  x_{n+1} = \left(\frac{1 + \frac{\Delta t}{2} \alpha }{1 - \frac{\Delta t}{2} \alpha}\right) x_n \\ \textbf{Stability,}  \\ {\therefore} \ &\left| \frac{x_{n+1}}{x_n}\right|<1 \\ {or,} \ &\left| \frac{1 + \frac{\Delta t}{2} \alpha }{1 - \frac{\Delta t}{2} \alpha} \right| < 1 \end{align}$

## Euler's Explicit
$ \begin{align}  x_{n+1}&=x_n + f(x_n,t_n) \ dt  \end{align}$

$ \begin{align} \text{ Let }, \dot{x}&=\alpha x \\ &{\therefore} \quad \frac{x_{n+1} - x_n}{\Delta t} = \alpha x_n \\ &{or}, \quad x_{n+1} - x_n = \alpha \Delta t x_n \\ &{or}, \quad x_{n+1}= x_n + \alpha \Delta t x_n = x_n\left(1 + \alpha \Delta t \right) \\ \textbf{Stability,}  \\ {\therefore} \ &\left| \frac{x_{n+1}}{x_n}\right|<1 \\ {or,} \ &\left| 1 + \alpha \Delta t \right| < 1 \end{align}$

## Predator-Corrector 

$ \begin{align} \text{Predator Step} : x_{n+1}&=x_n + f(x_n,t_n) \ dt \qquad\qquad\qquad\qquad\qquad \ \ \text{Euler's forward method} \\ \text{Corrector Step} : x_{n+1} &=x_n + \left( \frac{f(x_n,t_n) + f(x_{n+1},t_{n+1})}{2} \right) \ dt \qquad\quad \text{Trapezoidal method} \end{align}$


$ \begin{align} \text{ Let }, \dot{x}&=\alpha x \\ &{\therefore} \quad \frac{x_{n+1} - x_n}{\Delta t} = \alpha x_n \\ &{or}, \quad x_{n+1} - x_n = \Delta t \alpha x_n \\ &{or}, \quad x_{n+1}= x_n + \Delta t \alpha x_n = x_n\left(1 + \Delta t \alpha\right) = x^{predator} \\ x_{n+1} &=  x_n + \left( \frac{f(x_n,t_n) + f(x_{n+1}^{predator},t_{n+1})}{2} \right) \ dt \\ &= x_n + \frac{d t}{2}\left( \dot{x}_n + \dot{x}_{n+1}^{predator} \right) \\ &= x_n + \frac{\Delta t}{2}\left( \alpha x_n + \alpha x_{n+1}^{predator} \right)  \\ &= x_n + \frac{\Delta t}{2}\left( \alpha x_n + \alpha x_n\left(1 + \alpha \Delta t\right)  \right)  \\ &= x_n + \frac{\Delta t}{2}\left( \alpha x_n + \alpha x_n + \alpha^2 \Delta t x_n  \right)  \\ &= x_n + \frac{\Delta t}{2}\left(2\alpha x_n + \alpha^2 \Delta t x_n  \right) \\ &= x_n + (\alpha \Delta t) x_n + \frac{(\alpha\Delta t)^2 x_n}{2} \\ &= \left[ 1 + (\alpha \Delta t) + \frac{(\alpha\Delta t)^2 }{2} \right] x_n \\ \textbf{Stability,}  \\ {\therefore} \ &\left| \frac{x_{n+1}}{x_n}\right|<1 \\ {or,} \ &\left| 1 + (\alpha \Delta t) + \frac{(\alpha\Delta t)^2 }{2} \right| < 1 \end{align}$

## RK2

$ \begin{align} x_{n+1} &= x_n + k_2 \\ \text{where} \\ k_1 &= f(x_n, t_n) \ dt \\ k_2 &= f\left(x_n + \frac{k_1}{2}, t_n + \frac{dt}{2}\right) \ dt \end{align}$


$ \begin{align} \text{ Let , } &\dot{x}=\alpha x \\ k_1 &= f(x_n, t_n) \ dt = \alpha \Delta t x_n  \\ k_2 &= f\left( x_n + \frac{k_1}{2}, t_n+\frac{\Delta t}{2}\right) \ dt = \alpha \Delta t \left( x_n + \frac{\alpha \Delta t x_n}{2}\right)= \alpha \Delta t \left( 1 + \frac{\alpha \Delta t }{2}\right) x_n \\ x_{n+1} &= x_n + k_2 = x_n + \alpha \Delta t \left( 1 + \frac{\alpha \Delta t }{2}\right) x_n = \left[ 1 + \alpha \Delta t + \frac{\left(\alpha \Delta t \right)^2}{2}\right] x_n \\ \textbf{Stability,} \\ {\therefore} \ &\left| \frac{x_{n+1}}{x_n}\right|<1 \\ {or,} \ &\left| 1 + \alpha \Delta t + \frac{\left(\alpha \Delta t \right)^2}{2}\right| < 1 \end{align}$

## RK3
$ \begin{align} x_{n+1} &= x_n + \left(\frac{k_1 + 4k_2 + k_3}{6}\right) \\ \text{where} \\ k_1 &= f(x_n, t_n) \ dt \\ k_2 &= f\left(x_n + \frac{k_1}{2}, t_n + \frac{dt}{2}\right) \ dt \\  k_3 &= f\left(x_n - k_1 + 2k_2, t_n + dt \right) \end{align}$


$ \begin{align} \text{ Let, } &\dot{x}=\alpha x  \\ {k_1 } &= f(x_n, t_n) \ dt = \alpha \Delta t x_n \\ {k_2} &= f\left( x_n + \frac{k_1}{2}, t_n+\frac{\Delta t}{2}\right) \ dt \\ &= \alpha \Delta t \left( x_n + \frac{\alpha \Delta t x_n}{2}\right) \\ &= \alpha \Delta t \left( 1 + \frac{\alpha \Delta t }{2}\right) x_n \\ &= \left( \alpha \Delta t + \frac{(\alpha \Delta t)^2 }{2}\right) x_n  \\ {k_3} &= f\left( x_n-k_1+2k_2, t_n+ \Delta t \right) \ dt \\ &= \alpha\Delta t \left( x_n - \alpha \Delta t x_n + 2\alpha \Delta t \left( 1 + \frac{\alpha \Delta t }{2} \right)\right) \\ &= \alpha\Delta t \left( 1 - \alpha \Delta t  + 2\alpha \Delta t \left( 1 + \frac{\alpha \Delta t }{2} \right)\right) x_n \\ &= \left[ \alpha\Delta t  - (\alpha \Delta t)^2  + 2(\alpha \Delta t)^2 + (\alpha \Delta t )^3 \right] x_n \\ &= \left[ \alpha\Delta t + (\alpha \Delta t)^2 + (\alpha \Delta t )^3 \right] x_n \\ {x_{n+1}} &= x_n + \left(\frac{k_1 + 4k_2 + k_3}{6}\right) \\ &= x_n +  \frac{ 1}{6}\left[ \alpha \Delta t x_n + 4 \left( \alpha \Delta t + \frac{(\alpha \Delta t)^2 }{2}\right) x_n + \left[ \alpha\Delta t + (\alpha \Delta t)^2 + (\alpha \Delta t )^3 \right] x_n \right] \\ &= x_n +  \frac{ 1}{6}\left[ \alpha \Delta t  + 4 \left( \alpha \Delta t + \frac{(\alpha \Delta t)^2 }{2}\right) + \left\{ \alpha\Delta t + (\alpha \Delta t)^2 + (\alpha \Delta t )^3 \right\} \right] x_n \\ &= x_n +  \frac{ 1}{6}\left[ \alpha \Delta t  + 4 (\alpha \Delta t) + 2(\alpha \Delta t)^2 + \alpha\Delta t + (\alpha \Delta t)^2 + (\alpha \Delta t )^3  \right] x_n \\ &= x_n +  \frac{ 1}{6}\left[ 6( \alpha \Delta t) + 3(\alpha \Delta t)^2 + (\alpha \Delta t )^3  \right] x_n \\ &= x_n +  \left[ ( \alpha \Delta t) + \frac{(\alpha \Delta t)^2}{2} + \frac{(\alpha \Delta t )^3}{6}  \right] x_n \\ &= \left[ 1 +   ( \alpha \Delta t) + \frac{(\alpha \Delta t)^2}{2} + \frac{(\alpha \Delta t )^3}{6}  \right] x_n  \\ \textbf{Stability,} \\ 
 {\therefore} \ \ &\left| \frac{x_{n+1}}{x_n}\right|<1 \\ {or, }  \ \ &\left| 1 +   ( \alpha \Delta t) + \frac{(\alpha \Delta t)^2}{2} + \frac{(\alpha \Delta t )^3}{6} \right| < 1 \end{align}$
 
 
 ## RK3
$ \begin{align} x_{n+1} &= x_n + \left(\frac{k_1 + 3k_3}{4}\right) \\ \text{where} \\ k_1 &= f(x_n, t_n) \ dt \\ k_2 &= f\left(x_n + \frac{k_1}{3}, t_n + \frac{dt}{3}\right) \ dt \\  k_3 &= f\left(x_n + \frac{2k_2}{3}, t_n + \frac{2 \ dt}{3} \right) \end{align}$


$ \begin{align} \text{ Let , } &\dot{x}=\alpha x \\ {k_1 } &= f(x_n, t_n) \ dt = \alpha \Delta t x_n \\ {k_2} &= f\left( x_n + \frac{k_1}{3}, t_n+\frac{d t}{3}\right) \ dt \\ &= \alpha \Delta t \left( x_n + \frac{\alpha \Delta t x_n}{3}\right) \\ &= \alpha \Delta t \left( 1 + \frac{\alpha \Delta t }{3}\right) x_n \\ &= \left( \alpha \Delta t + \frac{(\alpha \Delta t)^2 }{3}\right) x_n \\ {k_3} &= f\left(x_n + \frac{2}{3}k_2, t_n + \frac{2}{3}dt \right) \\ &= \alpha \Delta t \left(x_n + \frac{2}{3}\left( \alpha \Delta t + \frac{(\alpha \Delta t)^2 }{3}\right) x_n \right) \\ &= \alpha \Delta t \left(1 + \frac{2}{3}\left( \alpha \Delta t + \frac{(\alpha \Delta t)^2 }{3}\right)\right) x_n  \\ &= \left(\alpha \Delta t + \frac{2}{3}(\alpha \Delta t)^2 + \frac{2}{9}(\alpha \Delta t)^3  \right)x_n \\ x_{n+1} &= x_n + \left(\frac{k_1 + 3k_3}{4}\right) \\  &= x_n + \frac{1}{4}\left[ \alpha \Delta t x_n + 3\left(\alpha \Delta t + \frac{2}{3}(\alpha \Delta t)^2 + \frac{2}{9}(\alpha \Delta t)^3  \right)x_n \right] \\  &= x_n + \frac{1}{4}\left[ (\alpha \Delta t)  + 3(\alpha \Delta t) + 2(\alpha \Delta t)^2 + \frac{2}{3}(\alpha \Delta t)^3  \right] x_n \\  &= x_n + \frac{1}{4}\left[ 4(\alpha \Delta t) + 2(\alpha \Delta t)^2 + \frac{2}{3}(\alpha \Delta t)^3  \right] x_n  \\  &= x_n + \left[ (\alpha \Delta t) + \frac{1}{2}(\alpha \Delta t)^2 + \frac{1}{6}(\alpha \Delta t)^3  \right] x_n \\    &= \left[ 1 + (\alpha \Delta t) + \frac{1}{2}(\alpha \Delta t)^2 + \frac{1}{6}(\alpha \Delta t)^3  \right] x_n \\ \textbf{Stability,} \\ {\therefore} \qquad &\left| \frac{x_{n+1}}{x_n}\right|<1 \\ {or,} \ \ &\left| 1 + (\alpha \Delta t) + \frac{1}{2}(\alpha \Delta t)^2 + \frac{1}{6}(\alpha \Delta t)^3  \right| < 1 \end{align}$

## RK4
$ \begin{align} x_{n+1} &= x_n + \left(\frac{k_1 + 2k_2 + 2k_3 + k_4}{6}\right) \\ \text{where} \\ k_1 &= f(x_n, t_n) \ dt \\ k_2 &= f\left(x_n + \frac{k_1}{2}, t_n + \frac{dt}{2}\right) \ dt \\ k_3 &= f\left(x_n + \frac{k_2}{2}, t_n + \frac{dt}{2}\right) \ dt \\ k_4 &= f(x_n + k_3, t_n + dt) \ dt \end{align}$

x_n + \frac{\alpha \Delta t }{2}x_n \right)= \alpha \Delta t \left( 1 + \frac{\alpha \Delta t }{2}\right) x_n \\ &= \left( \alpha \Delta t + \frac{(\alpha \Delta t)^2 }{2}\right) x_n \\ {k_3} &= f\left( x_n + \frac{k_3}{2}, t_n+\frac{d t}{2}\right) \ dt \\ &= \alpha \Delta t \left( x_n +  \frac{\alpha \Delta t}{2} \left( 1 + \frac{\alpha \Delta t }{2}\right) x_n \right) \\ &= \alpha \Delta t \left( 1 +  \frac{\alpha \Delta t}{2} + \left( \frac{\alpha \Delta t }{2}\right)^2\right)  x_n \\ &= \left( \alpha \Delta t +  \frac{(\alpha \Delta t)^2}{2} + \frac{(\alpha \Delta t)^3}{4} \right)  x_n \\ {k_4} &= f(x_n+k_3, t_n+dt) \ dt \\ &= \alpha \Delta t \left( x_n +   \left( \alpha \Delta t +  \frac{(\alpha \Delta t)^2}{2} + \frac{(\alpha \Delta t)^3}{4} \right)  x_n \right) \\ &=  \left( 1+(\alpha \Delta t)^2 +  \frac{(\alpha \Delta t)^3}{2} + \frac{(\alpha \Delta t)^4}{4} \right)  x_n \\ x_{n+1} &= x_n + \left(\frac{k_1 + 2k_2 + 2k_3 + k_4}{6}\right) \\ &= x_n + \frac{1}{6}\left[ \alpha \Delta t x_n + 2\left( \alpha \Delta t + \frac{(\alpha \Delta t)^2 }{2}\right) x_n + 2\left( \alpha \Delta t +  \frac{(\alpha \Delta t)^2}{2} + \frac{(\alpha \Delta t)^3}{4} \right)  x_n  + \left( \alpha \Delta t + (\alpha \Delta t)^2 +  \frac{(\alpha \Delta t)^3}{2} + \frac{(\alpha \Delta t)^4}{4} \right)  x_n \right] \\ &=  x_n + \frac{1}{6}\left[ (\alpha \Delta t) + 2(\alpha \Delta t) + (\alpha \Delta t)^2 + 2(\alpha \Delta t) +  (\alpha \Delta t)^2 + \frac{(\alpha \Delta t)^3}{2} + \alpha \Delta t + (\alpha \Delta t)^2 +  \frac{(\alpha \Delta t)^3}{2} + \frac{(\alpha \Delta t)^4}{4}   \right]x_n \\ &=  x_n + \frac{1}{6}\left[ 6(\alpha \Delta t) + 3(\alpha \Delta t)^2 + (\alpha \Delta t)^3 + \frac{(\alpha \Delta t)^4}{4} \right]x_n  \\ &=  x_n + \left[ (\alpha \Delta t) + \frac{(\alpha \Delta t)^2}{2} + \frac{(\alpha \Delta t)^3}{6} + \frac{(\alpha \Delta t)^4}{24} \right]x_n   \\ &=  \left[ 1 +  (\alpha \Delta t) + \frac{(\alpha \Delta t)^2}{2} + \frac{(\alpha \Delta t)^2}{2} + \frac{(\alpha \Delta t)^3}{6} + \frac{(\alpha \Delta t)^4}{24} \right| < 1 \end{align}$
