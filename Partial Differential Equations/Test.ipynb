$\qquad$ Peak FLOPS/s = N * M * F                                                                                                                               
where:
    * N = Number of cores (20 cores)
    * M = Number of floating-point operations per clock cycle per core (20 FLOPs)
    * F = Average frequency of the processor in Hz (3 GHz)                                                                                                                               

Let's calculate the peak FLOPS/s:

```python
# Given data
num_cores = 20
flops_per_clock_cycle_per_core = 20
clock_speed_hz = 3e9  # 3 GHz in Hz

# Calculate peak FLOPS/s
peak_flops_per_second = num_cores * flops_per_clock_cycle_per_core * clock_speed_hz

print(f"The peak FLOPS/s of the processor is approximately {peak_flops_per_second:.2e} FLOPS/s.")
```

```python
def decimal_to_twos_complement(decimal_number, num_bits):
    # Calculate the maximum positive value for the given number of bits
    max_positive_value = 2**(num_bits - 1) - 1

    # Check if the decimal number is within the representable range
    if decimal_number < -max_positive_value or decimal_number > max_positive_value:
        raise ValueError(f"Decimal number {decimal_number} out of representable range for {num_bits}-bit 2's complement.")

    # If the number is negative, convert it to its two's complement
    if decimal_number < 0:
        decimal_number += 2**num_bits

    # Convert the decimal number to binary
    binary_representation = bin(decimal_number)[2:].zfill(num_bits)

    return binary_representation

# Example usage:
decimal_number = -3
num_bits = 4
twos_complement_binary = decimal_to_twos_complement(decimal_number, num_bits)
print(twos_complement_binary)
```

2. Write a program to find if the given year is leap or not.

```python
def is_leap_year(year):
    # A year is a leap year if it is divisible by 4
    # But not divisible by 100 unless it is also divisible by 400
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

# Test the program with some examples
years = [2020, 1900, 2000, 2021]

for year in years:
    if is_leap_year(year):
        print(f"{year} is a leap year.")
    else:
        print(f"{year} is not a leap year.")
```

3. A palindrom string is the one which reads same both forward and backward, i.e. ( "lal", "kaak", "01010").              
Write a program to find if given string is palindrome.


```python
def is_palindrome(input_string):
    # Convert the input string to lowercase to ignore case sensitivity
    input_string = input_string.lower()
    
    # Remove any spaces from the input string
    input_string = input_string.replace(" ", "")
    
    # Check if the input string is equal to its reverse
    if input_string == input_string[::-1]:
        return True
    else:
        return False

# Test the program with some examples
strings = ["lal", "kaak", "hello", "01010"]

for string in strings:
    if is_palindrome(string):
        print(f"{string} is a palindrome.")
    else:
        print(f"{string} is not a palindrome.")
```

Q4. Write a code to compute factorial of a given number.


```python
def factorial(n):
    # Base case: 0! and 1! are equal to 1
    if n == 0 or n == 1:
        return 1
    else:
        # Recursive case: factorial(n) = n * factorial(n-1)
        return n * factorial(n - 1)

# Test the program with some examples
num = [5, 0, 10]

for n in num:
    print(f"The factorial of {n} is: {factorial(n)}")
```

Q5. Write a code to compute greatest common divisor (GCD) of two integers a and b using Euclid's algorithm.

```python
def gcd(a, b):
    while b != 0:
        remainder = a % b # Calculate the remainder when a is divided by b
        a = b
        b = remainder
    return a

# Test the program with some examples
n1 = [36,100]
n2 = [48, 72]

for i,j in zip(n1,n2):
    print(f"The GCD of {i} and {j} is: {gcd(i,j)}")
```

Q6. Write a code to find if a number is prime or not.


```python
def is_prime(number):
    if number <= 1:
        return False
    elif number == 2:
        return True
    elif number % 2 == 0:
        return False
    else:
        # Check for factors from 3 to the square root of the number
        for i in range(3, int(number**0.5) + 1, 2):
            if number % i == 0:
                return False
        return True

# Test the program with some examples
num = [7, 12, 23, 49]

for n in num:
    if is_prime(n):
        print(f"{n} is a prime number.")
    else:
        print(f"{n} is not a prime number.")
```

1. Write a function to compute sum of digits of a number.

```python
def sum_of_digits(number):
    # Convert the number to a string to access individual digits
    num_str = str(number)
    
    # Initialize a variable to store the sum of digits
    sum_digits = 0
    
    # Iterate through each character in the string and add its integer value to the sum
    for digit in num_str:
        sum_digits += int(digit)
    
    return sum_digits

# Test the function with some examples
num = [12345, 9876, 0, 9999]

for n in num:
    print(f"The sum of digits of {n} is: {sum_of_digits(n)}")
```

2. Write a function to find the maximum number in an array.

```python
def find_max(arr):
    # Check if the array is empty
    if not arr:
        return None

    # Initialize a variable to store the maximum number
    max_num = arr[0]

    # Iterate through the array to find the maximum number
    for num in arr:
        if num > max_num:
            max_num = num

    return max_num

# Test the function with some examples
arrays = [[1, 5, 8, 3, 7, 2],
          [10, 20, 30, 40, 50],
          [-5, -2, -10, -3]]

for ar in arrays:
    print(f"The maximum number in {ar} is: {find_max(ar)}")
```

2. Make contour, density and surface plots for the electric potential of a line charge with charge density of $\lambda$ .  
Make appropriate scaling for the plot.                                                                                                                               
The electric potential (V) due to a line charge with charge density $𝜆$ at a distance 'r' from the line charge is given by:                                                                                                                           
$\qquad V=\frac{k⋅λ}{r}$                                                                                                                                 
where:                                                                                                                                
    * $V$ is the electric potential.                                                                                                   
    * $k$ is Coulomb's constant $\displaystyle{\bigg(k=\frac{1}{4\pi\epsilon_o}}$ , where $\epsilon_o$ is the vacuum permittivity $\bigg).$                        
    * $𝜆$ is the charge density (charge per unit length) of the line charge.                                                              
    * $r$ is the distance from the line charge.

```python
# Charge density 𝜆 (you can adjust this value)
lambda_ = 2.0

# Coulomb's constant
k = 8.9875e9  # N m^2/C^2

# Create a grid of points for the x and y coordinates
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate the distance from the line charge at each point in the grid
r = np.sqrt(X**2 + Y**2)

# Calculate the electric potential V using the formula
V = k * lambda_ / r

# Plot the contour plot
plt.figure(figsize=(4,3))
plt.contour(X, Y, V, levels=15, cmap='viridis')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Contour Plot of Electric Potential (V) of a Line Charge')
plt.colorbar(label='Electric Potential (V)')
plt.grid()
plt.show()

# Plot the density plot
plt.figure(figsize=(4,3))
plt.imshow(V, extent=[-5, 5, -5, 5], origin='lower', cmap='viridis', aspect='auto')
plt.colorbar(label='Electric Potential (V)')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Density Plot of Electric Potential (V) of a Line Charge')
plt.show()

# Plot the surface plot
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, V, cmap='viridis', rstride=1, cstride=1)
ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')
ax.set_zlabel('Electric Potential (V)')
ax.set_title('Surface Plot of Electric Potential (V) of a Line Charge')
plt.show()
```

4. Using the non-dimenstionalized wave functions of Hydrogen atom,                                                         
compute $\langle r\rangle,\langle r^2 \rangle$ and $\langle E\rangle$ for  $ψ_{100},ψ_{200},ψ_{210},$ and $ψ_{211}$ states of the H atom.

The expectation values 
$\langle r\rangle,\langle r^2 \rangle$ and $\langle E\rangle$ for the $1s (n=1,l=0,m=0), 2s (n=2,l=0,m=0), 2p_x(n=2,l=1,m=0),$ and $2p_y(n=2,l=1,m=1)$ states of the hydrogen atom, we need to use the non-dimensionalized wave functions of the hydrogen atom.

The non-dimensionalized wave functions for these states are as follows:
* $1s$ state $\displaystyle{(n=1,l=0,m=0)\rightarrow ψ_{1,0,0}(r,\theta,\phi)=\frac{1}{\sqrt{\pi}}e^{-r}}$                                              
* $2s$ state $\displaystyle{(n=2,l=0,m=0)\rightarrow ψ_{2,0,0}(r,\theta,\phi)=\frac{1}{\sqrt{32\pi}}(2-r)e^{-r/2}}$                                    
* $2p_x$ state $\displaystyle{(n=2,l=1,m=0)\rightarrow ψ_{2,1,0}(r,\theta,\phi)=\frac{1}{\sqrt{32\pi}}re^{-r/2}\cos(\theta)}$                                              
* $2p_y$ state $\displaystyle{(n=2,l=1,m=1)\rightarrow ψ_{2,1,\pm 1}(r,\theta,\phi)=\frac{1}{\sqrt{32\pi}}re^{-r/2}\sin(\theta)e^{\pm i\phi}}$

To compute the expectation values, we need to integrate the product of the wave function and the corresponding operator over all space.

The radial expectation value $\langle r\rangle$ for a state $\psi_{nlm}$ is given by:                                                        
$$\displaystyle{\langle r\rangle \ =\int_0^\infty r^2 |\psi_{nlm}(r,\theta,\phi)|^2 dr}$$                                                                      
The radial squared expectation value $\langle r^2\rangle$ for a state $\psi_{nlm}$ is given by:                                              
$$\displaystyle{\langle r^2\rangle \ =\int_0^\infty r^4 |\psi_{nlm}(r,\theta,\phi)|^2 dr}$$                                                                   
The energy expectation value $\langle E\rangle$ for a state $\psi_{nlm}$ is given by:                                                        
$$\qquad\qquad\qquad\qquad\qquad\qquad\displaystyle{ \langle E\rangle \ =\int_0^\infty |\psi_{nlm}(r,\theta,\phi)|^2 \bigg(-\frac{1}{2} \frac{d^2}{dr^2} -\frac{1}{r}\frac{d}{dr}+\frac{l(l+1)}{2r^2}-\frac{Z}{r}\bigg)}$$                                                           
Where $Z$ is the atomic number (for hydrogen, $Z=1$).

```python
# Constants
Z = 1  # Atomic number for hydrogen (Z=1)

# Wave functions for the states
def psi_100(r, theta, phi):
    return 1 / np.sqrt(np.pi) * np.exp(-r)

def psi_200(r, theta, phi):
    return 1 / np.sqrt(32 * np.pi) * (2 - r) * np.exp(-r / 2)

def psi_210(r, theta, phi):
    return 1 / np.sqrt(32 * np.pi) * r * np.exp(-r / 2) * np.cos(theta)

def psi_211(r, theta, phi):
    return 1 / np.sqrt(64 * np.pi) * r * np.exp(-r / 2) * np.sin(theta) * np.exp(1j * phi)

# Radial expectation value <r>
def radial_expectation_value(psi):
    integrand = lambda r: r**2 * np.abs(psi(r, 0, 0))**2
    return si.integrate.quad(integrand, 0, np.inf)[0]

# Radial squared expectation value <r^2>
def radial_squared_expectation_value(psi):
    integrand = lambda r: r**4 * np.abs(psi(r, 0, 0))**2
    return si.integrate.quad(integrand, 0, np.inf)[0]

# Numerical second derivative of the wave function
def numerical_second_derivative(psi, r, h=1e-5):
    return (psi(r + h, 0, 0) - 2 * psi(r, 0, 0) + psi(r - h, 0, 0)) / h**2

# Energy expectation value <E>
def energy_expectation_value(psi):
    integrand = lambda r: np.abs(psi(r, 0, 0))**2 * (-1/2 * numerical_second_derivative(psi, r) - Z / r * psi(r, 0, 0))
    return si.integrate.quad(integrand, 1e-8, np.inf)[0]

# Compute the expectation values for each state
states = [(psi_100, "1s"), (psi_200, "2s"), (psi_210, "2p_x"), (psi_211, "2p_y")]

for psi, state_name in states:
    r_expectation = radial_expectation_value(psi)
    r_squared_expectation = radial_squared_expectation_value(psi)
    energy_expectation = energy_expectation_value(psi)

    print(f"State {state_name}:")
    print(f"<r>: {r_expectation}")
    print(f"<r^2>: {r_squared_expectation}")
    print(f"<E>: {energy_expectation}")
    print()
```

###  Lagrange interpolating polynomials 1D

$\displaystyle P_n(x)=\sum_{j=0}^{n-1} L_j(x)y_j \qquad L_j(x)=\prod_{i,i\ne j} \frac{(x-x_i)}{(x_j-x_i)} P_n(x)=\sum_{j=0}^{n-1} L_j(x)y_j  \qquad E_n(x)=f(x)-P_n(x)=\frac{f^{(n)}(\zeta)}{n!}\prod (x-x_i)$

```python
def Lagrange_interpolate(x,x_arr, y_arr):
    n = len(x_arr)
    ans = 0
    for j in range(n):
        den,num=1,1
        for i in range(n):
            if j!=i :
                num *= (x - x_arr[i])
                den *= (x_arr[j] - x_arr[i])
        ans += (num/den)*y_arr[j]
    return ans
```

###  Lagrange interpolating polynomials in 2D

$\displaystyle L_j(x)=\sum_i\prod_{i',i'\ne j'} \frac{(x-x_{i'})}{(x_i-x_{i'})}\tilde{P}(x_i,y)  \qquad \tilde{P}(x_i,y)=\sum_j\prod_{j',j'\ne j} \frac{(y-y_{j'})}{(y_j-y_{j'})}f(x_i,y_j) \qquad P_n(x)=\sum_{i}\sum_{j} L_{i,j}(x,y)y_{i,j} $

```python
def f(x, y):
    return np.exp(x + y)

x_range = [0, 1]
y_range = [0, 1]
x_points = np.linspace(x_range[0], x_range[1], 100)
y_points = np.linspace(y_range[0], y_range[1], 100)
X, Y = np.meshgrid(x_points, y_points)
Z = f(X, Y)
num_points_x = [2, 3, 4]
num_points_y = [2, 3, 4]

fig, axs = plt.subplots(len(num_points_x), len(num_points_y), figsize=(12, 10))
for i, nx in enumerate(num_points_x):
    for j, ny in enumerate(num_points_y):
        x_interp = np.linspace(x_range[0], x_range[1], nx)
        y_interp = np.linspace(y_range[0], y_range[1], ny)

        # Lagrange interpolating polynomials and interpolate point by point
        P = np.zeros_like(X)
        for xi in range(X.shape[0]):
            for yi in range(Y.shape[1]):
                for kx in range(len(x_interp)):
                    for ky in range(len(y_interp)):
                        Lx = np.prod([(X[xi, yi] - x_interp[kx]) / (x_interp[jx] - x_interp[kx])
                                      for jx in range(len(x_interp)) if jx != kx])
                        Ly = np.prod([(Y[xi, yi] - y_interp[ky]) / (y_interp[jy] - y_interp[ky])
                                      for jy in range(len(y_interp)) if jy != ky])
                        P[xi, yi] += Lx * Ly * f(x_interp[kx], y_interp[ky])
        ax = axs[i, j]
        ax.contourf(X, Y, Z, levels=20, cmap='viridis')
        ax.contour(X, Y, P, colors='r', linewidths=1)
        ax.set_title(f'({nx}x{ny}) Interpolation Points')
fig.text(0.5, 0.08, 'x', ha='center')
fig.text(0.08, 0.5, 'y', va='center', rotation='vertical')
plt.tight_layout()
plt.show()
```

## Trapezoidal rule

${\displaystyle \int _{a}^{b}f(x)\,dx\approx (b-a)\cdot {\tfrac {1}{2}}\bigg(f(a)+f(b)\bigg).}$

```python
def trapz(f,a,b,N=50):
    x = np.linspace(a,b,N+1) 
    y = f(x)
    y_right = y[1:] 
    y_left = y[:-1] 
    dx = (b - a)/N
    T = (dx/2) * np.sum(y_right + y_left)
    return T
```

## Simpson's 1/3 rule

${\displaystyle \int _{a}^{b}f(x)\,dx\approx {\frac {b-a}{6}}\left[f(a)+4f\left({\frac {a+b}{2}}\right)+f(b)\right].}$

```python
def simps_1_3_rule(f,a,b,N=50):
    if N % 2 == 1:
        raise ValueError("N must be an even integer.")
    dx = (b-a)/N
    x = np.linspace(a,b,N+1)
    y = f(x)
    S = dx/3 * np.sum(y[0:-1:2] + 4*y[1::2] + y[2::2])
    return S
```

## Simpson's 3/8 rule                                                                              
$\qquad\displaystyle \int _{a}^{b}f(x)\,dx \approx {\frac {3}{8}}h\left[f(a)+3f\left({\frac {2a+b}{3}}\right)+3f\left({\frac {a+2b}{3}}\right)+f(b)\right]$                                                                               
$\qquad\qquad\qquad\quad\displaystyle ={\frac {b-a}{8}}\left[f(a)+3f\left({\frac{2a+b}{3}}\right)+3f\left({\frac {a+2b}{3}}\right)+f(b)\right] $                                                                              
where ${\displaystyle h=\frac{(b-a)}{3}}$ is the step size.                                                                              
The error of this method is                                                                              
$\qquad{\displaystyle -{\frac {3}{80}}h^{5}f^{(4)}(\xi )=-{\frac {(b-a)^{5}}{6480}}f^{(4)}(\xi ),}$                                                                              
where $\xi$  is some number between $a$ and $b$.

```python
def simpsons_3_8_rule(func, a, b, n):
    if n % 3 != 0:
        raise ValueError("Number of intervals must be a multiple of 3.")
    
    h = (b - a) / n
    integral = func(a) + func(b)
    
    for i in range(1, n):
        x = a + i * h
        if i % 3 == 0:
            integral += 2 * func(x)
        else:
            integral += 3 * func(x)
    
    integral *= (3 * h) / 8
    return integral
```

Using a large number of points (8-10 or more) for derivative computations, especially in finite difference methods, can lead to several issues:

Higher Sensitivity to Noise: As you increase the number of points, the finite difference methods become more sensitive to noise in the data.                                                                                                                                     
Small variations in the function values can result in large fluctuations in the computed derivatives,                                        
leading to less accurate results.

Numerical Instability: Finite difference methods involve subtracting or adding small values to the function evaluations.                  
When subtracting two nearly equal values, you can experience a significant loss of precision due to subtractive cancellation,                                                                                                                                      
leading to numerically unstable results.

Round-off Errors: When subtracting or adding small values to a large value, the relative round-off error can become significant.

Higher Computational Cost: Calculating derivatives using a larger number of points requires more function evaluations and computations,

Inaccurate Results Near Discontinuities: Finite difference methods assume a certain degree of smoothness in the function.                  
If you're working with functions that have discontinuities or sharp changes, using more points can lead to inaccurate derivatives.

Higher Order Methods: When you need accurate derivatives, it's often more effective to use higher-order methods

### Euler's Forward Method  or, explicit method
$\qquad\begin{equation*} x^{(n+1)} = x^{(n)} + (Δt) f(x^{(n)},t_{n})   .\end{equation*}$

```python
def euler_forward(f, t_init, t_fin, dt, init_condition):
    n = int((t_fin - t_init) / dt) + 1     # (n-1) divisions
    t = np.linspace(t_init, t_fin, n)
    x = np.zeros(n)
    x[0] = init_condition
    for i in range(n-1):
        x[i+1] = x[i] + f(x[i], t[i])*dt
    return t,x
```

### Simpson's 1/3 Method
$\qquad\begin{equation*} x^{(n+1)} = x^{(n-1)} + (1/3) (Δt)[ f(x^{(n-1)},t_{n-1}) + 4f(x^{(n)},t_{n}) +  f(x^{(n+1)},t_{n+1})]  .\end{equation*}$

```python
def simpsons_1_3(f, t_init, t_fin, dt, init_condition):    
    n = int((t_fin - t_init) / dt) + 1
    t = np.linspace(t_init, t_fin, n)
    x = np.zeros(n)
    x[0] = init_condition
    for i in range(1, n):
        if i == n - 1:  # If it's the last point, use backward Euler method
            x[i] = x[i-1] + dt * f(x[i-1], t[i-1])
        else:
            x[i] = x[i-1] + (dt / 3)*(f(x[i-1], t[i-1]) + 4*f(x[i-1] + (dt/2)*f(x[i-1], t[i-1]), t[i-1] + dt/2) \
                   + f(x[i-1] + dt*f(x[i-1] + (dt/2)*f(x[i-1], t[i-1]), t[i-1] + dt/2), t[i-1] + dt))
    return t, x
```

### Simpson's 3/8 Method

```python
def simpsons_3_8(f, t_init, t_fin, dt, init_condition):    
    n = int((t_fin - t_init) / dt) + 1
    t = np.linspace(t_init, t_fin, n)
    x = np.zeros(n)
    x[0] = init_condition
    for i in range(n - 3):
        k1 = f(x[i], t[i])
        k2 = f(x[i] + (dt / 3) * k1, t[i] + dt / 3)
        k3 = f(x[i] + (2 * dt / 3) * k2, t[i] + 2 * dt / 3)
        k4 = f(x[i] + dt * k1 - 2 * dt * k2 + 3 * dt * k3, t[i] + dt)
        x[i+1] = x[i] + (dt / 8) * (k1 + 3 * k2 + 3 * k3 + k4)
    return [t[:-2], x[:-2]]
```

2. Why do the numerical solutions of ODEs have errors ?                                                                                                                               
$\text{Ans }:$ Numerical solutions of ordinary differential equations (ODEs) have errors due to several factors, both inherent to the nature of numerical methods and due to the limitations of computers.                                                                                                                                
Here are some of the main reasons for the errors in numerical solutions of ODEs:                                                                                                                               
$\bf\text{Discretization Error:}$ Numerical methods convert continuous ODEs into discrete equations by dividing the continuous domain into discrete time steps or grid points.                                                                                                                                
This discretization introduces errors because it approximates the continuous behavior of the system.                                                                                                                               
$\bf\text{Truncation Error:}$ Numerical methods use finite approximations of derivatives, leading to truncation errors. The accuracy of these approximations depends on the chosen step size and the order of the method.                                                                                                                               
$\bf\text{Round-off Error:}$ Computers use finite-precision arithmetic, which means that real numbers are approximated using a finite number of bits. This can lead to round-off errors where calculations involve approximated values, especially in cases of repeated calculations or very large/small numbers.                                                                                                                               
$\bf\text{Stability Issues:}$ Some numerical methods, particularly explicit methods like the Euler method, have stability constraints. If the step size is too large, these methods can lead to numerical instability, causing solutions to diverge from the true solution.                                                                                                                               
$\bf\text{Initial Condition Error:}$ The initial condition used in numerical methods might not be exactly the same as the actual initial condition of the ODE. Even small deviations can lead to errors in the subsequent calculations.                                                                                                                               
$\bf\text{Accumulated Errors:}$ Numerical errors can accumulate over time as the calculation progresses through multiple steps. Small errors in each step can compound, leading to larger discrepancies from the true solution.                                                                                                                               
$\bf\text{Sensitivity to Parameters:}$ Some ODEs are highly sensitive to changes in parameters or initial conditions. Small variations in these values can lead to significantly different behaviors in the solution.                                                                                                                               
$\bf\text{Numerical Method Selection:}$ Different numerical methods have different accuracy and stability properties. Choosing an inappropriate method for a specific problem can lead to larger errors.                                                                                                                               
$\bf\text{Stiffness:}$ Some ODEs exhibit stiffness, where the system's behavior changes rapidly in some regions and slowly in others. Numerical methods can struggle to accurately capture both behaviors.                                                                                                                               
$\bf\text{Nonlinearity:}$ Nonlinear ODEs can be challenging to solve numerically because small changes in the initial conditions or parameters can lead to complex and unexpected behaviors.                                                                                                                               
Due to these factors, it's important to choose appropriate numerical methods, adjust step sizes, and consider the specific characteristics of the problem to achieve accurate numerical solutions. Additionally, error analysis and comparisons with analytical solutions are valuable for understanding the accuracy and reliability of numerical results.

### Euler's Backward Method  or, implicit method
$\qquad\begin{equation*} x^{(n+1)} = x^{(n)} + (Δt) f(x^{(n+1)},t_{n+1})   .\end{equation*}$

```python
def euler_backward(f, t_init, t_fin, dt, init_condition):
    n = int((t_fin - t_init) / dt) + 1
    t_values=np.linspace(t_init, t_fin, n)
    x_values = np.zeros(n)
    x_values[0] = init_condition
    for i in range(1, n):
        x_next = x_values[i - 1] + dt * f(x_values[i - 1], t_values[i - 1])
        for _ in range(10):  # Use Newton-Raphson method
            x_new = x_values[i - 1] + dt * f(x_next, t)
            if abs(x_new - x_next) < 1e-6:
                break
            x_next = x_new
        x_values[i] = x_next 
    return t_values, x_values
```

### Trapezoidal Method      or ,      semi-implicit method
$\qquad\begin{equation*} x^{(n+1)} = x^{(n)} + (0.5)(Δt) [f(x^{(n)},t_{n}) + f(x^{(n+1)},t_{n+1})]  .\end{equation*}$

```python
# Trapezoid method implementation
def trapezoid_method(f, t_init, t_fin, dt, init_condition):
    n = int((t_fin - t_init) / dt) + 1
    t = np.linspace(t_init, t_fin, n)
    x = np.zeros(n)
    x[0] = init_condition
    for i in range(1, n):
        t_i = t[i - 1]
        x_i = x[i - 1]
        x_new = x_i + 0.5 * dt * (f(x_i, t_i) + f( x_i + dt * f(x_i,t_i),t_i + dt))   # update x
        x[i] = x_new
    return t, x
```

###  Predictor-Corrector Scheme  (PC)
$ \text{Predictor step}: x^{(n+1)^*} = x^{(n)}+(\Delta t) f( x^{(n)}, t_n)  $                                                                            
$\text{Corrector step}:  x^{(n+1)} = x^{(n)}+(½) (\Delta t) [f( x^{(n)}, t_n)+f( x^{(n+1)*}, t_{n+1})]$

```python
def PC(f, tinit, tfinal, dt, init_cond):
    n=int((tfinal-tinit)/dt)+1
    t = np.linspace(tinit,tfinal,n)
    x = np.zeros(n)
    x[0] = init_cond
    
    for i in range(1, n):
        x_pred = x[i - 1] + dt * f(x[i - 1], t[i - 1])
        x[i] = x[i - 1] + 0.5 * dt * (f(x[i - 1], t[i - 1]) + f(x_pred, t[i]))
    return t, x
```

### Second-order Runge-Kutta scheme (RK2)

$\text{The predictors for RK2 is} $                                                                            
$\qquad k_1 = (\Delta t) f(x^{(n)},t_n); \quad x^{(n+½)^*} = x^{(n)} + k_1/2 ,$                                                                            
$\text{Now, the corrector for RK2 is } $                                                                            
$\qquad x^{(n+1)} = x^{(n)} + (\Delta t) f(x^{(n+½)^*},t_{n+½}) $

```python
def RK2(f, tinit, tfinal, dt, init_cond):
    n = int((tfinal - tinit) / dt) + 1
    t = np.linspace(tinit, tfinal, n)
    x = np.zeros(n)
    x[0] = init_cond

    for i in range(n - 1):
        k1 = dt * f(x[i], t[i])  
        x[i + 1] = x[i] + dt * f(x[i] + k1/2, t[i] + dt/2)
    return t, x
```

### Third-order Runge-Kutta scheme (RK3)

$\text{The predictors for RK3 are} $                                                                            
$\qquad k_1 = (\Delta t) f(x^{(n)},t_n);  \qquad \qquad \ x^{(n+½)^*} = x^{(n)} + k_1/2 , $                                                                            
$\qquad k_2 = (\Delta t) f(x^{(n+½)^*},t_{n+½}); \qquad x^{(n+½)^{**}} = x^{(n)} -k_1 + 2k_2 , $                                                                            
$\qquad k_3 = (\Delta t) f(x^{(n+1)^{**}},t_{n+1})$                                                                            
$\text{Now, the corrector for RK3 is }$                                                                            
$\qquad x^{(n+1)} = x^{(n)} + (k_1 + 4k_2 + k_3)/6 $

```python
def RK3(f, tinit, tfinal, dt, init_cond):
    n = int((tfinal - tinit) / dt) + 1
    t = np.linspace(tinit, tfinal, n)
    x = np.zeros(n)
    x[0] = init_cond
    
    for i in range(n - 1):
        k1 = dt * f(x[i], t[i])
        k2 = dt * f(x[i] + k1/2, t[i] + dt/2)
        k3 = dt * f(x[i] - k1 + 2*k2r, t[i] + dt)
        x[i + 1] = x[i] + (k1 + 4*k2 + k3) / 6
    return t, x
```

### Fourth-order Runge-Kutta  scheme (RK4)

$\text{The predictors for RK4 are} $                                                                            
$\qquad k_1 = (\Delta t) f(x^{(n)},t_n);  \quad\qquad x^{(n+½)^*} = x^{(n)} + k_1/2 , $                                                                            
$\qquad k_2 = (\Delta t) f(x^{(n+½)^*},t_{n+½}); \  \ x^{(n+½)^{**}} = x^{(n)} + k_2/2 ,$                                                                            
$\qquad k_3 = (\Delta t) f(x^{(n+½)^{**}},t_{n+½}); \ \ x^{(n+1)^{***}} = x^{(n)} + k_3 , $                                                                            
$\qquad k_4 = (\Delta t) f(x^{(n+1)^{***}},t_{n+1}) .$                                                                            
$\text{Now, the corrector for RK4 is } $                                                                                                            
$\qquad x^{(n+1)} = x^{(n)} + (k_1 + 2k_2 +2k_3 + k_4)/6 $

```python
def RK4(f, tinit, tfinal, dt, init_cond):
    n=int((tfinal-tinit)/dt)+1
    t = np.linspace(tinit,tfinal,n)
    x = np.zeros(n)
    x[0] = init_cond
    for i in range( n-1 ):
        k1=f(x[i],t[i])
        k2=f(x[i]+(dt*k1/2),t[i]+(dt/2))
        k3=f(x[i]+(dt*k2/2),t[i]+(dt/2))
        k4=f(x[i]+dt*k3,t[i]+dt)
        x[i+1] = x[i] + (dt/6)*(k1 + 2*(k2 + k3) + k4)
    return [t, x]
```

### Leap-frog method (LF)
$\qquad\displaystyle\begin{equation*}\displaystyle{x^{(n+1)} = x^{(n−1)}+2(\Delta t) f( x^{(n)}, t_n)}\end{equation*}$

```python
def LF(f, tinit, tfinal, dt, init_cond):
    n = int((tfinal - tinit) / dt) + 1
    t = np.linspace(tinit, tfinal, n)
    x = np.zeros(n)
    x[0] = init_cond
    
    for i in range(1, n):
        x_half = x[i - 1] + 0.5 * dt * f(x[i - 1], t[i - 1])  
        x[i] = x[i - 1] + dt * f(x_half, t[i - 1])  
    return t, x
```

### Second-order Adams-Bashforth method  (AB2)
$\qquad\displaystyle{x^{(n+1)} = x^{(n)} + (\Delta t) f( x^{(n)}, t_n) ,}$                                                                            

$\qquad\displaystyle{x^{(n+1)} = x^{(n)} + \frac{(\Delta t)}{2} [3 f( x^{(n)}, t_n) − f( x^{(n−1)}, t_{n−1})]}$

```python
def AB2(f, tinit, tfinal, dt, init_cond):
    n = int((tfinal - tinit) / dt) + 1    
    t = np.linspace(tinit, tfinal, n)
    x = np.zeros(n)    
    x[0] = init_cond
    
    for i in range(1, n):
        x_e = x[i-1] + dt * f(x[i-1], t[i])  
        if i > 1:
            x_ab2 = x[i-1] + (dt / 2) * (3 * f(x[i-1], t[i]) - f(x[i-2], t[i-2]))
        else:
            x_ab2 = x_e        
        x[i] = x_ab2
    return t, x
```

$\displaystyle \text{Separate variables }:$
                                                                                                               
$\displaystyle\quad \frac{dx}{dt} = x^2 - 50x $

$\displaystyle or,\quad\int \frac{1}{x(x - 50)} dx = \int dt $
                                                                                                               
$\displaystyle\text{Now, we can write it as two separate integrals }:$

### Fourier Transformtion

Inverse Fourier Transform ( Fourier to Real ) : $\displaystyle f(x) = \sum_{n=-\infty}^\infty {\hat{f}_n}\exp(ik_nx) \qquad$ 

where :                                                                                                    
   * $\displaystyle {\hat{f}_n} =$ Fourier coefficient or Fourier mode

   * $\displaystyle {k_n} = \frac{2n\pi}{L} = $ wavenumbers

The function $\exp(ik_nx)$ form an orthonormal basis , i.e., 

$\qquad\qquad\displaystyle{\int_0^L \exp(ik_nx) \ \exp(-ik_mx) \ dx = L\delta_{m,n}}$

where Kronecker delta function ,  $\begin{equation*} \delta_{m,n}  = \left\{\begin{array}{rl}1 & \text{if } n=m,\\0 & \text{if } n\ne m.\end{array} \right.\end{equation*}$

From Orthonormality property we get the invert as :

Forward Fourier Transform ( Real to Fourier ) : $\displaystyle {\hat{f}_n} = \frac{1}{L} \int_0^L f(x) \exp(-ik_nx) \ dx \qquad$  

Energy density $ \displaystyle{= \frac{|f(x)|^2}{2} }$

Total Energy $ \displaystyle{= \frac{1}{2\pi}\int_{-\infty}^\infty \frac{|f(x)|^2}{2} \ dx}$

Modal energy $ \displaystyle{= \frac{|\hat{f}_n|^2}{2} }$

Parseval's theorem : $\displaystyle{\frac{1}{2L}\int_0^L \frac{|f(x)|^2}{2} \ dx = \frac{1}{2}\sum_{-\infty}^\infty \frac{|\hat{f}_n|^2}{2} }$

Theoritical Analysis :-                                                                                                                               
To find the amplitude of the Fourier mode $\big(\mathbf{k} = 1\big)$ for the function $\big(f(x) = 4\cos(x) + 2\sin(2x)\big)$,                                                                                                                                
we need to calculate the coefficients $a_1$ and $b_1$ in the Fourier series representation.                                                            
The amplitude of the Fourier mode $\big(\mathbf{k}\big)$ is given by:                                                                                                                               
$\displaystyle A_{\mathbf{k}} = \sqrt{a_{\mathbf{k}}^2 + b_{\mathbf{k}}^2}$                                                                                                                               
For $\big(\mathbf{k} = 1\big)$ , the coefficients $a_1$ and $b_1$ are calculated as follows:                                                                                                                               
$\quad\displaystyle a_1 = \frac{1}{\pi} \int_{0}^{2\pi} \left(4\cos(x) + 2\sin(2x)\right) \cos(x) dx$                        
$\quad\displaystyle b_1 = \frac{1}{\pi} \int_{0}^{2\pi} \left(4\cos(x) + 2\sin(2x)\right) \sin(x) dx$                                                                                                                               
Let's calculate these integrals:                                                                                                                               
$\quad\displaystyle a_1 = \frac{1}{\pi} \int_{0}^{2\pi} \left(4\cos^2(x) + 2\sin(2x)\cos(x)\right) dx$                                                                                                                               
Using trigonometric identities $\displaystyle \big(\cos^2(x) = \frac{1}{2}(1 + \cos(2x))\big)$ and $\displaystyle \big(\sin(2x) = 2\sin(x)\cos(x)\big)$, we can simplify $a_1:$                                                                                                                               
$\quad\displaystyle a_1 = \frac{1}{\pi} \int_{0}^{2\pi} \left(2 + 2\sin(2x)\cos(x)\right) dx$                                                                                                                               
Now, calculate $b_1$:                                                                                                                               
$\quad\displaystyle b_1 = \frac{1}{\pi} \int_{0}^{2\pi} \left(4\cos(x)\sin(x) + 2\sin^2(2x)\right) dx$                                                                                                                               
Again, using trigonometric identities $\displaystyle \big(\sin(2x) = 2\sin(x)\cos(x)\big)$ and $\displaystyle \big(\sin^2(2x) = \frac{1}{2}(1 - \cos(4x))\big)$, we can simplify $b_1:$                                                                                                                               
$\quad\displaystyle b_1 = \frac{1}{\pi} \int_{0}^{2\pi} \left(4\sin(x)\cos^2(x) + 1 - \cos(4x)\right) dx$                                                                                                                               
Now, let's calculate these integrals:                                                                                                                               
$\qquad \displaystyle a_1 = \frac{1}{\pi} \left[2x - \frac{1}{4}\sin(2x)\right]_{0}^{2\pi} = 2$                                                                                                                               
$\qquad\displaystyle b_1 = \frac{1}{\pi} \left[\sin(x) - \frac{1}{4}\sin(4x)\right]_{0}^{2\pi} = 0$                                                                                                                               
Now that we have $\big(a_1 = 2\big)$ and $\big(b_1 = 0\big),$ we can calculate the amplitude $A_1$ for $k = 1:$                                                                                                                               
$\qquad\displaystyle  A_1 = \sqrt{a_1^2 + b_1^2} = \sqrt{2^2 + 0^2} = 2$                                                                                                                               
So, the amplitude of the Fourier mode $\big(\mathbf{k} = 1\big)$  for the function $\big(f(x) = 4\cos(x) + 2\sin(2x)\big)$ is $2$.

10. Consider a function $f(x,y)=4\cos(x)\sin(y)$ that is periodic in a $(2π)^2$ box. What is the amplitude of $\hat{f}_{\vec{k} =(1,1)}$ $($Fourier mode at ${\vec{k} =(1,1)})?$                                                                                                                               
Theoritical Analysis :                                                                                                                               
To find the amplitude of the Fourier mode $\big( \mathbf{k} = (1, 1) \big)$ for the function $\big( f(x, y) = 4\cos(x)\sin(y) \big)$ that is periodic in a $(2\pi)^2$ box,                                                                                                                                
we need to calculate the Fourier coefficients $ a_{\mathbf{k}}$ and $ b_{\mathbf{k}}$ for this mode.                                                                                                                                
The amplitude of the Fourier mode $\mathbf{k}$ is given by:                                                                                                                               
$\displaystyle  A_{\mathbf{k}} = \sqrt{a_{\mathbf{k}}^2 + b_{\mathbf{k}}^2} $                                                                                                                               
The coefficients $ a_{\mathbf{k}}$ and $b_{\mathbf{k}}$ are calculated as follows:                                                                                                                               
$\quad\displaystyle  a_{\mathbf{k}} = \frac{1}{(2\pi)^2} \int_{0}^{2\pi} \int_{0}^{2\pi} f(x, y) \cos(k_x x) \cos(k_y y) dx dy $                                                                                                                                
$\quad\displaystyle  b_{\mathbf{k}} = \frac{1}{(2\pi)^2} \int_{0}^{2\pi} \int_{0}^{2\pi} f(x, y) \sin(k_x x) \sin(k_y y) dx dy $                                                                                                                               
For $\big( \mathbf{k} = (1, 1) \big)$, these become:                                                                                                                               
$\quad \displaystyle a_{\mathbf{k}} = \frac{1}{(2\pi)^2} \int_{0}^{2\pi} \int_{0}^{2\pi} 4\cos(x)\sin(y) \cos(x) \cos(y) dx dy $                                                                                                                                        
$\quad \displaystyle  b_{\mathbf{k}} = \frac{1}{(2\pi)^2} \int_{0}^{2\pi} \int_{0}^{2\pi} 4\cos(x)\sin(y) \sin(x) \sin(y) dx dy $                                                                                                                               
Now, let's calculate these integrals:                                                                                                                               
$\quad \displaystyle  a_{\mathbf{k}} = \frac{4}{(2\pi)^2} \int_{0}^{2\pi} \int_{0}^{2\pi} \cos^2(x) \sin^2(y) dx dy $                         
$\quad \displaystyle  b_{\mathbf{k}} = \frac{4}{(2\pi)^2} \int_{0}^{2\pi} \int_{0}^{2\pi} \cos(x) \sin(x) \cos(y) \sin(y) dx dy $                                                                                                                               
Solving these integrals :                                                                                                                               
$\quad\displaystyle  a_{\mathbf{k}} = \frac{1}{\pi^2} \int_{0}^{2\pi} \cos^2(x) dx \int_{0}^{2\pi} \sin^2(y) dy $                        
$\quad\displaystyle  b_{\mathbf{k}} = \frac{1}{\pi^2} \int_{0}^{2\pi} \cos(x) \sin(x) dx \int_{0}^{2\pi} \cos(y) \sin(y) dy $                                                                                                                               
Now, calculate these integrals separately:                                                                                                                               
$\qquad\displaystyle \int_{0}^{2\pi} \cos^2(x) dx = \pi $                                                                                                 
$\qquad\displaystyle  \int_{0}^{2\pi} \sin^2(y) dy = \pi $                                                                                                      
$\qquad\displaystyle  \int_{0}^{2\pi} \cos(x) \sin(x) dx = 0 $                                                                                                
$\qquad\displaystyle  \int_{0}^{2\pi} \cos(y) \sin(y) dy = 0 $                                                                                                                               
So, we have:                                                                                                                               
$\qquad\displaystyle  a_{\mathbf{k}} = \frac{1}{\pi^2} \cdot \pi \cdot \pi = 1 $                                                                          
$\qquad\displaystyle  b_{\mathbf{k}} = 0 $                                                                                                                               
Now, we can calculate the amplitude $ A_{\mathbf{k}} $ for $ \mathbf{k} = (1, 1) :$                                                                                                                               
$\qquad\displaystyle  A_{\mathbf{k}} = \sqrt{a_{\mathbf{k}}^2 + b_{\mathbf{k}}^2} = \sqrt{1^2 + 0^2} = 1 $                                                                                                                               
So, the amplitude of the Fourier mode $\big( \mathbf{k} = (1, 1) \big)$ for the function $\big( f(x, y) = 4\cos(x)\sin(y) \big)$ is $1.$


1. What are the advantages and disadvantages of using Finite Difference method for solving diffusion equations ?

The Finite Difference Method (FDM) is a numerical technique commonly used to solve differential equations, including diffusion equations. Here are some advantages and disadvantages of using the Finite Difference Method for solving diffusion equations:

Advantages:

1. Simplicity: FDM is relatively easy to implement and understand, making it accessible for students and researchers new to numerical methods. It involves discretizing the domain into a grid and approximating derivatives using finite differences.

2. Versatility: FDM can be applied to a wide range of diffusion equations, including one-dimensional, two-dimensional, and three-dimensional problems. It can handle various boundary conditions and source terms.

3. Efficiency: For simple geometries and small to moderately sized problems, FDM can be computationally efficient. It typically requires less computational power compared to more advanced methods like Finite Element or Finite Volume methods.

4. Local Control: FDM allows for local control of grid spacing, meaning you can refine the grid in areas where the solution changes rapidly and coarsen it where the solution is relatively smooth. This adaptability can lead to more accurate results in critical regions.

5. Stability: FDM can provide stable solutions for a wide range of time step sizes and grid resolutions if proper stability conditions are met (e.g., the Courant-Friedrichs-Lewy condition for time-stepping).

Disadvantages:

1. Accuracy: FDM may not provide highly accurate results for problems with complex geometries or rapidly changing solutions. Convergence rates can be slow in such cases.

2. Grid Dependence: The accuracy of FDM solutions is often highly dependent on the choice of grid size and discretization schemes. Fine grids can lead to more accurate results but also increase computational cost.

3. Limited Applicability: FDM is generally best suited for problems with regular geometries and well-behaved solutions. It may not be the most appropriate choice for problems with irregular boundaries or discontinuities.

4. Stability Constraints: FDM's stability can be limited by the Courant-Friedrichs-Lewy (CFL) condition, which imposes restrictions on the time step size to ensure stability. This can lead to long simulation times for problems with rapid transient behavior.

5. Boundary Treatment: Properly handling boundary conditions in FDM can be challenging, especially for problems with complex or non-standard boundary conditions. Special techniques like ghost points or extrapolation may be required.

6. Higher Dimensions: As the problem dimensionality increases, the number of grid points and the associated computational cost grow exponentially, making FDM less efficient for high-dimensional problems.

In summary, the Finite Difference Method is a useful and accessible numerical technique for solving diffusion equations and other partial differential equations. However, its suitability depends on the specific problem at hand, and users should carefully consider its advantages and disadvantages in their particular context.

### Bisection Method

$\qquad\quad\displaystyle{x_{n+1}=\frac{x_n - x_{n-1}}{2}}$

```python
def bisection_method(f, a, b, tol, N=100):
    x = np.zeros(N)
    x[0] = a
    x[1] = b
    for i in range(2,N):
        mid = (x[i-2] + x[i-1])/2
        if (abs(mid-x[i-1]) < tol):
            x[i] = mid
            break
        if f(x[i-2])*f(mid) > 0:
            x[i] = mid
        else:
            x[i-1] = x[i-2]; x[i] = mid
    return (i, x[i])
```

$\qquad\quad\displaystyle{x_{1}=x_0 - \frac{f(x_0)}{f'(x_0)}}$

For $(n+1)^{th}$ step : $\quad\displaystyle{x_{n+1}=x_n - \frac{f(x_n)}{f'(x_n)}}$

```python
def newton_raphson(f, df, a, tol, N=100):
    x = np.zeros(N)
    x[0] = a
    for i in range(1,N):
        x[i] = x[i-1] - f(x[i-1])/df(x[i-1])
        if (abs(x[i]-x[i-1]) < tol):
            break
    return (i, x[i])
```

4. In 2D, a potential $V(x,y)=x^2−y^2$ satisfies the Laplace's equation.                                                            
What is the average value of potential $V$ over a disk centered at $(0,0)$ and with a unit radius?                                                            
Ans :- To find the average value of the potential $V(x, y) = x^2 - y^2$ over a disk centered at $(0, 0)$ with a unit radius,                                                            
we can use the formula for the average value of a function over a region.                                                             
In this case, the region is the disk, which can be described by the inequality $x^2 + y^2 \leq 1$ (a unit circle centered at the origin).                                                            
The average value of a function $f(x, y)$ over a region $R$ is given by the integral :                                                            
$\qquad f_{\text{Average}} = \frac{1}{\text{Area}(R)} \iint_R f(x, y) \, dx \, dy$                                                            
In this case, $f(x, y) = x^2 - y^2$, and $R$ is the unit disk, so :                                                            
$\qquad V_{\text{Average}} = \frac{1}{\pi \cdot 1^2} \iint_{x^2 + y^2 \leq 1} (x^2 - y^2) \, dx \, dy$                                                            
To evaluate this integral, it's convenient to switch to polar coordinates, where $x = r \cos(\theta)$ and $y = r \sin(\theta)$.                                                             
Also, the limits of integration change to $0 \leq r \leq 1$ and $0 \leq \theta \leq 2\pi$.                               
The Jacobian determinant of the coordinate transformation is $r$.                                                            
So, the integral becomes :                                                            
$\qquad V_{\text{Average}} = \frac{1}{\pi} \int_0^1 \int_0^{2\pi} (r^2 \cos^2(\theta) - r^2 \sin^2(\theta)) \cdot r \, d\theta \, dr$                                                            
Now, let's evaluate this integral step by step :                                                            
 Integrate with respect to $\theta$ from $0$ to $2\pi$ :                                                            
    $\qquad\qquad\int_0^{2\pi} (r^2 \cos^2(\theta) - r^2 \sin^2(\theta)) \, d\theta = 0$                                                            
  This is because $\cos^2(\theta)$ and $\sin^2(\theta)$ both integrate to $0$ over one period.                                                            
 Now, you're left with:                                                            
    $\qquad\qquad V_{\text{Average}} = \frac{1}{\pi} \int_0^1 0 \cdot r \, dr = 0$                                                            
So, the average value of the potential $V$ over the unit disk centered at $(0,0)$ with a unit radius is $0$.                
Therefore, the correct answer is $0$.                                                                                                                       
$\qquad\qquad\qquad\boxed{V_{\text{Average}}=0}$

```python
def potential(x, y):
    return x**2 - y**2

x_min, x_max = -1, 1
y_min, y_max = lambda x: -((1 - x**2)**0.5), lambda x: ((1 - x**2)**0.5)

V_average, err = si.integrate.dblquad(lambda y, x: potential(x, y), x_min, x_max, y_min, y_max)
print("Average value of potential over the disk:", round(V_average, 2))
```

```python
def gauss_elmbk(a, b):
    m, n = a.shape              
    b = b.reshape(n,1)
    M = np.hstack((a,b))                       # augmented_mat
    display(Math(r'A|B = '+latex(sp.Matrix(M))))
    for k in range(n):                       
        for j in range(k+1,n):                      
            q = float(M[j][k]) / M[k][k]       # lambda
            for m in range(k, n+1):            # Forward Elimination - Subtracting rows
                M[j][m] -=  q * M[k][m] 
    display(Math(r'A|B = '+latex(sp.Matrix(np.round(M,2)))))
    
    x = np.zeros(n)
    x[n-1] = float(M[n-1][n]) / M[n-1][n-1]  
    for i in range (n-1,-1,-1):                 # Backwards Substitution 
        z = 0.0                                     
        for j in range(i+1,n):                     
            z = z  + float(M[i][j])*x[j]   
        x[i] = float(M[i][n] - z) / M[i][i]
    return x
```

4. For quadrature, what are the advantages and disadvantages of Monte Carlo method over Newton-Cotes method.                                    
Ans :-

**Monte Carlo Method:**

Advantages:
1. **Versatility**: Monte Carlo methods are highly versatile and applicable to a wide range of problems, including high-dimensional integrals or problems with complex geometries. They work well for problems where other methods, like Newton-Cotes, may struggle.

2. **Convergence**: Monte Carlo methods tend to have consistent convergence properties, meaning that as the number of samples (random points) increases, the estimate of the integral becomes more accurate. This makes them useful for problems with no known analytical solution.

3. **Parallelization**: Monte Carlo methods are inherently parallelizable, making them suitable for modern multi-core and distributed computing environments.

4. **Handling Stochasticity**: Monte Carlo methods can naturally handle stochastic or probabilistic integrals and problems where the function to be integrated is not deterministic.

Disadvantages:
1. **Variance**: Monte Carlo methods can have high variance, which means that the estimates may be noisy and require a large number of samples to obtain accurate results. Variance reduction techniques may be necessary.

2. **Computational Cost**: In many cases, Monte Carlo methods require a large number of random samples to achieve reasonable accuracy, making them computationally expensive.

3. **Deterministic Precision**: If high-precision deterministic results are required, Monte Carlo methods may not be the best choice, as they inherently involve random sampling.

**Newton-Cotes Methods (e.g., Trapezoidal Rule, Simpson's Rule):**

Advantages:
1. **Deterministic**: Newton-Cotes methods provide deterministic results. The accuracy of the result depends on the chosen integration rule and the number of subintervals, but the result is not subject to random variability.

2. **Low Variance**: These methods typically have lower variance compared to Monte Carlo methods, which can lead to more predictable and stable results.

3. **Ease of Implementation**: Newton-Cotes methods are relatively straightforward to implement and understand, making them a good choice for simple integration problems.

4. **Deterministic Precision**: If high-precision deterministic results are required, Newton-Cotes methods can provide such results without the need for extensive sampling.

Disadvantages:
1. **Applicability**: Newton-Cotes methods may not work well for problems with irregular or complex geometries, highly oscillatory functions, or high-dimensional integrals.

2. **Limited to Specific Rules**: Different Newton-Cotes rules (e.g., Trapezoidal Rule, Simpson's Rule) have different convergence rates and may require a different number of intervals for accurate results. Choosing the right rule can be a trial-and-error process.

8. Explain why Metropolis algorithm works while simulating equilibrium statistical systems.                                                        
Ans :-  The Metropolis algorithm is a powerful and widely used algorithm in the field of computational statistical physics for simulating equilibrium statistical systems. It's part of a class of algorithms known as Markov Chain Monte Carlo (MCMC) methods. The Metropolis algorithm works effectively for simulating equilibrium statistical systems for several reasons:

a. **Ergodicity**:
   - The Metropolis algorithm is designed to be ergodic, which means it can explore the entire phase space of a system, including all possible microstates.
   - Ergodicity ensures that, given sufficient time, the algorithm can sample states from the entire equilibrium distribution of the system, not just a limited subset of states.

b. **Detailed Balance**:
   - The Metropolis algorithm satisfies the principle of detailed balance. In other words, it ensures that the system reaches a state of detailed balance, where the rate of transitions between states is balanced by the rate of transitions in the opposite direction.
   - This property guarantees that the algorithm converges to the correct equilibrium distribution, making it suitable for simulating equilibrium systems.

c. **Markov Property**:
   - The Metropolis algorithm has the Markov property, which means that the next state of the system depends only on the current state and not on the entire history of states.
   - This property simplifies the simulation process and ensures that the simulation is memoryless and self-contained.

d. **Rejection Sampling**:
   - The Metropolis algorithm employs a rejection sampling scheme that allows it to generate proposed states and then either accept or reject them based on a probability criterion.
   - This mechanism enables the algorithm to explore the phase space efficiently, as it proposes changes to the system configuration and accepts these changes based on the Metropolis acceptance probability, which depends on the energy change and temperature of the system.

e. **Adaptability to Various Potentials**:
   - The Metropolis algorithm can be applied to systems with a wide range of interaction potentials and Hamiltonians.
   - It doesn't require explicit knowledge of the partition function, which makes it flexible and applicable to diverse physical systems.

f. **Temperature Control**:
   - The algorithm can simulate systems at different temperatures by adjusting the acceptance probability criterion accordingly.
   - This feature is crucial for studying the temperature-dependent behavior of equilibrium systems.

g. **Parallelization**:
   - The Metropolis algorithm can be parallelized easily, allowing for efficient simulations on modern computing architectures.

h. **Widely Tested and Validated**:
   - The Metropolis algorithm has been extensively tested and validated against known analytical results for various statistical systems,                                                                                                                    
   and it has been found to provide accurate results.

In summary, the Metropolis algorithm works while simulating equilibrium statistical systems due to its ergodicity, adherence to detailed balance,                                                                                                             
Markov property, rejection sampling mechanism, adaptability to different potentials, temperature control, parallelizability, and extensive validation.                                                                             
These properties make it a robust and versatile tool for studying a wide range of equilibrium systems in statistical physics.

9. Simulate 2D Ising model for T = 1, 2, 2.5, 3, 4, 10. Compute the decay/growth rate of M for these runs.                           
initial configuration : $\langle M\rangle\approx 0.7$

##### Ising Phase Transition using Metropolis algorithm

```python
N = 100
M = 100
eps = 0.01
m_init = 0.7
max_steps = 100000

def mat_init(N,M,m_init):
    A = np.random.randint(0,2,(N+2,M+2))
    A[0,:]=0
    A[N+1,:]=0
    A[:,0]=0
    A[:,M+1]=0
    for i in range(1,N+1):
        for j in range(1,M+1):
            random_no = np.random.rand()
            if (A[i,j] == 0):
                if (random_no > m_init):
                    A[i,j] = -1
                else:
                    A[i,j] = 1
    return A

def energy(A):
    return (-np.sum(A[1:N+1,1:M+1]*A[0:N,1:M+1]  \
                +  A[1:N+1,1:M+1]*A[2:N+2,1:M+1] + A[1:N+1,1:M+1]*A[1:N+1,0:M] + A[1:N+1,1:M+1]*A[0:N,2:M+2]))

def Em(energy,T, A, N, M, max_steps, eps):
    E = np.zeros(max_steps+1)
    m = np.zeros(max_steps+1)
    Enow = energy(A)
    E[0] = Enow
    m[0] = np.sum(A)/(N*M)
    
    for step in range(1,max_steps):
        i = np.random.randint(1,N+1)
        j = np.random.randint(1,M+1)
        A[i,j] = -1*A[i,j]
        Enext = energy(A)
        #print("step, Enow, Enext, <m> = ",step, Enow, Enext, np.sum(A)/(N*M))
        dE = Enext-Enow
        
        if np.random.rand() < np.exp(-dE/T) :
                Enow = Enext                     # accept change
        else:
                A[i,j] = -1*A[i,j]
        E[step] = Enow
        m[step] = np.sum(A)/(N*M)
    return E, m

t0 = datetime.now()
fig = plt.figure(figsize = (6,4))
ax1 = fig.add_subplot(1,1,1)

for T in [1, 2, 2.5, 3, 4, 10]:
    A = mat_init(N,M,m_init)
    E,m = Em(energy,T, A, N, M, max_steps, eps)
    ax1.plot(m, label='T={}'.format(T))

ax1.set_xlabel('n')
ax1.set_ylabel(r'$\langle M\rangle$')
ax1.set_xticks([0, 50000, 100000])
ax1.set_yticks([0, 0.35, 0.7, 1])
ax1.set_xlim(0,100000)
fig.suptitle('Ising Model 2D', fontsize=24)
plt.tight_layout()
ax1.legend(loc= (1.1,0.1), fontsize=18)
plt.show()

t1 = datetime.now()
print (" time elapsed in Ising = ", t1-t0)
```
