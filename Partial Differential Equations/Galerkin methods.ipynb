```python
def RK4(f, t_span, const, dt, x0):
    t = np.arange(t_span[0], t_span[1], dt)
    X = np.zeros((len(t), len(x0)))
    X[0] = x0
    
    for i in range(1, len(t)):
        k1 = dt * f(t[i-1], X[i-1], const)
        k2 = dt * f(t[i-1] + 0.5*dt, X[i-1] + 0.5*k1, const)
        k3 = dt * f(t[i-1] + 0.5*dt, X[i-1] + 0.5*k2, const)
        k4 = dt * f(t[i-1] + dt, X[i-1] + k3, const)
        X[i] = X[i-1] + (k1 + 2*k2 + 2*k3 + k4)/6 
    return t, X
```

<span style="color:red">Problem 15.3                                                                                                                                                 
Show that an infinite square well (extending from zero to $L$) with bottom that moves up and down sinusoidally                                             
according to $\displaystyle V(t) = V_{\tiny 0}\sin(2\pi f t)$ cannot cause “transitions” from one energy eigenstate to another – if we start at $t = 0$                  
in the state $\displaystyle \psi(x) = \psi_j(x) = \frac{2}{L}\sin(\frac{j\pi x}{L}),$ there is no time at which $\psi(x, t)$ has spatial component $\psi_k(x)$ for $k\ne j$                                                                                                                                                  
(i.e. show that $\displaystyle\int^L_0 \psi(x, t )^∗\, \psi_k(x)\,dx = 0$ for all $t$ ). <span>                                                               

**Hamiltonian**

Infinite square well with time-dependent bottom:

$\qquad\qquad\displaystyle H(t)=H_0 + V(t), \quad V(t)=V_0 \sin(2\pi f t)$

Here $V(t)$ is **only time-dependent (no $x$)**, i.e. a multiple of the identity.

**Eigenstates of $H_0$**

$\qquad\qquad\displaystyle \psi_j(x)=\sqrt{\frac{2}{L}}\sin\!\left(\frac{j\pi x}{L}\right), \quad H_0 \psi_j = E_j \psi_j$

**Effect of $V(t)$**

Since $V(t)$ is spatially uniform:

$\qquad\qquad\displaystyle H(t)\psi_j = \big(E_j + V(t)\big)\psi_j$

So **each eigenstate remains an eigenstate** (no mixing).

**Time evolution**

Schrödinger equation 

$\qquad\qquad\displaystyle i\hbar \frac{\partial \psi}{\partial t} = (E_j + V(t))\psi$

Solution

$\qquad\qquad\displaystyle \psi(x,t)=\psi_j(x)\exp\!\left[-\frac{i}{\hbar}\left(E_j t + \int_0^t V(t')dt'\right)\right]$

Projection onto other states

$\qquad\qquad\displaystyle \int_0^L \psi(x,t)^* \psi_k(x)\,dx = e^{i\theta(t)} \int_0^L \psi_j(x)\psi_k(x)\,dx$

Using orthogonality

$\qquad\qquad\displaystyle \int_0^L \psi_j \psi_k dx = \delta_{jk}$

Thus 

$\qquad\qquad\displaystyle \boxed{\int_0^L \psi(x,t)^* \psi_k(x)\,dx = 0 \quad (k\ne j)}$


**Conclusion**

A purely time-dependent potential adds only a **global phase**, so **no transitions occur between eigenstates**.

**Separation of variables**

Let $u(x,t)=X(x)T(t)$. Then

$\qquad\qquad\displaystyle\frac{1}{v^2}\frac{T''}{T}=\frac{X''}{X}=-\lambda$

**Spatial eigenproblem**

$\qquad\qquad\displaystyle X'' + \lambda X = 0$

BCs
$\qquad\qquad\displaystyle X(0)=0,\quad X'(1)=0$

General solution: $\qquad\qquad\displaystyle X=A\sin(\sqrt{\lambda}x)+B\cos(\sqrt{\lambda}x)$

From $X(0)=0\Rightarrow B=0$

$\qquad\qquad\displaystyle X = A\sin(\sqrt{\lambda}x)$

Apply $X'(1)=0$ 

$\qquad\qquad\displaystyle \sqrt{\lambda}\cos(\sqrt{\lambda})=0 \Rightarrow \cos(\sqrt{\lambda})=0$

$\qquad\qquad\displaystyle \sqrt{\lambda_n}=\frac{(2n-1)\pi}{2}$

**Eigenfunctions**

$\qquad\qquad\displaystyle \boxed{\phi_n(x)=\sin\!\left(\frac{(2n-1)\pi x}{2}\right)}$

**Time equation**

$\qquad\qquad\displaystyle T_n'' + v^2 \lambda_n T_n = 0$

$\qquad\qquad\displaystyle T_n(t)=A_n\cos(\omega_n t)+B_n\sin(\omega_n t), \quad \omega_n = v\sqrt{\lambda_n} =v\frac{(2n-1)\pi}{2}$

General solution

$\qquad\qquad\displaystyle u(x,t)=\sum_{n=1}^\infty \left[A_n\cos(\omega_n t)+B_n\sin(\omega_n t)\right]\phi_n(x)$

Initial conditions

When $u(x,0)=u_0 x(2-x)$

$\qquad\qquad\displaystyle A_n = 2\int_0^1 u_0 x(2-x)\sin\!\left(\frac{(2n-1)\pi x}{2}\right)\,dx$

Then $\dot{u}(x,0)=0 \Rightarrow B_n=0$

Thus

$\qquad\qquad\displaystyle\boxed{u(x,t)=\sum_{n=1}^\infty A_n \cos(\omega_n t)\sin\!\left(\frac{(2n-1)\pi x}{2}\right)}$

**Key insight**

Mixed BCs $\rightarrow$ **half-wave modes** $\frac{(2n-1)}{2}$, not integer modes.

```python
# symbols
x, y, L = sp.symbols('x y L', positive=True)
a1 = sp.Function('a1')(y)
a2 = sp.Function('a2')(y)

# basis
phi1 = sp.sin(sp.pi*x/L)
phi2 = sp.sin(2*sp.pi*x/L)

# potential
V = a1*phi1 + a2*phi2

# derivatives
Vx = sp.diff(V, x)
Vy = sp.diff(V, y)

Ex = -Vx
Ey = -Vy
E2 = Ex**2 + Ey**2

# divergence term
divE = sp.diff(Ex, x) + sp.diff(Ey, y)
term1 = divE*(1 - E2)

# nonlinear term
term2 = (Ex*Ex*sp.diff(Ex, x) + Ex*Ey*sp.diff(Ey, x) + Ey*Ex*sp.diff(Ex, y) + Ey*Ey*sp.diff(Ey, y))
r = sp.simplify(term1 + term2)

# Galerkin equations
eq1 = sp.integrate(phi1 * r, (x, 0, L))
eq2 = sp.integrate(phi2 * r, (x, 0, L))

sp.simplify(eq1), sp.simplify(eq2)
```

The Galerkin approach works even when the basis set is itself discrete (i.e. the functions $\{φ_j\}^{\infty}_{j=1}$ are vectors $\{v_j\}^n_{j=1}$ ).       
Consider a sourced wave equation (think of the potential associated with a time-varying charge distribution) in two dimensions :                              
$\qquad\qquad\displaystyle -\ddot{u}+\nabla^2 u = −f(x, y)\cos(\omega t), \qquad\qquad(15.58)$                                                            
where $f(x, y)$ is some spatial function, and the source’s time dependence is given explicitly.                                                                   
Take $\{v_j\}^n_{j=1}$  to be the eigenvectors of the discrete Laplacian matrix, $\mathbb{D}$ (from, for example, Section 4.3),                                
with eigenvalues $\{\lambda_j\}^n_{j=1}$ so that: $\mathbb{D}\,\mathbf{v}_j = \lambda_j \mathbf{v}_j$ .                                                          
Since these eigenvectors are complete, we know that the function $f(x, y),$ evaluated on our two-dimensional grid, can be written as :                       
$\qquad\qquad\displaystyle \mathbf{f} = \sum^n_{j=1} \beta_j\,\mathbf{v}_j\;, \qquad\qquad (15.59)$                                                             
for some set of coefficients $\{\beta_j\}^n_{j=1}.$                                                                                                            
The discretized PDE is : $−\mathbf{\ddot{u}} + \mathbb{D}\,\mathbf{u} = −\mathbf{f}\cos(\omega t)$.                                                               
Solve for the temporal evolution of the coefficients in the (finite) Galerkin expansion                                                                              
$\qquad\qquad\displaystyle \mathbf{u} = \sum^n_{j=1}  \alpha_j(t)\,\mathbf{v}_j \,,\qquad\qquad (15.60)$                                                           
given the coefficients of the decomposition at $t = 0, \alpha_j (0) ≡ \alpha_{j0}$, and assuming we begin “from rest,” so that $\dot{\alpha}_j(0)=0$.          
The boundary conditions, implicit in the definition of $\mathbb{D}$, are $u = 0$ for all four sides of the two-dimensional domain. <span>                         

**Galerkin expansion**

$\qquad\qquad\displaystyle \mathbf{u}(t)=\sum_{j=1}^n \alpha_j(t)\,\mathbf{v}_j$

Given

$\qquad\qquad\displaystyle \mathbb{D}\mathbf{v}_j=\lambda_j \mathbf{v}*j,\quad \mathbf{f}=\sum*{j=1}^n \beta_j \mathbf{v}_j$

Substitute into PDE

$\qquad\qquad\displaystyle -\sum \ddot{\alpha}_j \mathbf{v}_j + \sum \alpha_j \lambda_j \mathbf{v}_j = -\sum \beta_j \mathbf{v}_j \cos(\omega t)$

**Orthogonality**

Project onto $\mathbf{v}_k$

$\qquad\qquad\displaystyle -\ddot{\alpha}_k + \lambda_k \alpha_k = -\beta_k \cos(\omega t)$

ODE for each mode

$\qquad\qquad\displaystyle\boxed{\ddot{\alpha}_k - \lambda_k \alpha_k = \beta_k \cos(\omega t)}$

**Solve ODE**

Homogeneous

$\qquad\qquad\displaystyle \alpha_k^{(h)} = C_1 \cos(\sqrt{\lambda_k}t) + C_2 \sin(\sqrt{\lambda_k}t)$

Particular (try $A\cos\omega t$) 

$\qquad\qquad\displaystyle A = \frac{\beta_k}{\lambda_k - \omega^2}$

So

$\qquad\qquad\displaystyle \alpha_k(t)=C_1 \cos(\sqrt{\lambda_k}t) + C_2 \sin(\sqrt{\lambda_k}t) + \frac{\beta_k}{\lambda_k - \omega^2}\cos(\omega t)$

Apply initial conditions

$\qquad\qquad\displaystyle \alpha_k(0)=\alpha_{k0}\, \quad \dot{\alpha}_k(0)=0$

From $\dot{\alpha}_k(0)=0 ⇒ C_2=0$

From $\alpha_k(0)$

$\qquad\qquad\displaystyle C_1 + \frac{\beta_k}{\lambda_k - \omega^2} = \alpha_{k0}$

$\qquad\qquad\displaystyle C_1 = \alpha_{k0} - \frac{\beta_k}{\lambda_k - \omega^2}$

Thus

$\qquad\qquad\displaystyle\boxed{\alpha_k(t)=\left(\alpha_{k0} - \frac{\beta_k}{\lambda_k - \omega^2}\right)\cos(\sqrt{\lambda_k},t) +\frac{\beta_k}{\lambda_k - \omega^2}\cos(\omega t)}$

**Special case (resonance)**

If $\lambda_k=\omega^2$, then

$\qquad\qquad\displaystyle \alpha_k(t)=\alpha_{k0}\cos(\omega t) + \frac{\beta_k}{2\omega}t\sin(\omega t)$

**Key insight**

Each mode behaves like an **independent driven harmonic oscillator**.

$\qquad\displaystyle V_2(x,y)=a_1(y)\sin\left(\frac{\pi x}{L}\right) + a_2(y)\sin\left(\frac{2\pi x}{L}\right)$                                                 
appropriate to a two-dimensional, square region with side length $L$ (with $V_2 = 0$ at $x = 0,L$), using an ODE solver with shooting.                                     
Set up your RK4 routine using your result from Problem 15.8 to define the vector function $G$.                                                                        
For boundary conditions, use                                                                                                                                       
$\qquad\qquad\displaystyle V(x,0)=0.01\sin(\pi x), \quad V(x,1)=0.01\sin(2\pi x)$                                                                        
or equivalently:                                                                                                                                                 
$\qquad\qquad\displaystyle a_1(0)=0.01,\; a_2(0)=0,\quad a_1(1)=0,\; a_2(1)=0.01$                                                                                 
Define a function $u$ that takes trial values of $a_1'(0), a_2'(0)$, runs the ODE solver with $N_s=100,\,\Delta y=0.01$, and returns                        
$\qquad\qquad\displaystyle u = a_1(1)^2 + (a_2(1)-0.01)^2$                                                                                                       
Minimize $u$ using steepest descent (Algorithm 3.1).                                                                                                     
Compute gradients via finite differences with step $\delta=10^{-4}$, normalize the gradient, and update with step size $\eta$.                                   
Start from the origin, use :
* $\eta=10^{-4}$ for first 100 steps
* then divide $\eta$ by $10$ every $100$ steps                                                                                                                
  Stop when $u \le 10^{-4}$
  
<span style="color:red">Find optimal $a_1'(0), a_2'(0)$, then construct $V_2(x,y)$ and plot contours.


```python
def G(y, Y, const):
    L = const
    a1, a2, b1, b2 = Y
    db1 = - (np.pi/L)**2 * a1 + 0.5 * a1 * (a1**2 + a2**2)
    db2 = - (2*np.pi/L)**2 * a2 + 0.5 * a2 * (a1**2 + a2**2)    
    return np.array([b1, b2, db1, db2])

def shoot(s):
    s1, s2 = s   
    Y0 = np.array([0.01, 0.0, s1, s2])
    t, sol = RK4(G, [0, Ns*dy], L, dy, Y0)
    a1, a2 = sol[-1, 0], sol[-1, 1]
    return a1**2 + (a2 - 0.01)**2

def grad_u(s):
    g = np.zeros(2)
    for i in range(2):
        ds = np.zeros(2)
        ds[i] = delta
        g[i] = (shoot(s + ds) - shoot(s - ds)) / (2*delta)
    norm = np.linalg.norm(g)
    return g / norm if norm != 0 else g

L = 1.0
dy = 0.01
Ns = 100
N  = 1000
delta = 1e-4
s = np.array([0.0, 0.0])
eta = 1e-4
history = []

for k in range(N+1):
    g = grad_u(s)
    s = s - eta*g
    val = shoot(s)
    history.append(val)
    
    if k % 100 == 0 and k != 0:
        eta /= 10
    
    if k % 10 == 0:
        clear_output(wait=True)
        plt.plot(history)
        plt.yscale('log')
        plt.title(f"Iter {k}, u = {val:.2e}")
        plt.xlabel("iteration")
        plt.ylabel("u")
        plt.xlim(0, N)
        plt.grid()
        plt.show()
    
    if val <= 1e-4:
        break

print("Optimal slopes:", s)
print("Final error:", val)
```

```python
Y0 = np.array([0.01, 0.0, s[0], s[1]])
t_vals, sol = RK4(G, [0, Ns*dy], L, dy, Y0)
A = sol[:, :2]   # a1, a2
x = np.linspace(0,1,100)
V2 = np.zeros((len(t_vals), len(x)))
for j in range(len(t_vals)):
    a1, a2 = A[j]
    V2[j,:] = a1*np.sin(np.pi*x) + a2*np.sin(2*np.pi*x)

for j in range(0, len(t_vals), 5):
    clear_output(wait=True)

    fig = plt.figure(figsize=(12,5))
    plt.plot(x, V2[j])
    plt.ylim(-0.02, 0.02)
    plt.xlabel('x')
    plt.xlabel(r'$V_2(x,t)$')
    plt.title(f"t = {t_vals[j]:.2f}")
    plt.grid()
    plt.show()
    time.sleep(0.03)
```

<span style="color:red">Problem 15.16                                                                                                                     
Using the eigenvectors of the finite-difference approximation to the Laplace operator,                                                                    
construct the solution to the forced wave equation as in Problem 15.9.                                                                                        
That is, write coefficient functions $a_j(t)$ and use them in the sum (15.60) to generate $u(x,y,t)$.                                                                 
Take a grid with $N_x = N_y = 29,\,\Delta x = \Delta y = \frac{1}{N_x+1}$.                                                                                  
Use a time-varying source                                                                                                                                    
$\qquad\qquad\displaystyle f(x_j,y_k)= 1 \text{ if } j=15,\;k=15,\quad \text{else } 0$                                                                        
Initial condition $u(x,y,0)=0$.                                                                                                                             
Use driving frequency $\omega=1$.                                                                                                                               
Tasks :                                      
1. Plot $a_1(t)$ for $t\in[0,1]$
2. Plot $a_{841}(t)$ for $t\in[0,10]$
3. Make a contour movie of $u(x,y,t)$ for $t\in[0,2], \; \Delta t=0.01$ <span>


**Solution** 

Discrete wave equation

$\qquad\qquad\displaystyle
-\ddot{\mathbf u} + \mathbb{D}\mathbf u = -\mathbf f \cos(\omega t)$

Expand

$\qquad\qquad\displaystyle \mathbf u(t)=\sum_{j=1}^{841} a_j(t),\mathbf v_j,\quad
\mathbb{D}\mathbf v_j=\lambda_j \mathbf v_j$

Substitute and project

$\qquad\qquad\displaystyle 
-\sum \ddot a_j \mathbf v_j + \sum a_j \lambda_j \mathbf v_j
= -\sum \beta_j \mathbf v_j \cos(\omega t)$

Project onto $\mathbf v_k$ :

$\qquad\qquad\displaystyle\boxed{\ddot a_k - \lambda_k a_k = \beta_k \cos(\omega t)}$

Let $\lambda_k<0 ,$ define

$\qquad\qquad\displaystyle\omega_k=\sqrt{-\lambda_k}$

Homogeneous

$\qquad\qquad\displaystyle a_k^{(h)}=C_1\cos(\omega_k t)+C_2\sin(\omega_k t)$

Particular

$\qquad\qquad\displaystyle a_k^{(p)}=\frac{\beta_k}{\lambda_k-\omega^2}\cos(\omega t)$

Apply ICs

$\qquad\qquad\displaystyle a_k(0)=0,\quad \dot a_k(0)=0$

Gives 

$\qquad\qquad\displaystyle C_2=0,\quad C_1=-\frac{\beta_k}{\lambda_k-\omega^2}$

Thus

$\qquad\qquad\displaystyle\boxed{a_k(t)=\frac{\beta_k}{\lambda_k-\omega^2}\Big[\cos(\omega t)-\cos(\omega_k t)\Big]}$

gives 

$\qquad\qquad\displaystyle\boxed{u(x,y,t)=\sum_{k=1}^{841}a_k(t)\,v_k(x,y)}$

**Source projection**

$\qquad\qquad\displaystyle \beta_k = v_k(15,15)$

**Final insight (exam line)**

Each mode behaves as a **forced oscillator**, and the full solution is a superposition of these driven modes with frequencies $\omega$ and $\omega_k$.


```python
Nx = Ny = 29
dx = 1/(Nx+1)

# Laplacian
main = -2*np.ones(Nx)
off = np.ones(Nx-1)
T = np.diag(main) + np.diag(off,1) + np.diag(off,-1)
I = np.eye(Nx)
D = (np.kron(I,T) + np.kron(T,I)) / dx**2

# eigenpairs
eigvals, eigvecs = np.linalg.eigh(D)

# source at center
f = np.zeros((Nx,Ny))
f[14,14] = 1
beta = eigvecs.T @ f.flatten()
omega = 1.0

# coefficients
def a_all(t, eigvals, beta, omega):
    wj = np.sqrt(-eigvals)
    return beta/(eigvals - omega**2) * (np.cos(omega*t) - np.cos(wj*t))

t1 = np.linspace(0,1,200)
t2 = np.linspace(0,10,500)
a1 = [a_all(t, eigvals, beta, omega)[-1] for t in t1]
a841 = [a_all(t, eigvals, beta, omega)[0] for t in t2]

plt.figure(figsize=(14,5))
plt.subplot(121)
plt.plot(t1, a1)
plt.title(r"$a_1(t)$")

plt.subplot(122)
plt.plot(t2, a841)
plt.title(r"$a_{841}(t)$")
plt.show()
```

```python
def u_t(t, eigvals, eigvecs, beta, omega, Nx, Ny):
    return (eigvecs @ a_all(t, eigvals, beta, omega)).reshape((Nx,Ny))

x = np.linspace(0,1,Nx)
y = np.linspace(0,1,Ny)
X, Y = np.meshgrid(x,y, indexing='ij')

for t in np.arange(0,2,0.01):
    U = u_t(t, eigvals, eigvecs, beta, omega, Nx, Ny)
    clear_output(wait=True)
    plt.figure(figsize=(8,7))
    plt.contourf(X, Y, U, levels=30)
    plt.colorbar()
    plt.title(f"t = {t:.2f}")
    plt.show()
    time.sleep(0.03)
```

<span style="color:red">Problem 15.17                                                                                                                                                
We can treat the heat equation in the same, fully discrete manner as we did for the wave equation in Problem 15.16.                                      
Using                                                                                                                                                      
$\qquad\qquad\displaystyle u = \sum_{j=1}^n \alpha_j(t )v_j \qquad\qquad (15.64)$                                                                                
for the eigenvectors $\{v_j\}^n_{j=1}$ of the discrete Laplacian operator, find the “exact” solution to $\mathbb{D}u = \dot{u}$                               
(with the heat coefficient set to unity) – you will again need to solve for the $\{\alpha_j (t)\}$ analytically, as in Problem 15.9.                                                                                                                                        
As your initial condition, take $\displaystyle u(x, y, 0) =\sin(\frac{2\pi x}{L})\sin(\frac{3\pi y}{L})$ for a square grid with $L = 1$.                            
Use the same discretization parameters as in Problem 15.16, namely $N_x = N_y = 29, \Delta x = \Delta y = \frac{1}{N_x+1}$.                                   
Plot the coefficients $\alpha_1(t)$ and $\alpha_{841}(t)$ as functions of time, and                                                                        
make a movie of the three dimensional plots of the solution $u(t)$ for $t = 0 \rightarrow 0.03$ in steps of $\Delta t = 0.001$.<span>                           

Solution 

Expansion in eigenbasis

$\qquad\qquad\displaystyle u(t)=\sum_{j=1}^{n}\alpha_j(t)\,v_j\,\quad \mathbb{D}v_j=\lambda_j v_j$

Substitute into equation

$\qquad\qquad\displaystyle \dot{u}=\mathbb{D}u \Rightarrow \sum \dot{\alpha}_j v_j=\sum \alpha_j \lambda_j v_j$

Project onto $v_k$

$\qquad\qquad\displaystyle\boxed{\dot{\alpha}_k=\lambda_k \alpha_k}$

Solve ODE

$\qquad\qquad\displaystyle \boxed{\alpha_k(t)=\alpha_{k0}\,e^{\lambda_k t}}$

Initial condition

$\qquad\qquad\displaystyle u(x,y,0)=\sin(2\pi x)\sin(3\pi y)$

This corresponds to **one mode only**, so

$\qquad\qquad\displaystyle \alpha_{k0}\neq 0 \text{ for one } k ,\quad \text{others }=0$

Thus

$\qquad\qquad\displaystyle\boxed{u(t)=\sum_{j=1}^n \alpha_{j0} e^{\lambda_j t} v_j}$

**Key insight**

Each mode evolves independently with exponential decay

$\qquad\qquad\displaystyle \alpha_k(t)\sim e^{\lambda_k t}, \quad \lambda_k<0$ ⇒ higher modes decay faster.


```python
Nx = Ny = 29
dx = 1/(Nx+1)

# build Laplacian
main = -2*np.ones(Nx)
off = np.ones(Nx-1)
T = np.diag(main) + np.diag(off,1) + np.diag(off,-1)
I = np.eye(Nx)
D = (np.kron(I,T) + np.kron(T,I)) / dx**2
eigvals, eigvecs = np.linalg.eigh(D)

# initial condition
u0 = np.sin(2*np.pi*X)*np.sin(3*np.pi*Y)
u0_flat = u0.flatten()

# coefficients
alpha0 = eigvecs.T @ u0_flat

t_vals = np.arange(0,0.031,0.001)
alpha1 = alpha0[-1] * np.exp(eigvals[-1]*t_vals)
alpha_last = alpha0[0] * np.exp(eigvals[0]*t_vals)

plt.figure(figsize=(16,6))
plt.subplot(121)
plt.plot(t_vals, alpha1)
plt.title(r"$a_1(t)$")

plt.subplot(122)
plt.plot(t_vals, alpha_last)
plt.title(r"$a_{841}(t)$")
plt.show()
```

```python
def u_t(t, alpha0, eigvals, eigvecs, Nx, Ny):
    alpha = alpha0 * np.exp(eigvals*t)
    u = eigvecs @ alpha
    return u.reshape((Nx,Ny))

x = np.linspace(0,1,Nx)
y = np.linspace(0,1,Ny)
X, Y = np.meshgrid(x,y)

for t in t_vals:
    U = u_t(t, alpha0, eigvals, eigvecs, Nx, Ny)
    clear_output(wait=True)    
    fig = plt.figure(figsize=(6,6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, U, cmap='hot')
    ax.set_title(f"t = {t:.3f}")
    plt.show()
    time.sleep(0.05)
```
