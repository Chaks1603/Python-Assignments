2. Plot the eight figure curve :

5. Write a function called `parametric_plots` which takes input parameters `a` and `k` and plot the parametric equation :
    $$\displaystyle{x(t)=2k\cos(t)-a\cos(kt) \\ y(t)=2k\sin(t)-a\sin(kt) }$$
for $\displaystyle{t\in[0,2\pi]}$    

```python
def parametric_plots(a,k):
    t=np.arange(0,2*np.pi,0.01)
    def x(t):
        return (2*k*np.cos(t))-(a*np.cos(k*t))
    
    def y(t):
        return (2*k*np.sin(t))-(a*np.sin(k*t))
    
    plt.plot(x(t),y(t))
    plt.axis('equal')
    plt.title('a = {}, k = {}'.format(a,k),size=30)
    
    return None

for k in range(2,10,1):
    plt.subplot(3,3,k)
    parametric_plots(2,k)
plt.show()
```

6. Write a function called `Hypotrochoid_plots` which takes input parameters `R`,`r` and `d` and plot the parametric equation :
    $$\displaystyle{x(t)=(R-r)\cos(t)+d\cos(\frac{R-r}{r}t) \\ y(t)=(R-r)\sin(t)-d\sin(\frac{R-r}{r}t)}$$
for $\displaystyle{t\in[0,6\pi]}$    

```python
def Hypotrochoid_plots(R,r,d):
    t=np.arange(0,6*np.pi,0.00001)
    def x(t):
        return ((R-r)*np.cos(t))+(d*np.cos((R-r)/r*t))
    
    def y(t):
        return ((R-r)*np.sin(t))-(d*np.sin((R-r)/r*t))
    
    plt.plot(x(t),y(t))
    plt.axis('equal')
    plt.title('R = {}, r = {}, d = {}'.format(R,r,d),size=30)
    plt.show()
    return None

Hypotrochoid_plots(5,3,5)
Hypotrochoid_plots(5,1,1)
Hypotrochoid_plots(5,2,2)
Hypotrochoid_plots(5,1,5)
```

Q2.Consider the following function:
$$ \displaystyle{ y(x)=\sqrt{\frac{100(1-0.01x^2)^2 + 0.02x^2}{(1-x^2+0.1x^2)}}}$$
Generate a 2×2 subplot of y(x) for 0 ≤ x ≤ 100 using plot, semilogx, semilogy, and loglog.
Use a fine enough discretization in x to make the plot appear smooth. 
Give each plot axis labels and a title. Turn on the grid. 
Which plot seems to convey the most information?.

```python
def f(x):
    num=100*(1-(0.01*x*x))**2+(0.02*x*x)
    dnum=(1-x*x)**2+(0.1*x*x)
    sr=(num/dnum)**0.5
    return sr

x=np.arange(0,100,0.01)
y=f(x)

plt.figure(figsize = (20, 8))
plt.subplot(1, 4, 1)
plt.plot(x,y)
plt.title("Plot",fontsize = 18)
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.subplot(1, 4, 2)
plt.loglog(x,y)
plt.title("Loglog",fontsize = 18)
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(which="both")
plt.subplot(1, 4, 3)
plt.semilogx(x,y)
plt.title("Semilogx",fontsize = 18)
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(which="both")
plt.subplot(1, 4, 4)
plt.semilogy(x,y)
plt.title("Semilogy",fontsize = 18)
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.tight_layout()
plt.show()
```

Q7. Write a function my_polygon(n) that plots a regular polygon with n sides and radius 1.
Recall that the radius of a regular polygon is the distance from its centroid to the vertices. 
Use plt.axis("equal") to make the polygon look regular. 
Remember to give the axes a label and a title. 
Use plt.title to title the plot according to the number of sides. 
Hint: This problem is significantly easier if you think in polar coordinates.
Recall that a complete revolution around the unit circle is 2π radians. 
Note that the first and last point on the polygon should be the point associated
with the polar coordinate angles, 0 and 2π, respectively.

```python
def  my_polygon(n):
    r=1
    t=np.arange(0,2*np.pi+(2*np.pi/n),(2*np.pi/n))
    x=r*np.sin(t)
    y=r*np.cos(t)  
    plt.plot(x,y)
    plt.title('%i side polygon'%n)
    plt.axis("equal")
    #plt.show()
    return None

plt.figure(figsize = (20,16))
for n in range(3,10,1):
    plt.subplot(3,3,n)
    my_polygon(n)
```

Q9. Write a function with my_poly_plotter(n,x) that plots the polynomials $p_k(x) = x^k$ for $k = 1, . . . , n$. Make sure your plot has axis labels and a title.

```python
def my_poly_plotter(n,x):
    p=1
    for k in range(1,n+1,1):
        p *= x**k
        plt.plot(x,p)
    plt.title('Polynomials upto degree %i'%n)
    return None
n = 5 
x = np.linspace(-1, 1, 200)
my_poly_plotter(n,x)
```

Q10. Assume you have three points at the corner of an equilateral triangle, $P1= (0, 0),P2 = (0.5,√2/2),and P3 = (1, 0)$.   
Generate another set of points pi = (xi, yi ) such that p1 = (0, 0) and pi+1 is the midpoint between pi and P1 with 33% probability, the midpoint between pi and P2 with 33% probability, and the midpoint between pi and P3 with 33% probability.                                                                                                                             
Write a function my_sierpinski(n) that generates the points pi for $i = 1, . . . , n$.                                      
The function should make a plot of the points using blue dots (i.e., "b." as the third argument to the plt.plot function).

```python
def Sierpinski_Triangle(n):
    import random as R
    vertex_coordinates = [[-200.0, -100.0], [0.0, 200.0], [200.0, -100.0]]
    x_coordinates = [vertex_coordinates[0][0]]
    y_coordinates = [vertex_coordinates[0][1]]
    for i in range(n):
        r = R.randrange(3)
        x_coordinates.append((x_coordinates[-1] + vertex_coordinates[r][0]) / 2)
        y_coordinates.append((y_coordinates[-1] + vertex_coordinates[r][1]) / 2)
    fig, ax = plt.subplots(figsize=(10,8))
    ax.scatter(x_coordinates, y_coordinates, s = 0.1)
    ax.set(title=f'Sierpinski triangle for {n} points')
    plt.show()
    return None

k=8   
for i in range(2,k):
    Sierpinski_Triangle(10**i)
```

Q11. Assume you are generating a set of points (xi, yi ) where x1 = 0 and y1 = 0. 

The points (xi, yi ) for i = 2, . . . , n are generated according to the following probabilistic relationship:
    
       With 1% probability:
       xi = 0,
       yi = 0.16yi−1;
       With 7% probability:
       xi = 0.2xi−1 − 0.26yi−1,
       yi = 0.23xi−1 + 0.22yi−1 + 1.6;
       With 7% probability:
       xi =−0.15xi−1 +0.28yi−1,
       yi = 0.26xi−1 + 0.24yi−1 + 0.44;
       With 85% probability:
       xi = 0.85xi−1 + 0.04yi−1,
       yi =−0.04xi−1 +0.85yi−1 +1.6.
       Write a function my_fern(n) that generates the points (xi, yi ) for i = 1, . . . , n and plots them using blue dots.          Also use plt.axis("equal") and plt.axis("off") to make the plot more attractive.

```python
def Barnsley_fern(n):
    x_coo,y_coo = [],[]
    x_coo.append(0)
    y_coo.append(0)
    curr=0
    for i in range(1,n):
        r = randint(1,100)
        if r==1:
            x_coo.append(0)
            y_coo.append(0.16*(y_coo[curr])-1)
        if r>=2 and r<=86:
            x_coo.append(0.85*(x_coo[curr])+0.04*(y_coo[curr]))
            y_coo.append(-0.04*(x_coo[curr])+0.85*(y_coo[curr])+1.6)
        if r>=87 and r<=93:
            x_coo.append(0.2*(x_coo[curr])-0.26*(y_coo[curr]))
            y_coo.append(0.23*(x_coo[curr])+0.22*(y_coo[curr])+1.6)
        if r>=94 and r<=100:
            x_coo.append(-0.15*(x_coo[curr])+0.28*(y_coo[curr]))
            y_coo.append(0.26*(x_coo[curr])+0.24*(y_coo[curr])+0.44)
        curr+=1
    fig, ax = plt.subplots(figsize=(10,10)) 
    plt.title(f'Barnsley Fern for {n} points')
    plt.scatter(x_coo,y_coo,s=0.3,edgecolor='green')
    return None

k=6
for i in range(2,k):
    Barnsley_fern(10**i)
```

Q12. Write a function my_parametric_plotter (x,y,t) where x and y are function objects x(t) and y(t), respectively, and t is a one-dimensional array. The function my_parametric_plotter should produce the curve (x(t ), y(t), t) in a 3D plot. Be sure to give your plot a title and label your axes.

```python
def my_parametric_plotter(F, G, t):
    mpl.rcParams['legend.fontsize'] = 10
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    z = np.linspace(-2, 2, 100)
    ax.plot(F, G, z, label='parametric curve')
    plt.title('My Parametric Plot')
    plt.show()
    return None

f = lambda t: np.sin(t)
g = lambda t: t**2
t = np.linspace(0, 6*np.pi, 100)
my_parametric_plotter(f(t), g(t), t)
```

Q13. Write a function my_surface_plotter(f, x, y, option) where f is a function object f(x,y).
The function my_surface_plotter should produce a 3D surface plot of f(x,y) using plot_surface
if the option used is the string "surface". It should produce a contour plot of f(x,y) if
the option used is the string "contour". Assume that x and y are one-dimensional arrays or lists.
Remember to give the plot a title and label your axes.

```python
def my_surface_plotter(f, x, y, option) :

    x1 = np.arange(-1, 1.1, 0.2)
    y1 = np.arange(-2, 2.1, 0.2)
    X, Y = np.meshgrid(x1, y1)
    Z = f(X,Y)
    if option=="surface":
        fig = plt.figure(figsize = (12,10))
        ax = plt.axes(projection="3d")
        ax.plot_surface(X, Y, Z, cmap = plt.cm.cividis)
        ax.set_title("Surface plot")
        ax.set_xlabel("x", labelpad=20)
        ax.set_ylabel("y", labelpad=20)
        ax.set_zlabel("z", labelpad=20)
        plt.show()
    elif option=="contour":
        fig, ax = plt.subplots(figsize = (12,10))
        ax.contour(X, Y, Z,  cmap=plt.cm.coolwarm)
        ax.set_title("Contour plot")
        ax.set_xlabel("x", labelpad=20)
        ax.set_ylabel("y", labelpad=20)
        plt.show()
    else :
        print ('stop')
    return None

f = lambda x, y: np.cos(y)*np.sin(np.exp(x))
x = np.linspace(-1, 1, 20)
y = np.linspace(-2, 2, 40)

my_surface_plotter(f, x,y, "surface")

my_surface_plotter(f, x,y, "contour")
```
