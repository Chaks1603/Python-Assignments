Q1. The effective 1-D equation for is a dimensionless form is
$$\displaystyle( \frac{d}{d \rho^2} - \frac{l(l+1)}{\rho^2} + \frac{2Z}{\rho} - 2\gamma)u = 0 \qquad \text{ where } \displaystyle\rho = \frac{r}{a}, \gamma=-\frac{E}{\frac{e^2}{a}},a=\frac{\hbar^2}{me^2}=\text{ Bohr Radius}$$

Use Shooting method to find out the eigenvalues ($\gamma$) 

and normalized eigenvectors in 1s,2s, 3s, 3p and 3d states (Take Z=1)

```python
import scipy.integrate as SI
import scipy.optimize as SOP
import numpy as np
import matplotlib.pyplot as plt
```

```python
xmax=10# higher higher excited states
#oe=-1 # set -1 for odd
X=np.linspace(1e-10,xmax,100)
PSI0=[0,1]
z,l=1,0
def f(PSI,x,l,z,e):# function for Schr eqn
    psi, psid=PSI
    dpsi=psid
    dpsid=((l*(l+1)/(x*x))-(2*z/x)+(2*e))*psi
    return [dpsi, dpsid]
def schr(e): # function for bisection
    P=SI.odeint(f,PSI0,X,args=(l,z,e))
    shy =P[-1,0] # 
    return(shy)
ecr=SOP.bisect(schr,0.4,1) # set app  limits
print ('eigenvalue is ', ecr)
X1=np.linspace(1e-10,xmax,500)
P=SI.odeint(f,PSI0, X1,args=(l,z,ecr))
si=P[:,0]
N=1.0/SI.simps(si*si, X1)
si*=np.sqrt(N)
plt.figure(figsize=(20, 8))
plt.plot(X1,si)
plt.title('1s',fontsize=40)
plt.grid()
plt.show()
```

Q3. Instead of Coulombic potential if a Screened coulomb potential $\large\displaystyle V(r) = -\frac{e^2}{r} \exp(-\frac{r}{b})$ is used
then the effective 1-D equation for is a dimensionless form is
$$\large\displaystyle( \frac{d}{d \rho^2} - \frac{l(l+1)}{\rho^2} + \frac{2\exp(-\frac{r}{\lambda})}{\rho} - 2\gamma)u = 0  \qquad \text{ where }b = \lambda a$$
a) Find 1s, 2s, 3s states for $\lambda = 10,20,50$ 

b) compare the eigenvalues of 3s,3p and 3d states with $\lambda = 10$

```python
import scipy.integrate as SI 
import scipy.optimize as SOP 
import numpy as np 
import matplotlib.pyplot as plt 
Xmax=[16,20,80] # higher higher excited states 
PSI0=[0,1]
```

```python
lam=10
#Z=[1,2,4] 
#l=0 
plt.figure(figsize=(14,6))
a=[0.3,0.01,0.001] 
b=[0.6,0.09,0.005] 
for i in range(3):    
    xmax=Xmax[i] 
    X=np.linspace(1e-10,xmax,100) 
    n=0 
    m=0 
    def f(PSI,x,lam,e):# function for Schr eqn 
        psi, psid=PSI 
        dpsi=psid 
        dpsid=(-(2*np.exp(-x/lam))/x+(2*e))*psi
        return [dpsi, dpsid] 
    def schr(e): # function for bisection 
        P=SI.odeint(f,PSI0,X,args=(lam,e)) 
        shy =P[-1,0] #  
        return(shy) 
    n=a[i]
    m=b[i] 
    ecr=SOP.bisect(schr,n,m) # set app  limits 
    print(n,m) 
    print ('eigenvalue is ', ecr) 
    X1=np.linspace(1e-10,xmax,500) 
    P=SI.odeint(f,PSI0, X1,args=(lam,ecr)) 
    si=P[:,0] 
    X=np.linspace(0,xmax,500) 
    N=1.0/SI.simps(si*si, X) 
    si*=np.sqrt(N) 
    plt.plot(X1,si) 
    plt.grid()
    plt.title(r'Screened potential for $\lambda=10$',fontsize=25)
    plt.legend(['$1s$','$2s$','$3s$'],fontsize=30)
plt.show()
```
