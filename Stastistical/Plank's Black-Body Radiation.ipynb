# $$\textbf{Laws of Radiation}$$

Q.Plot Planckâ€™s law for Black Body radiation and compare it with Raleigh-Jeans Law at high temperature and low temperature

$\displaystyle{\text{Plank's law }:    u_{\lambda}\text{d}\lambda = \frac{8\pi hc}{\lambda ^5} \frac{1}{e^{hc/\lambda kT} -1}\text{d}\lambda}$

$\displaystyle{\text{Wein's law }:    u_{\lambda}\text{d}\lambda = \frac{8\pi hc}{\lambda ^5} \frac{1}{e^{hc/\lambda kT} }\text{d}\lambda}$

$\displaystyle{\text{Rayleigh-Jeans law }: u_{\lambda}\text{d}\lambda = \frac{8\pi kT}{\lambda ^4}\text{d}\lambda} $


```python
from pylab import *
import numpy as np
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
plt.rcParams['text.usetex'] = True
plt.rcParams["figure.figsize"] = (16,12)
plt.rcParams['font.size'] = 30
plt.rcParams['legend.fontsize'] = 30
```

```python
h = 6.626e-34              # Plank's const
c = 3.0e+8                 # speed of light
k = 1.38e-23               # Boltzmann constant (joule per kelvin)
wavelengths = np.arange(1e-9, 3e-6, 1e-9) 
```

```python
def Planck(wav, T):
    return (8.*np.pi*h*c)/((wav**5)*(np.exp((h*c)/(wav*k*T))-1.0))     # intensity

M = [ 4000.,5000.,6000.,7000.]

for t in M:
    plt.plot(wavelengths*1e9, Planck(wavelengths, t),label='%i K'%t)
plt.title('Planks Black-Body Radiation')
plt.ylabel(r'Energy Density $u(x)$')
plt.xlabel(r'Wavelengths $\lambda$')
plt.legend()
plt.show()
```

```python
def Wein(wav,T):
    return (8*np.pi*h*c)/(wav**5 * np.exp((h*c)/(wav*k*T)))                    
```

```python
def Rayleigh_Jeans(wav,T):
    return (8*np.pi*k*T)/wav**4 
```

```python
plt.plot(wavelengths*1e9, Planck(wavelengths, 7000.), 'r-', label='Planck')
plt.plot(wavelengths*1e9, Wein(wavelengths, 7000.), 'g--', label='Wein')
plt.plot(wavelengths*1e9, Rayleigh_Jeans(wavelengths, 7000.), 'k--', label='Rayleigh')
plt.ylabel(r'Energy Density $u(x)$')
plt.xlabel(r'Wavelengths $\lambda$')
plt.ylim(0,4e6)                          
plt.legend()
plt.show()  
```

```python
for t in [ 4000.,5500.,7000.]:
    plt.plot(wavelengths*1e6, Planck(wavelengths, t),label='P:%i K'%t)
    plt.plot(wavelengths*1e6, Wein(wavelengths, t), '--',label='W:%i K'%t)
    plt.plot(wavelengths*1e6, Rayleigh_Jeans(wavelengths, t), '--',label='RJ:%i K'%t)
plt.title('Planks Black-Body Radiation')
plt.legend(bbox_to_anchor=(1, 0.7))
plt.ylabel(r'Energy Density $u(x)$')
plt.xlabel(r'Wavelengths $\lambda$')
plt.ylim(0,4e6)   
plt.legend()
plt.show()
```

# $$ \textbf{Radiation Itensity vs Frequency}$$

$\displaystyle{\text{Plank's law }:    u(x) = \frac{x^3}{e^{x} -1}}$

$\displaystyle{\text{Wein's law }:    u(x) = \frac{x^3}{e^{x} }}$

$\displaystyle{\text{Rayleigh-Jeans law }: u(x) = {x^2}}$

$\displaystyle{\text{where } x=\frac{\hbar \omega} {k_B T} \text{ where } \omega=\frac{1}{\nu}}$


```python
def planck(x):
    return x**3/(np.exp(x)-1)
def wein(x):
    return x**3/np.exp(x)
def rayleigh_jeans(x):
    return x**2

v=np.arange(0, 14,0.01)
plt.plot(v,planck(v),label='Planck')
plt.plot(v,wein(v),'--',label="Weins'")
plt.plot(v,rayleigh_jeans(v),'--',label="Rayleigh-Jeans'")
plt.title('The spectral distribution of energy in the blackbody radiation')
plt.ylabel(r'Energy Density $u(x)$')
plt.xlabel(r'frequency $\nu$')
plt.legend()
plt.ylim(0,2)
plt.show()
```
