Q. Plot the following functions with energy at different temperatures  
   a) Maxwell-Boltzmann distribution ,                                                                              
   b)Fermi-Dirac distribution ,                                                                                           
   c) Bose-Einstein distribution
   
$$\large\displaystyle{f(T,\eta)= \frac{1}{e^{(\epsilon-\mu)e/kT} + \eta} }$$   
   
$$ \large\displaystyle{\begin{equation*} \eta = \left\{\begin{array}{rl}  0 & \text{Maxwell-Boltzmann distribution },\\  1 & \text{Fermi-Dirac distribution },\\ -1 & \text{Bose-Einstein distribution }. \end{array} \right.\end{equation*}}$$

```python
from pylab import *
import numpy as np
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
plt.rcParams['text.usetex'] = True
plt.rcParams["figure.figsize"] = (15,8)
plt.rcParams['font.size'] = 20
plt.rcParams['legend.fontsize'] = 20
```

```python
E = np.linspace(-0.5,0.5,1001) # energy range  
e = 1.6e-19                    # electric charge  
k = 1.38e-23                   # Boltzmann constant(joule per kelvin)  
u = 0                          # considering chemeical potential of the substance is zero  
```

```python
def Fn(T,a):  
    return 1/((np.exp(((E-u)*e)/(k*T)))+a)
```

```python
T = [0.0001, 100.,300.,500.,700.]
system = [ 0, 1, -1]
```

```python
for t in T:
    plt.plot(E,Fn(t,0),label='T = %i K'%t)  
plt.ylim(-0.05,2) 
plt.xlim(-0.2,0.6)  
plt.xlabel('Energy(eV)')  
plt.ylabel('$f(E)$')   
plt.legend(loc='best')  
plt.title("Maxwell-Boltzmann distribution for u=0")   
plt.show()
```

```python
for t in T:
    plt.plot(E,Fn(t,1),label='T = %i K'%t)
plt.legend(loc='best')  
plt.ylim(-0.05,1.05)  
plt.xlabel('Energy(eV)')  
plt.ylabel('$f(E)$')  
plt.title("Fermi-Dirac distribution for u=0")
plt.show()
```

```python
E=np.linspace(-1.5,1.5,1001)
for t in [0.001,900,2500]:
    plt.plot(Fn(t,1),E,linewidth=4,label='T = %i K'%t)
plt.legend(loc='best')  
#plt.ylim(-0.05,1.05)  
plt.ylabel('Energy(eV)')  
plt.xlabel('$f(E)$')  
plt.title("Fermi-Dirac distribution for u=0")
plt.show()
```

```python
E_= np.linspace(0,10,1001)
u_= 5      # E_f
S_= ['c-','r-.']
T_= [0.0001,2500]
for s_,t_ in zip(S_,T_):
    plt.plot(E_,0.682E28*Fn(t_,1)*E_**0.5/1e28,s_,linewidth=4,label='T = %i K'%t_ )  
plt.xlabel('Energy(eV)')  
plt.ylabel('$f(E)$')   
plt.legend(loc='best')    
plt.grid()
plt.title("Fermi-Dirac distribution for Metal at u = %0.2f eV"%u)
plt.show()
```

```python
for t in T:
    plt.plot(E,Fn(t,-1),label='T = %i K'%t)   
plt.xlim(-0.1,0.51)  
plt.ylim(-0.05,4)  
plt.xlabel('Energy(eV)')  
plt.ylabel('$f(E)$')  
plt.legend(loc='best')        
plt.title("Bose-Einstein distribution for u=0")
plt.show()
```

```python
for s in system:
    plt.plot(E,Fn(500,s),label='$\eta$ = %.d '%s)  
plt.legend(loc='best')  
plt.ylim(-0.1,4)  
plt.xlim(-0.4,0.4)  
plt.xlabel('Energy(eV)')  
plt.ylabel('$f(E)$')  
plt.title("Temperature = 500 K")
plt.text(-0.09, 3, "MB", fontsize=25, color="C0")   
plt.text(-0.3, 1.1, "FD", fontsize=25, color="C1")  
plt.text(0.03, 3, "BE", fontsize=25, color="C2") 
plt.show()
```

```python
fig = plt.figure(figsize=(28,20))
gs = fig.add_gridspec(3,3)
ax1=fig.add_subplot(gs[0,0])
ax2=fig.add_subplot(gs[0,1])
ax3=fig.add_subplot(gs[0,2])
ax4=fig.add_subplot(gs[1,0])
ax5=fig.add_subplot(gs[2,0])
ax6=fig.add_subplot(gs[1:,1:])

fig.suptitle(r'$f(T,\eta)= \frac{1}{e^{(\epsilon-\mu)e/kT} + \eta}$',fontsize=30)

for t in T:
    ax1.plot(E,Fn(t,0),label='T = %i K'%t)  
ax1.set_ylim(-0.05,2) 
ax1.set_xlim(-0.2,0.6)  
ax1.set_xlabel('Energy(eV)')  
ax1.set_ylabel('$f(E)$')   
ax1.legend(loc='best')  
ax1.set_title("Maxwell-Boltzmann distribution for u=0") 

for t in T:
    ax2.plot(E,Fn(t,1),label='T = %i K'%t)
ax2.legend(loc='best')  
ax2.set_ylim(-0.05,1.05)  
ax2.set_xlabel('Energy(eV)')  
ax2.set_ylabel('$f(E)$')  
ax2.set_title("Fermi-Dirac distribution for u=0")

for t in T:
    ax3.plot(E,Fn(t,-1),label='T = %i K'%t)   
ax3.set_xlim(-0.1,0.51)  
ax3.set_ylim(-0.05,4)  
ax3.set_xlabel('Energy(eV)')  
ax3.set_ylabel('$f(E)$')  
ax3.legend(loc='best')        
ax3.set_title("Bose-Einstein distribution for u=0")

E = np.linspace(-1.5,1.5,1001)
for t in [0.001,900,2500]:
    ax4.plot(Fn(t,1),E,linewidth=4,label='T = %i K'%t)
ax4.legend(loc='best')  
ax4.set_ylabel('Energy(eV)')  
ax4.set_xlabel('$f(E)$')  
ax4.set_title("Fermi-Dirac distribution for u=0")

E_= np.linspace(0,10,1001)
u_= 5      # E_f
S_= ['c-','r-.']
T_= [0.0001,2500]
for s_,t_ in zip(S_,T_):
    ax5.plot(E_,0.682E28*Fn(t_,1)*E_**0.5/1e28,s_,linewidth=4,label='T = %i K'%t_ )  
ax5.set_xlabel('Energy(eV)')  
ax5.set_ylabel('$f(E)$')   
ax5.legend(loc='best')    
ax5.grid()
ax5.set_title("Fermi-Dirac distribution for Metal at u = %0.2f eV"%u)

for s in system:
    ax6.plot(E,Fn(500,s),label='$\eta$ = %.d '%s)  
ax6.legend(loc='best', fontsize=30)  
ax6.set_ylim(-0.1,4)  
ax6.set_xlim(-0.4,0.4)  
ax6.set_xlabel('Energy(eV)')  
ax6.set_ylabel('$f(E)$')  
ax6.set_title("Temperature = 500 K")
ax6.text(-0.09, 3, "MB", fontsize=25, color="C0")   
ax6.text(-0.3, 1.1, "FD", fontsize=25, color="C1")  
ax6.text(0.03, 3, "BE", fontsize=25, color="C2") 

plt.tight_layout()
plt.show()
```
