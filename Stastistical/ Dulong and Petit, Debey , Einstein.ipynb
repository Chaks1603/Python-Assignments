```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
plt.rcParams['text.usetex'] = True
plt.rcParams["figure.figsize"] = (20,12)
plt.rcParams['font.size'] = 30
plt.rcParams['legend.fontsize'] = 30
```

Debyeâ€™s formula for the heat capacity $C_V$ of a solid is $C_V = 9Nkg(u)$, where
$$\displaystyle{g(u)=u^3\int_0^{1/u} \frac{x^4 e^x}{(e^x-1)^2} dx}$$
The terms in this equation are                                                                                             
$\displaystyle{ \qquad\qquad N = \text{number of particles in the solid} \\
  \qquad\qquad k = \text{Boltzmann constant}\\
  \qquad\qquad u = T/\Theta_D \\
  \qquad\qquad T = \text{absolute temperature}\\
  \qquad\qquad \Theta_D = \text{Debye temperature} \\}$
Compute $g(u)$ from $u = 0$ to $1.0$ in intervals of $0.05$ and plot the results.

```python
def f(x):
    if x == 0: 
        return 0.0
    else: 
        return (x**4)*np.exp(x)/(np.exp(x) - 1.0)**2
    
u = np.arange(0.0,1.01,0.05)
g = np.zeros(len(u))

for i in range(len(u)):
    if i != 0:
        I = quad(lambda x:f(x),0,1.0/u[i],limit=200)[0] 
        g[i]= (u[i]**3)*I*3
        
plt.plot(u,g,'--o')
plt.xlabel(r'$u \rightarrow $')
plt.ylabel(r'$g(u) \rightarrow $')
plt.grid(True)
plt.title(r'$\displaystyle{g(u)=u^3\int_0^{1/u} \frac{x^4 e^x}{(e^x-1)^2} dx}$',y=1.1)
plt.show()
```

```python
def f(x):
    if x == 0: 
        return 0.0
    else: 
        return (x**4)*np.exp(x)/(np.exp(x) - 1.0)**2
    
u = np.arange(0.0,1.01,0.05)
Cv = np.zeros(len(u))
N = 6.022 * 10**23
k = 1.380649 * 10**(-23)

for i in range(len(u)):
    if i != 0:
        I = quad(lambda x:f(x),0,1.0/u[i],limit=200)[0] 
        Cv[i]= (u[i]**3)*I*9*N*k
        
plt.plot(u,Cv,'-o')
plt.xlabel(r'$u \rightarrow $')
plt.ylabel(r'$C_V \rightarrow $')
plt.grid(True)
plt.title('Debey heat capacity')
plt.show()
```

Einsteinâ€™s formula for the heat capacity $C_V$ of a solid is $C_V = 3Nkg(u)$, where
$$\displaystyle{g(u)=u^2 \frac{ e^{u}}{(e^{u}-1)^2} \text{ where }u=\frac{1}{x}}$$
The terms in this equation are                                                                                             
$\displaystyle{ \qquad\qquad N = \text{number of particles in the solid} \\
  \qquad\qquad k = \text{Boltzmann constant}\\
  \qquad\qquad u = T/\Theta_D \\
  \qquad\qquad T = \text{absolute temperature}\\
  \qquad\qquad \Theta_E = \text{Einstein temperature} \\}$
  
Compute $g(u)$ from $u = 0$ to $1.0$ in intervals of $0.05$ and plot the results.

```python
def fE(x):
    c = (np.pi/6)**(1/3);
    return  c*c*np.exp(c/x)/(np.exp(c/x)-1)**2
    
u = np.arange(0.0,1.01,0.05)
gE = np.zeros(len(u))

for i in range(len(u)):
    if i != 0:
        #I = quad(lambda u:f(u),0,1.0/u[i],limit=200)[0] 
        gE[i]= (1/u[i])**2*fE(u[i])
plt.plot(u,gE,'-o')
plt.xlabel(r'$u \rightarrow $')
plt.ylabel(r'$g(u) \rightarrow $')
plt.grid(True)
plt.title(r'$\displaystyle{g(u)=u^2 \frac{e^x}{(e^x-1)^2}}$',y=1.1)
plt.show()
```

From Dulong and Petit's law : $\displaystyle{ \\ \ \qquad\displaystyle{ C_V = 3Nk_B => \frac{C_V}{3Nk_B}=1}}$

Debyeâ€™s theory suggest : $\displaystyle{ \\ \qquad\qquad\displaystyle{C_V= 9Nk_B u^3\int_0^{1/u} \frac{x^4 e^x}{(e^x-1)^2} dx=>\frac{C_V}{3Nk_B}= 3 u^3\int_0^{1/u} \frac{x^4 e^x}{(e^x-1)^2} dx}}$

Einsteinâ€™s theory suggest:$\displaystyle{ \\ \qquad\qquad\displaystyle{C_V= 3Nk_Bu^2 \frac{ e^{u}}{(e^{u}-1)^2} =>\frac{C_V}{3Nk_B}= u^2 \frac{ e^{u}}{(e^{u}-1)^2} \text{ where }u=\frac{1}{x}}}$

```python
def fD(x):
    if x == 0: 
        return 0.0
    else: 
        return (x**4)*np.exp(x)/(np.exp(x) - 1.0)**2

def fE(x):
    c = (np.pi/6)**(1/3)
    return  c*c*np.exp(c/x)/(np.exp(c/x)-1)**2
    
u = np.arange(0.0,2.01,0.02)
gD = np.zeros(len(u))
gE = np.zeros(len(u))
gDP= np.ones(len(u))

for i in range(len(u)):
    if i != 0:
        I = quad(lambda x:fD(x),0,1.0/u[i],limit=200)[0] 
        gD[i]= 3*(u[i]**3)*I
        gE[i]= (1/u[i])**2*fE(u[i])
        
plt.plot(u,gD,'-o',label='Debye')
plt.plot(u,gE,'--o',label='Einstein')
plt.plot(u,gDP,'--',label='Dulong-Petit')
plt.xlabel(r'$u \rightarrow $')
plt.ylabel(r'$g(u) \rightarrow $')
plt.legend()
plt.grid(True)
plt.title('heat capacity')
plt.show()
```
