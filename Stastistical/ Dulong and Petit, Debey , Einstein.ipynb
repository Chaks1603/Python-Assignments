Debyeâ€™s formula for the heat capacity $C_V$ of a solid is $C_V = 9Nkg(u)$, where
$$\displaystyle{g(u)=u^3\int_0^{1/u} \frac{x^4 e^x}{(e^x-1)^2} dx}$$
The terms in this equation are                                                                                             
$\displaystyle{ \qquad\qquad\begin{aligned} N &= \text{number of particles in the solid} \\
  \qquad\qquad k &= \text{Boltzmann constant}\\
  \qquad\qquad u &= T/\Theta_D \\
  \qquad\qquad T &= \text{absolute temperature}\\
  \qquad\qquad \Theta_D &= \text{Debye temperature} \\ \end{aligned}}$
  
Compute $g(u)$ from $u = 0$ to $1.0$ in intervals of $0.05$ and plot the results.

```python
def f(x):
    if x == 0: 
        return 0.0
    else: 
        return (x**4)*np.exp(x)/(np.exp(x) - 1.0)**2
    
u = np.arange(0.0,1.01,0.05)
g = np.zeros(len(u))

for i in range(len(u)):
    if i != 0:
        I = quad(lambda x:f(x),0,1.0/u[i],limit=200)[0] 
        g[i]= (u[i]**3)*I*3
        
plt.plot(u,g,'--o')
plt.xlabel(r'$u \rightarrow $')
plt.ylabel(r'$g(u) \rightarrow $')
plt.grid(True)
plt.title(r'$\displaystyle{g(u)=u^3\int_0^{1/u} \frac{x^4 e^x}{(e^x-1)^2} dx}$',y=1.1)
plt.show()
```

Einsteinâ€™s formula for the heat capacity $C_V$ of a solid is $C_V = 3Nkg(u)$, where
$$\displaystyle{g(u)=u^2 \frac{ e^{u}}{(e^{u}-1)^2} \text{ where }u=\frac{1}{x}}$$
The terms in this equation are                                                                                             
$\displaystyle{ \qquad\qquad\begin{aligned} N &= \text{number of particles in the solid} \\
  \qquad\qquad k &= \text{Boltzmann constant}\\
  \qquad\qquad u &= T/\Theta_D \\
  \qquad\qquad T &= \text{absolute temperature}\\
  \qquad\qquad \Theta_E &= \text{Einstein temperature} \end{aligned} }$
  
Compute $g(u)$ from $u = 0$ to $1.0$ in intervals of $0.05$ and plot the results.

```python
def fE(x):
    c = (np.pi/6)**(1/3);
    return  c*c*np.exp(c/x)/(np.exp(c/x)-1)**2
    
u = np.arange(0.0,1.01,0.05)
gE = np.zeros(len(u))

for i in range(len(u)):
    if i != 0:
        #I = quad(lambda u:f(u),0,1.0/u[i],limit=200)[0] 
        gE[i]= (1/u[i])**2*fE(u[i])
plt.plot(u,gE,'-o')
plt.xlabel(r'$u \rightarrow $')
plt.ylabel(r'$g(u) \rightarrow $')
plt.grid(True)
plt.title(r'$\displaystyle{g(u)=u^2 \frac{e^x}{(e^x-1)^2}}$',y=1.1)
plt.show()
```

Debyeâ€™s theory suggest : $\displaystyle{ \\ \qquad\qquad\displaystyle{C_V= 9Nk_B u^3\int_0^{1/u} \frac{x^4 e^x}{(e^x-1)^2} dx=>\frac{C_V}{3Nk_B}= 3 u^3\int_0^{1/u} \frac{x^4 e^x}{(e^x-1)^2} dx}}$

Einsteinâ€™s theory suggest:$\displaystyle{ \\ \qquad\qquad\displaystyle{C_V= 3Nk_Bu^2 \frac{ e^{u}}{(e^{u}-1)^2} =>\frac{C_V}{3Nk_B}= u^2 \frac{ e^{u}}{(e^{u}-1)^2} \text{ where }u=\frac{1}{x}}}$

**Problem:**

Plot the **specific heat of solids** as a function of temperature $T$ according to the following models:

**(a)** **Dulong-Petit law**

**(b)** **Einstein formula** â€” take three different values of the Einstein temperature $\theta_E$

**(c)** **Debye formula** â€” take three different values of the Debye temperature $\theta_D$

Compare the results graphically.

---

**(a) Dulong-Petit Law:**

$$
C_V = 3 R
$$

* Constant value independent of $T$
* $R = 8.314\,\mathrm{J/mol\cdot K}$ (gas constant)


**(b) Einstein Model:**

$$
C_V = 3 R \frac{(\theta_E/T)^2 e^{\theta_E/T}}{(e^{\theta_E/T}-1)^2}
$$

* $\theta_E$ = Einstein temperature
* Consider three values: e.g., $\theta_E = 100, 300, 600\,\mathrm{K}$


**(c) Debye Model:**

$$
C_V = 9 R \left(\frac{T}{\theta_D}\right)^3 \int_0^{\theta_D/T} \frac{x^4 e^x}{(e^x - 1)^2} dx
$$

* $\theta_D$ = Debye temperature
* Consider three values: e.g., $\theta_D = 200, 400, 800\,\mathrm{K}$
* Integral is done numerically


```python
R = 8.314  # J/mol.K
T = np.linspace(1, 1000, 500) # Temperature range

C_DP = 3*R                     # (a) Dulong-Petit Law

theta_E_vals = [100, 300, 600] # (b) Einstein Model

def C_Einstein(T, theta_E):
    x = theta_E / T
    return 3*R * (x**2 * np.exp(x) / (np.exp(x)-1)**2)

theta_D_vals = [100, 300, 600]  # (c) Debye Model

def C_Debye(T, theta_D):
    C = np.zeros_like(T)
    for i, Ti in enumerate(T):
        integrand = lambda x: (x**4 * np.exp(x)) / (np.exp(x)-1)**2
        integral, _ = quad(integrand, 0, theta_D/Ti)
        C[i] = 9*R * (Ti/theta_D)**3 * integral
    return C

plt.figure(figsize=(10,6))

plt.plot(T, [C_DP]*len(T), label='Dulong-Petit', color='black', linewidth=2)

for theta_E in theta_E_vals:
    plt.plot(T, C_Einstein(T, theta_E), label=rf'Einstein $\theta_E$={theta_E}K')

for theta_D in theta_D_vals:
    plt.plot(T, C_Debye(T, theta_D), '--', label=rf'Debye $\theta_D$={theta_D}K')

plt.xlabel('Temperature T (K)')
plt.ylabel('Specific Heat C_V (J/mol.K)')
plt.title('Specific Heat of Solids')
plt.legend(fontsize=14)
plt.grid(True)
plt.show()
```
