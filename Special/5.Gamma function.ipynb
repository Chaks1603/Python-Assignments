## Gamma function :
##    $$\displaystyle{\Gamma(x)= (x-1)! \quad \text{for real integer in +ve axis}
\\ \Gamma(x)=\frac{\Gamma(x+1)}{x}\quad \text{for real integer in -ve axis}}$$

```python
from sympy import *
from sympy import exp
from sympy.plotting import plot
x,t,y,v=var('x,t,y,v')

def G(x): 
    def g(x):
        try:
            y=factorial(x-1)
        except:
            y = 0
        return y
    try:
        v = g(x+1)/x
    except:
        v = 0
    return v

p0=plot(G(x),(x,-6.01,5.01),legend=True,label=r'$P_{0}(x)$',show=False,ylim=[-8,10])   

p0.size=(15,6)
p0.title=('Gamma Function')
p0.show()
```
