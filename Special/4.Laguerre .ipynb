##  Laguerre Function :$ L_{n}(x)= \displaystyle{\sum_{m=0}^n (-1)^m\frac{n!}{(m!)^2(n-m)!}x^m }$

```python
from sympy import *
from sympy.plotting import plot
x,n, m=var('x,n,m')

def L(n, x): 
    return summation(((((-1)**m)*(x**m))*factorial(n))/((factorial(m)**2)*factorial(n-m)),[m,0,n])
    
#print("\t\t\t\t\t\t\t  Laguerre Function")
p0=plot(L(0,x),(x,-2.,7.),legend=True,label=r'$L_{0}(x)$',show=False,ylim=[-4,3])    
for i in range(1,6):
    p0.extend(plot(L(i,x),(x,-2.,7.),legend=True,label=r'$L_{%i}(x)$'%i,show=False))
p0.title=('Laguerre polynomial')
p0.size=(15,8)
p0.show()
```

##  Associated Laguerre Function :$ L_{n}^m (x)= \displaystyle{\sum_{k=0}^n (-1)^k\frac{(n+m)!}{(k!)(n-k)!(k+m)!}x^k}$

```python
from sympy import *
from sympy.plotting import plot
x,n,m,k=var('x,n,m,k')

def L(n,m,x): 
    return summation(((((-1)**k)*(x**k))*factorial(n+m))/(factorial(k)*factorial(n-k)*factorial(m+k)),[k,0,n])
    
#print("\t\t\t\t\t\t\t  Laguerre Function")
m=2
p0=plot(L(0,m,x),(x,-2.,7.),legend=True,label=r'$L_{0}(x)$',show=False,ylim=[-4,5])    
for i in range(1,6):
    p0.extend(plot(L(i,m,x),(x,-2.,7.),legend=True,label=r'$L_{%i}(x)$'%i,show=False))
p0.title=('Associated Laguerre polynomial')
p0.size=(15,8)
p0.show()
```
