##  Laguerre Function :$ L_{n}(x)= \displaystyle{\sum_{m=0}^n (-1)^m\frac{n!}{(m!)^2(n-m)!}x^m }$

```python
from sympy import *
from sympy.plotting import plot
x,n, m=var('x,n,m')

def L(n, x): 
    return summation(((((-1)**m)*(x**m))*factorial(n))/((factorial(m)**2)*factorial(n-m)),[m,0,n])
    
#print("\t\t\t\t\t\t\t  Laguerre Function")
p0=plot(L(0,x),(x,-2.,7.),legend=True,label=r'$L_{0}(x)$',show=False,ylim=[-4,3])    
p1=plot(L(1,x),(x,-2.,7.),legend=True,label=r'$L_{1}(x)$',show=False)  
p2=plot(L(2,x),(x,-2.,7.),legend=True,label=r'$L_{2}(x)$',show=False)
p3=plot(L(3,x),(x,-2.,7.),legend=True,label=r'$L_{3}(x)$',show=False)
p4=plot(L(4,x),(x,-2.,7.),legend=True,label=r'$L_{4}(x)$',show=False)
p5=plot(L(5,x),(x,-2.,7.),legend=True,label=r'$L_{5}(x)$',show=False)
p0.extend(p1)
p0.extend(p2)
p0.extend(p3)
p0.extend(p4)
p0.extend(p5)
p0.title=('Laguerre polynomial')
p0.size=(15,8)
p0.show()
```

##  Associated Laguerre Function :$ L_{n}^m (x)= \displaystyle{\sum_{k=0}^n (-1)^k\frac{(n+m)!}{(k!)(n-k)!(k+m)!}x^k}$

```python
from sympy import *
from sympy.plotting import plot
x,n,m,k=var('x,n,m,k')

def L(n,m,x): 
    return summation(((((-1)**k)*(x**k))*factorial(n+m))/(factorial(k)*factorial(n-k)*factorial(m+k)),[k,0,n])
    
#print("\t\t\t\t\t\t\t  Laguerre Function")
m=2
p0=plot(L(0,m,x),(x,-2.,7.),legend=True,label=r'$L_{0}(x)$',show=False,ylim=[-4,5])    
p1=plot(L(1,m,x),(x,-2.,7.),legend=True,label=r'$L_{1}(x)$',show=False)  
p2=plot(L(2,m,x),(x,-2.,7.),legend=True,label=r'$L_{2}(x)$',show=False)
p3=plot(L(3,m,x),(x,-2.,7.),legend=True,label=r'$L_{3}(x)$',show=False)
p4=plot(L(4,m,x),(x,-2.,7.),legend=True,label=r'$L_{4}(x)$',show=False)
p5=plot(L(5,m,x),(x,-2.,7.),legend=True,label=r'$L_{5}(x)$',show=False)
p0.extend(p1)
p0.extend(p2)
p0.extend(p3)
p0.extend(p4)
p0.extend(p5)
p0.title=('Associated Laguerre polynomial')
p0.size=(15,8)
p0.show()
```
