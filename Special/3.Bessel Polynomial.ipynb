### Bessel Polynomial functions of the first kind of integer orders : $ \displaystyle J_{n} (x) = \sum_{m=0}^\infty  \frac{{(-1)^m}}{{m!}\Gamma{(n+m+1)} }(\frac{x}{2})^{2m+n}$

```python
from sympy import *
from sympy.plotting import plot
x,n,m=var('x,n,m')

def besselJ(n,x):
    return summation((((-1)**m)*((x/2)**(2*m+n)))/(factorial(m)*gamma(m+n+1)),[m,0,oo])

p0=plot(besselJ(0,x),(x,-30,30),label=r'$J_{0}(x)$',legend=True,show=False)

for i in range(1,4):
    p0.extend(plot(besselJ(i,x),(x,-30,30),label=r'$J_{%i}(x)$'%i,legend=True,show=False))
p0.size=(15,6)
p0.title=('Bessel polynomial')
p0.show()

besselJ(n,x)
```

### Bessel Polynomial functions of the first kind of integer orders : $ \displaystyle J_{-n} (x) = \sum_{m=0}^\infty  \frac{{(-1)^m}}{{m!}\Gamma{(m-n+1)} }(\frac{x}{2})^{2m-n}$

```python
from sympy import *
from sympy.plotting import plot
x,n,m=var('x,n,m')

def besselI(n,x):
    return summation((((-1)**m)*((x/2)**(2*m-n)))/(factorial(m)*gamma(m-n+1)),[m,0,oo])

p0=plot(besselI(0,x),(x,-30,30),label=r'$J_{0}(x)$',legend=True,show=False,ylim=[-0.6,1.1])    
for i in range(1,4):
    p0.extend(plot(besselI(i,x),(x,-30,30),label=r'$J_{%i}(x)$'%i,legend=True,show=False))
p0.size=(15,9)
p0.title=('Bessel polynomial')
p0.show()

besselI(n,x)
```

### Bessel Polynomial functions of the second kind of integer orders :

## $$J_{n} (x) = \displaystyle{\sum_{m=0}^{\infty} \frac{(-1)^m}{m!\Gamma(m+n+1)} ({\frac{x}{2}})^{2m+n}}$$
## $$J_{-n} (x) = \displaystyle{\sum_{m=0}^{\infty} \frac{(-1)^m}{m!\Gamma(m-n+1)} ({\frac{x}{2}})^{2m-n}}$$
## $$Y_{n} (x) = \displaystyle{\lim_{\alpha\rightarrow n}\frac{{J_{\alpha} (x)}\cos(\alpha\pi) - {J_{-\alpha} (x)}}{\sin(\alpha\pi)} }$$


```python
from sympy import symbols, cos, sin

x, n = symbols('x n')

def besselY(n, x):
    if n == 0:
        return -cos(x) / x
    elif n == 1:
        return -cos(x) / x**2 - sin(x) / x
    else:
        y_minus2 = -cos(x) / x
        y_minus1 = -cos(x) / x**2 - sin(x) / x

        for k in range(2, n + 1):
            y = (2 * (k - 1) / x) * y_minus1 - y_minus2
            y_minus2 = y_minus1
            y_minus1 = y

        return y

p0=plot(besselY(0,x),(x,0.,10.0),label=r'$Y_{0}(x)$',legend=True,show=False,ylim=[-1.1,0.5])    
for i in range(1,5):
    p0.extend(plot(besselY(i,x),(x,0.,10),label=r'$Y_{%i}(x)$'%i,legend=True,show=False))
p0.size=(15,6)
p0.title=('Bessel polynomial')
p0.show()
```
