### Bessel Polynomial functions of the first kind of integer orders : $ \displaystyle J_{n} (x) = \sum_{m=0}^\infty  \frac{{(-1)^m}}{{m!}\Gamma{(n+m+1)} }(\frac{x}{2})^{2m+n}$

```python
from sympy import *
from sympy.plotting import plot
x,n,m=var('x,n,m')

def besselj(n,x):
    return summation((((-1)**m)*((x/2)**(2*m+n)))/(factorial(m)*gamma(m+n+1)),[m,0,oo])

p0=plot(besselj(0,x),(x,-30,30),label=r'$J_{0}(x)$',legend=True,show=False)    
for i in range(1,4):
    p0.extend(plot(besselj(i,x),(x,-30,30),label=r'$J_{%i}(x)$'%i,legend=True,show=False))
p0.size=(15,6)
p0.title=('Bessel polynomial')
p0.show()

besselj(n,x)
```

### Bessel Polynomial functions of the first kind of integer orders : $ \displaystyle J_{-n} (x) = \sum_{m=0}^\infty  \frac{{(-1)^m}}{{m!}\Gamma{(m-n+1)} }(\frac{x}{2})^{2m-n}$

```python
from sympy import *
from sympy.plotting import plot
x,n,m=var('x,n,m')

def besselj(n,x):
    return summation((((-1)**m)*((x/2)**(2*m-n)))/(factorial(m)*gamma(m-n+1)),[m,0,oo])

p0=plot(besselj(0,x),(x,-30,30),label=r'$J_{0}(x)$',legend=True,show=False,ylim=[-0.6,1.1])    
for i in range(1,4):
    p0.extend(plot(besselj(i,x),(x,-30,30),label=r'$J_{%i}(x)$'%i,legend=True,show=False))
p0.size=(15,9)
p0.title=('Bessel polynomial')
p0.show()

besselj(n,x)
```

### Bessel Polynomial functions of the second kind of integer orders :

## $$J_{n} (x) = \displaystyle{\sum_{m=0}^{\infty} \frac{(-1)^m}{m!\Gamma(m+n+1)} ({\frac{x}{2}})^{2m+n}}$$
## $$J_{-n} (x) = \displaystyle{\sum_{m=0}^{\infty} \frac{(-1)^m}{m!\Gamma(m-n+1)} ({\frac{x}{2}})^{2m-n}}$$
## $$Y_{n} (x) = \displaystyle{\lim_{\alpha\rightarrow n}\frac{{J_{\alpha} (x)}\cos(\alpha\pi) - {J_{-\alpha} (x)}}{\sin(\alpha\pi)} }$$
## $$H_{n} (x) = \displaystyle{J_{n}-iY_{n}}$$

```python
from sympy import *
from sympy.plotting import plot
from sympy import re, im, E, I
x,n,m,a=var('x,n,m,a')

def besselY(a,x):
    
    def Jb(n,x):
        return summation((((-1)**m)*((x/2)**(2*m+n)))/(factorial(m)*gamma(m+n+1)),[m,0,oo])
        
    def Ib(n,x):
        return summation((((-1)**m)*((x/2)**(2*m-n)))/(factorial(m)*gamma(m-n+1)),[m,0,oo])
        
    def Yb(n,x):
        return limit(((Jb(n,x)*cos(n*pi)-Ib(n,x))/sin(n*pi)),n,a)
  
    return Jb(n,x)*E - Yb(n,x)*I

p0=plot(bessely(0,x),(x,-30,30),label=r'$J_{0}(x)$',legend=True,show=False,ylim=[-4,1.1])    
for i in range(1,3):
    p0.extend(plot(bessely(i,x),(x,-30,30),label=r'$J_{%i}(x)$'%i,legend=True,show=False))
p0.size=(15,10)
p0.title=('Bessel polynomial')
p0.show()

display(besselY(a,x))
```
