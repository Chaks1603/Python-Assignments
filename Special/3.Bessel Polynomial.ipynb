### Bessel Polynomial functions of the first kind of integer orders : $ \displaystyle J_{n} (x) = \sum_{m=0}^\infty  \frac{{(-1)^m}}{{m!}\Gamma{(n+m+1)} }(\frac{x}{2})^{2m+n}$

```python
from sympy import *
from sympy.plotting import plot
x,n,m=var('x,n,m')

def besselJ(n,x):
    return summation((((-1)**m)*((x/2)**(2*m+n)))/(factorial(m)*gamma(m+n+1)),[m,0,oo])

p0=plot(besselJ(0,x),(x,-30,30),label=r'$J_{0}(x)$',legend=True,show=False)

for i in range(1,4):
    p0.extend(plot(besselJ(i,x),(x,-30,30),label=r'$J_{%i}(x)$'%i,legend=True,show=False))
p0.size=(15,5)
p0.title=('Bessel polynomial')
p0.show()

besselJ(n,x)
```

Recurrence Relation :

$ \displaystyle J_{n+1}(x)= \frac{2n}{x}J_n(x)-J_{n-1}(x)$

Let, $n\rightarrow n-1$

$ \displaystyle J_{(n-1)+1}(x)= \frac{2(n-1)+1}{x}J_{(n-1)}(x)-J_{(n-1)-1}(x)$

$ \displaystyle => J_{n}(x)= \frac{2(n-1)+1}{x}J_{n-1}(x)-J_{n-2}(x)$

```python
from sympy import *
from sympy.plotting import plot
x,n,J=var('x,n,J')

def J(n, x):
    if n == 0:
        return sin(x) / x
    elif n == 1:
        return sin(x) / x**2 - cos(x) / x
    else:
        return ((2*(n-1)+1)/x) * J(n-1, x) - J(n-2, x)

p0=plot(J(0, x), (x,-20,20),label=r'$J_{0}(x)$',legend=True,show=False)
for n in range(1,5):
     p0.extend(plot(J(n, x), (x,-20,20),label=r'$J_{%i}(x)$'%n,legend=True,show=False))
p0.size=(15,6)
p0.title=('Bessel polynomial')
p0.show()        
```

### Bessel Polynomial functions of the first kind of integer orders : $ \displaystyle J_{-n} (x) = \sum_{m=0}^\infty  \frac{{(-1)^m}}{{m!}\Gamma{(m-n+1)} }(\frac{x}{2})^{2m-n}=(-1)^n J_n(x)$

```python
from sympy import *
from sympy.plotting import plot
x,n,m=var('x,n,m')

def besselJI(n,x):
    return summation((((-1)**m)*((x/2)**(2*m-n)))/(factorial(m)*gamma(m-n+1)),[m,0,oo])

p0=plot(besselJI(0,x),(x,-30,30),label=r'$J_{0}(x)$',legend=True,show=False,ylim=[-0.6,1.1])    
for i in range(1,4):
    p0.extend(plot(besselJI(i,x),(x,-30,30),label=r'$J_{%i}(x)$'%i,legend=True,show=False))
p0.size=(15,5)
p0.title=('Bessel polynomial')
p0.show()

besselJI(n,x)
```

## $$I_{n} (x) = \displaystyle{\sum_{m=0}^{\infty} \frac{1}{m!\Gamma(m+n+1)} \bigg({\frac{x}{2}}\bigg)^{2m+n}}$$

```python
from sympy import *
from sympy.plotting import plot
x,n,m=var('x,n,m')

def besselI(n,x):
    return summation(((1)*((x/2)**(2*m+n)))/(factorial(m)*gamma(m+n+1)),[m,0,oo])

p0=plot(besselI(-1,x),(x,-30,30),ylim=[-10,15],label=r'$I_{-1}(x)$',legend=True,show=False) 

for i in range(0,4):
    p0.extend(plot(besselI(i,x),(x,-30,30),ylim=[-10,15],label=r'$I_{%i}(x)$'%i,legend=True,show=False))
p0.size=(15,6)
p0.title=('Bessel polynomial')
p0.show()

besselI(n,x)
```

### Bessel Polynomial functions of the second kind of integer orders :

## $$J_{n} (x) = \displaystyle{\sum_{m=0}^{\infty} \frac{(-1)^m}{m!\Gamma(m+n+1)} ({\frac{x}{2}})^{2m+n}}$$
## $$ \qquad\quad\qquad J_{-n} (x) = \displaystyle{\sum_{m=0}^{\infty} \frac{(-1)^m}{m!\Gamma(m-n+1)} ({\frac{x}{2}})^{2m-n}}=(-1)^n J_n(x)$$
## $$Y_{n} (x) = \displaystyle{\lim_{\alpha\rightarrow n}\frac{{J_{\alpha} (x)}\cos(\alpha\pi) - {J_{-\alpha} (x)}}{\sin(\alpha\pi)} }$$

## $$ N_{\nu}(x)=\frac{1}{\sin(\nu x)} [J_{\nu}(x)\cos(\nu x) - J_{-\nu}(x)]$$

```python
from sympy import symbols, cos, sin,nan
from sympy.plotting import plot

x, n, pi = symbols('x n pi')

def bessel2Y(n,x):
    return (1/sin(n*pi))*(besselJ(n,x)*cos(n*pi)-((-1)**n*besselJ(n,x)))


bessel2Y(n,x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def Y(n, x):
    if n == 0:
        return -np.cos(x) / x
    elif n == 1:
        return -np.cos(x) / x**2 - np.sin(x) / x
    else:
        return ((2 * n - 1) / x) * Y(n-1, x) - Y(n-2, x)

# Usage
x = np.linspace(1e-12,20, 500)
plt.figure(figsize=(10, 5))
for n in range(0,5):
    plt.plot(x, Y(n,x), label=f'$Y_{n}(x)$')
    plt.ylim(-1,1)
    plt.xlabel('x')
    plt.ylabel('Bessel Functions of the Second Kind')
    plt.title('Bessel Functions of the Second Kind')
    plt.legend()
    plt.grid(True)
```

```python
from sympy import symbols, cos, sin

x, n , Y= symbols('x n Y')

def besselY(n, x):
    if n == 0:
        return -cos(x) / x
    elif n == 1:
        return -cos(x) / x**2 - sin(x) / x
    else:
        return ((2 * n - 1) / x) * besselY(n-1, x) - besselY(n-2, x)


p0=plot(besselY(0,x),(x,1e-12,20.0),label=r'$Y_{0}(x)$',legend=True,show=False,ylim=[-0.51, 0.51])    
for i in range(1,5):
    p0.extend(plot(besselY(i,x),(x,1e-12,20),label=r'$Y_{%i}(x)$'%i,legend=True,show=False))
p0.size=(15,6)
p0.title=('Bessel polynomial')
p0.show()
```
