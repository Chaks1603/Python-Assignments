##  Laguerre Polynomial Function :$ L_{n}(x)= \displaystyle{\sum_{m=0}^n (-1)^m\frac{n!}{(m!)^2(n-m)!}x^m }$

```python
from sympy import *
from sympy.plotting import plot
x,n, m=var('x,n,m')

def L(n, x): 
    return summation(((((-1)**m)*(x**m))*factorial(n))/((factorial(m)**2)*factorial(n-m)),[m,0,n])
    
p0=plot(L(0,x),(x,-5.,20.),legend=True,label=r'$L_{0}(x)$',show=False,ylim=[-10,20])    
for i in range(1,6):
    p0.extend(plot(L(i,x),(x,-5.,20.),legend=True,label=r'$L_{%i}(x)$'%i,show=False))
p0.title=('Laguerre polynomial')
p0.size=(15,8)
p0.show()
```

##  Associated Laguerre Polynomial Function :$ L_{n}^m (x)= \displaystyle{\sum_{k=0}^n (-1)^k\frac{(n+m)!}{(k!)(n-k)!(k+m)!}x^k}$

```python
from sympy import *
from sympy.plotting import plot
x,n,m,k=var('x,n,m,k')

def L(n,m,x): 
    return summation(((((-1)**k)*(x**k))*factorial(n+m))/(factorial(k)*factorial(n-k)*factorial(m+k)),[k,0,n])
    
m=2
p0=plot(L(0,m,x),(x,-2.,10.),legend=True,label=r'$L_{0}(x)$',show=False,ylim=[-4,10])    
for i in range(1,6):
    p0.extend(plot(L(i,m,x),(x,-2.,10.),legend=True,label=r'$L_{%i}(x)$'%i,show=False))
p0.title=('Associated Laguerre polynomial')
p0.size=(15,8)
p0.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def L(n, x):
    if n == 0:
        return 1 + 0. * x
    elif n == 1:
        return 1 - x
    else:
        return ((2 * n - 1 - x) * L(n - 1, x) - (n - 1) * L(n - 2, x)) / n

x = np.arange(-5, 10, 0.001)

plt.figure(figsize=(10, 6))
for n in range(0, 5):
    plt.plot(x, L(n, x), label=f'$L_{n}(x)$')
plt.xlabel('x')
plt.ylim(-5,10)
plt.ylabel('Laguerre Polynomials')
plt.title('Laguerre Polynomials')
plt.legend()
plt.grid(True)
plt.show()
```
