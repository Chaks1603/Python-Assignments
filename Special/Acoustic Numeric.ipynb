```python
from sympy import *
import sympy as sp
from IPython.display import display, Math
from sympy.plotting import plot_parametric
from sympy.plotting import plot, PlotGrid
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
plt.rcParams["figure.figsize"] = (16,7)
plt.rcParams['font.size'] = 30
plt.rcParams['legend.fontsize'] = 20
init_printing()
from sympy import pi,exp,sin,sinh,cosh
from sympy.plotting import plot,plot_parametric,PlotGrid

t=symbols('t',real=True)
x=Function('x')(t)
m=var('m')
f=var('f')
g=var(r'\gamma')
w=var(r'\omega_0')
dx=x.diff(t)
d2x=x.diff(t,2)
```

# $$\displaystyle{\frac{d^2x}{dt^2}+2\gamma\frac{dx}{dt}+\omega^2_o x = 0}$$

```python
eq=Eq(d2x + 2*g*dx + w**2*x , 0)
display(eq)

s1=dsolve(eq)
display(Math('\qquad '+latex(s1))) 
M1=eq.subs({w:1, g:0})
display(M1)
s2=dsolve(M1, ics={x.subs(t,0): 1,x.diff(t).subs(t,0):0})
display(Math('\qquad '+latex(s2))) 
p0=plot(s2.rhs,(t,0,40),label=r'$\gamma=0$',legend=True,show=False)

for i in [0.2,1,2.5]:
    M=eq.subs({w:1,f:1,g:i})
    display(M)
    s2=dsolve(M, ics={x.subs(t,0): 1,x.diff(t).subs(t,0):0})
    display(Math('\qquad '+latex(s2)))     
    p0.extend(plot(s2.rhs,(t,0,40),label=r'$\gamma= {}$'.format(i),legend=True,show=False))
p0.title=(r'${\frac{d^2x}{dt^2}+2\gamma \frac{dx}{dt}+\omega^2_o x=f}$')
p0.show()
```

```python
display(eq)
display(Math('\qquad '+latex(s1))) 

M1=eq.subs({w:0, g:1})
display(M1)
s2=dsolve(M1, ics={x.subs(t,0): 1,x.diff(t).subs(t,0):0})
display(Math('\qquad '+latex(s2))) 
p0=plot(s2.rhs,(t,0,30),label=r'$\gamma=0$',legend=True,show=False)

for i in [0.2,1,2.5]:
    M=eq.subs({w:i, g:1})
    display(M)
    s2=dsolve(M, ics={x.subs(t,0): 1,x.diff(t).subs(t,0):0})
    display(Math('\qquad '+latex(s2)))   
    p0.extend(plot(s2.rhs,(t,0,30),label=r'$\omega=$ {}'.format(i),legend=True,show=False))
p0.size=(15,6)
p0.title=(r'${\frac{d^2x}{dt^2}+2\gamma \frac{dx}{dt}+\omega^2_o x=0}$')
p0.show()
```

# $$\displaystyle{\frac{d^2x}{dt^2}+2\gamma\frac{dx}{dt}+\omega^2_o x = f}$$

```python
eqF=Eq(d2x+2*g*dx+w**2*x,f)
display(eqF)

sf1=dsolve(eqF)
display(Math('\qquad '+latex(sf1)))

M1=eqF.subs({w:1,f:1,g:0})
display(M1)
s2=dsolve(M1, ics={x.subs(t,0): 0, x.diff(t).subs(t,0):0})
display(Math('\qquad '+latex(s2))) 
p0=plot(s2.rhs,(t,0,40),label=r'$\gamma=0$',legend=True,show=False)
p1=plot_parametric((s2.rhs/t,s2.rhs),(t,0,30),show=False,label=r'$\gamma=0$',legend=True,
                   axis=True,title='the x-y phase track',
                   xlabel='x',ylabel='y')
for i in [0.2,1,2.5]:
    M=eqF.subs({w:1,f:1,g:i})
    display(M)
    s2=dsolve(M, ics={x.subs(t,0): 0,x.diff(t).subs(t,0):0})
    display(Math('\qquad '+latex(s2)))     
    p0.extend(plot(s2.rhs,(t,0,40),label=r'$\gamma= {}$'.format(i),legend=True,show=False))
    p1.extend(plot_parametric((s2.rhs/t,s2.rhs),(t,0,30),label=r'$\gamma={}$'.format(i),legend=True,show=False))
p0.title=(r'${\frac{d^2x}{dt^2}+2\gamma \frac{dx}{dt}+\omega^2_o x=f}$')
p2=PlotGrid(1,2, p0, p1, size=(20,10),show=False)
p2.show()
```

# $$\displaystyle{\frac{d^2x}{dt^2}+2\gamma\frac{dx}{dt}+\omega^2_o x = f\sin\omega t}$$

```python
eqF=Eq(d2x+2*g*dx+w**2*x,f*sin(0.98*w*t))
display(eqF)

sf1=dsolve(eqF)
display(Math('\qquad '+latex(sf1))) 

M1=eqF.subs({w:2,f:1,g:0.05})
display(M1)
s2=dsolve(M1, ics={x.subs(t,0): 0, x.diff(t).subs(t,0):0})
display(Math('\qquad '+latex(s2))) 
p0=plot(s2.rhs,(t,0,100),label=r'$x$',legend=True,show=False)
F=1*sin(0.98*2*t)
p0.extend(plot(F,(t,0,100),label=r'$f$',legend=True,show=False))
p0.size=(8,6)
p0.title=(r'${\frac{d^2x}{dt^2}+2\gamma \frac{dx}{dt}+\omega^2_o x=f\sin\omega t}$')

p1=plot_parametric((s2.rhs,F),(t,0,60),show=False,
                   axis=True,title='the x-y phase track',
                   xlabel='x',ylabel='y')

p2=PlotGrid(1,2, p0, p1, size=(20,10),show=False)
p2.show()
```

$$\displaystyle{m\ddot{x}_1=-\frac{mg}{l}x_1+k(x_2-x_1)=>\ddot{x}_1=-\frac{g}{l}x_1+\frac{k}{m}(x_2-x_1)}$$

$$\displaystyle{m\ddot{x}_2=-\frac{mg}{l}x_2-k(x_2-x_1)=>\ddot{x}_2=-\frac{g}{l}x_2-\frac{k}{m}(x_2-x_1)}$$

```python
m,g,l,k=var(r'm,\gamma,l,k')
t=symbols('t',real=True)
x1=Function(r'x_1')(t)
x2=Function(r'x_2')(t)
dx1=x1.diff(t,2)
dx2=x2.diff(t,2)

eq1=Eq(m*dx1,-m*g*x1/l+k*(x2-x1))
eq2=Eq(m*dx2,-m*g*x2/l-k*(x2-x1))
display(eq1,eq2)

s1=dsolve([eq1,eq2])
print('\n general soln. :\n')
display(*s1)

M1=eq1.subs({l : 2.1,g : 9.8,m : 0.3,k : 0.2})
M2=eq2.subs({l : 2.1,g : 9.8,m : 0.3,k : 0.2})
s2=dsolve([M1,M2], ics={x1.subs(t,0): 1.,x1.diff(t).subs(t,0):0.,
                        x2.subs(t,0): 0.,x2.diff(t).subs(t,0):0.})
display(*s2)
P1=plot_parametric((t,s2[0].rhs),(t,s2[1].rhs), (t, 0, 80),show=False)

P2=plot_parametric((s2[0].rhs,s2[1].rhs),(t,0,50),show=False)
P3=PlotGrid(1,2, P1, P2, size=(20,10),show=False)
P3.show()
```
