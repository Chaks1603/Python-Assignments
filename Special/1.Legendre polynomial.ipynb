### Legendre polynomial
$P_{n} (x)=\displaystyle{\sum_{m=0}^{n/2}  (-1)^m \frac{(2n-2m)!}{2^n m! (n-m)! (n-2m)!} x^{n-2m} }$ 

```python
from sympy import *
from sympy.plotting import plot
x,n, m=var('x,n,m')

def P(n, x): 
    return summation(((((-1)**m)*((x)**(n-2*m)))*factorial((2*n)-(2*m)))/
                     ((2**n)*factorial(m)*factorial(n-m)*factorial(n-2*m)),[m,0,n])

p0=plot(P(0,x),(x,-1,1),legend=True,label=r'$P_{0}(x)$',show=False)    
p1=plot(P(1,x),(x,-1,1),legend=True,label=r'$P_{1}(x)$',show=False)  
p2=plot(P(2,x),(x,-1,1),legend=True,label=r'$P_{2}(x)$',show=False)
p3=plot(P(3,x),(x,-1,1),legend=True,label=r'$P_{3}(x)$',show=False)
p4=plot(P(4,x),(x,-1,1),legend=True,label=r'$P_{4}(x)$',show=False)
p5=plot(P(5,x),(x,-1,1),legend=True,label=r'$P_{5}(x)$',show=False)
p0.extend(p1)
p0.extend(p2)
p0.extend(p3)
p0.extend(p4)
p0.extend(p5)
p0.size=(15,6)
p0.title=('Legendre polynomial')
p0.show()

P(n,x)
```

### using recursion relation : 
$P_{0}(x)=1 \qquad P_{1}(x)=x \qquad P_{n}(x)=\frac{(2n-1)xP_{n-1}(x)-P_{n-2}(x)}{n}$

```python
from sympy import *
from sympy.plotting import plot
x,n, p=var('x,n,p')

def P(n, x): 
    if(n == 0):
        return 1 # P0 = 1
    elif(n == 1):
        return x # P1 = x
    else:
        return (((2 * n)-1)*x * P(n-1, x)-(n-1)*P(n-2, x))/float(n)

p0=plot(P(0,x),(x,-1,1),legend=True,label=r'$P_{0}(x)$',show=False)    
p1=plot(P(1,x),(x,-1,1),legend=True,label=r'$P_{1}(x)$',show=False)  
p2=plot(P(2,x),(x,-1,1),legend=True,label=r'$P_{2}(x)$',show=False)
p3=plot(P(3,x),(x,-1,1),legend=True,label=r'$P_{3}(x)$',show=False)
p4=plot(P(4,x),(x,-1,1),legend=True,label=r'$P_{4}(x)$',show=False)
p5=plot(P(5,x),(x,-1,1),legend=True,label=r'$P_{5}(x)$',show=False)
p0.extend(p1)
p0.extend(p2)
p0.extend(p3)
p0.extend(p4)
p0.extend(p5)
p0.size=(15,6)
p0.title=('Legendre polynomial')
p0.show()
```

## Associated Legender polynomial
## $\displaystyle{P_{l}^m (x)=\frac{1}{2^l l!}(1-x^2)^{m/2} \sum_{r=m}^l \frac{(l+m)! (l!)^2 (x+1)^{l-r} (x-1)^{r-m}}{r! (l+m+1)! (l-r)! (r-m)!}} $

```python
from sympy import *
from sympy.plotting import plot
x,l,m,r=var('x,l,m,r')

def P(l,m,x): 
    return ((((1-x**2)**(m/2))/((2**l)*factorial(l)))*summation((factorial(l+m)*(factorial(l)**2)*
                ((x+1)**(l-r))*((x-1)**(r-m)))/(factorial(r)*factorial(l+m+1)*factorial(l-r)*factorial(r-m)) ,[r,m,l]))
m=1
p1=plot(P(1,m,x),(x,-1,1),legend=True,label=r'$P_{1}^1(x)$',show=False, ylim=[-0.06,0.20])    
p2=plot(P(2,m,x),(x,-1,1),legend=True,label=r'$P_{2}^1(x)$',show=False)  
p3=plot(P(3,m,x),(x,-1,1),legend=True,label=r'$P_{3}^1(x)$',show=False)
p4=plot(P(4,m,x),(x,-1,1),legend=True,label=r'$P_{4}^1(x)$',show=False)
p5=plot(P(5,m,x),(x,-1,1),legend=True,label=r'$P_{5}^1(x)$',show=False)
#p6=plot(P(6,m,x),(x,-1,1),legend=True,label=r'$P_{6}^1(x)$',show=False)
p1.extend(p2)
p1.extend(p3)
p1.extend(p4)
p1.extend(p5)
#p1.extend(p6)
p1.size=(15,8)
p1.title=('Associated Legendre polynomial')
p1.show()
```
