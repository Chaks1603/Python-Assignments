### Hermite polynomial using recursion relation : 
$H_{0}(x)=1 \qquad H_{1}(x)=2x \qquad H_{n}(x)=2xH_{n-1}(x)-2(n-1)H_{n-2}(x)$

```python
from sympy import *
from sympy.plotting import plot
x,n=var('x,n')

def H(n, x): 
    if(n == 0):
        return 1.0 + 0.*x # P0 = 1
    elif(n == 1):
        return 2.*x # P1 = x
    else:
        return (2.*x*H(n-1, x)-2*(n-1)*H(n-2, x))
    
#print("\t\t\t\t\t\t\t  Hermitian Function")
p0=plot(H(0,x),(x,-3.,3.),legend=True,label=r'$H_{0}(x)$',show=False,xlim=[-3,3], ylim=[-25,15])    
p1=plot(H(1,x),(x,-3.,3.),legend=True,label=r'$H_{1}(x)$',show=False)  
p2=plot(H(2,x),(x,-3.,3.),legend=True,label=r'$H_{2}(x)$',show=False)
p3=plot(H(3,x),(x,-3.,3.),legend=True,label=r'$H_{3}(x)$',show=False)
p4=plot(H(4,x),(x,-3.,3.),legend=True,label=r'$H_{4}(x)$',show=False)
#p5=plot(H(5,x),(x,-3.,3.),legend=True,label=r'$H_{5}(x)$',show=False)
p0.extend(p1)
p0.extend(p2)
p0.extend(p3)
p0.extend(p4)
#p0.extend(p5)
p0.title=('Hermite polynomial')
p0.size=(15,8)
p0.show()
```

##  Hermite Function :$ H_{n}(x)= \displaystyle{\sum_{m=0}^n (-1)^m\frac{n!}{(m!)(n-2m)!}(2x)^{(n-2m) }}$

```python
from sympy import *
from sympy.plotting import plot
x,n, m=var('x,n,m')

def H(n, x): 
    return summation(((((-1)**m)*((2*x)**(n-2*m)))*factorial(n))/((factorial(m))*factorial(n-2*m)),[m,0,n])
    
p0=plot(H(0,x),(x,-3.,3.),legend=True,label=r'$H_{0}(x)$',show=False,xlim=[-3,3], ylim=[-25,15])    
p1=plot(H(1,x),(x,-3.,3.),legend=True,label=r'$H_{1}(x)$',show=False)  
p2=plot(H(2,x),(x,-3.,3.),legend=True,label=r'$H_{2}(x)$',show=False)
p3=plot(H(3,x),(x,-3.,3.),legend=True,label=r'$H_{3}(x)$',show=False)
p4=plot(H(4,x),(x,-3.,3.),legend=True,label=r'$H_{4}(x)$',show=False)
p0.extend(p1)
p0.extend(p2)
p0.extend(p3)
p0.extend(p4)
p0.title=('Hermite polynomial')
p0.size=(15,8)
p0.show()

H(n,x)
```
