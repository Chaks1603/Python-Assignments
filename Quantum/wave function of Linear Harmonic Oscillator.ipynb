## Plot of wave function of Linear Harmonic Oscillator

let $\large \alpha = \frac{m \omega}{\hbar}$

$$ \Psi_n = {\frac{1}{\sqrt{{2^n}{n!}}}} {(\frac{\alpha}{\pi})}^{\frac{1}{4}} exp({-} \frac{(\sqrt{\alpha}x)^2}{2}) H_n(\sqrt{\alpha}x)$$

consider unit $\alpha = 1$

```python
from sympy import *
from sympy.abc import x

psi0=((1/pi)**(1/4))*exp(-x**2/2)*hermite(0,x)
p0=plot(psi0,(x,-5,5),ylabel='$\psi$',label='n=0',legend='True',line_color='r',show=False)

psi1=1/(sqrt(2)*(1/pi)**(1/4))*exp(-x**2/2)*hermite(1,x)
p1=plot(psi1,(x,-5,5),label='n=1',legend='True',line_color='g',show=False)

p0.extend(p1)

p0.show()
```

```python
psi=(psi0+psi1)/sqrt(2)
plot(psi,(x,-5,5),ylabel='$\psi$',yfontsize=16)
```

```python
plot(psi**2,(x,-5,5),ylabel='$\psi^2$',yfontsize=16)
```

$$ \Psi_n = {\frac{1}{\sqrt{{2^n}{n!}}}} {(\frac{\alpha}{\pi})}^{\frac{1}{4}} exp({-} \frac{(\sqrt{\alpha}x)^2}{2}) H_n(\sqrt{\alpha}x)$$

```python
a=1
p0=plot(pot(0,a),(x,-5,5),ylabel='$\psi$',label='n=0',legend='True',show=False)
p1=plot(pot(1,a),(x,-5,5),ylabel='$\psi$',label='n=1',legend='True',show=False)
p4=plot(pot(4,a),(x,-5,5),ylabel='$\psi$',label='n=4',legend='True',show=False)
p0.extend(p1)
p0.extend(p4)
p0.show()
```
