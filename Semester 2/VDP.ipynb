## For a Lienard System , Van der pol oscillator eqn.
## $${\ddot{x}+\mu(x^2-1)\dot{x}+x=0} $$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

mu = [0.0,0.5,1.0,1.5,2.0,2.5,3.0, 3.5, 4.0]
X0 = [1.0, 2.0]

def vanderpol(X, t, mu= 4.):
    x = X[0]
    y = X[1]
    dxdt = y
    dydt = mu * (1.0 - x**2) * y - x
    return np.array([dxdt, dydt])

def solve(X0, mu, t_start=0, t_end=102.6, t_num=5000):
    t = np.linspace(t_start, t_end, t_num)
    sol = odeint(vanderpol, X0, t, args=(mu,))
    x = sol[:, 0]
    y = sol[:, 1]
    return [x, y, t]

plt.figure(figsize=(16,15))
for mu in mu:
    [x, y, t] = solve(X0,mu)
    plt.plot(x,y,lw=0.9,label=r'$\mu=%.1f$'%mu)
    
plt.grid()
plt.legend(handletextpad=0,loc='upper left',fontsize=20)
plt.xlabel(r'$\Rightarrow$ x',fontsize=20)
plt.ylabel(r'$\Rightarrow \frac{dx}{dt}$',fontsize=20) 
plt.suptitle(r'Van der Pol $\frac{d^2x}{dt^2}-\mu(1-x^2)\frac{dx}{dt}+x=0 $',fontsize=30,y=0.94,fontweight='bold')
plt.show()
```
