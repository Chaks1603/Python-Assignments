 Q1. Show that for $\large \displaystyle{\lim_{\sigma \to 0} \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\pi)^2}{2 {\sigma}^2}} = \delta(x-\pi)} $ :  $$\int \cos x\delta(x-\pi) dx = -1$$

```python
import numpy as np
import scipy.integrate as si
import matplotlib.pyplot as plt
%matplotlib inline
sigma=10
E,s=[],[]
def f(x):
    return np.cos(x)*(1/(sigma*(2*np.pi)**0.5))*np.exp(-(x-np.pi)**2/(2*(sigma)**2))
for i in range(10):
    a=si.quad(lambda x:f(x),np.pi-5*(sigma),np.pi+5*(sigma),limit=200)[0]
    sigma=sigma*0.5
    E.append(a)
    s.append(sigma)
    plt.plot(s,E,'b-')
    plt.xlabel(r'$ \sigma \rightarrow $',fontsize=16)
plt.show()
```

 Q3. Plot the following function in the domain $(-4\pi \le x \le 4\pi) :$ $ \\ $ $$\large f(x)=\tan^{-1}x \textrm{  for  } 0\le x <2\pi \textrm{  and  } f(x+2\pi)=f(x)$$ $\\ $
     Reconstruct this function with its **Fourier Components**.

```python
import numpy as np
import scipy.integrate as si
import matplotlib.pyplot as plt
from pylab import *
%matplotlib inline

q1=[(-5*np.pi),(-4*np.pi),(-3*np.pi),(-2*np.pi),(-np.pi),
    0,np.pi,(2*np.pi),(3*np.pi),(4*np.pi),(5*np.pi)]
q2=[r'${-5\pi}$',r'${-4\pi}$',r'${-3\pi}$',r'${-2\pi}$',r'${-\pi}$',
    r'$0$',r'${\pi}$',r'${2\pi}$',r'${3\pi}$',r'${4\pi}$',r'${5\pi}$']

def f(X):
    return np.arctan(X%(2*np.pi))

x=np.linspace(-5*np.pi,5*np.pi,100000)
y=f(x)

fig = plt.figure(figsize=(20,5))
a = fig.add_subplot(111)
ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
ax.plot(x,y,'c-')
ax.legend(['$tan^{-1} x$'],fontsize=15)
xlim(1.1*x.min(), 1.1*x.max())
xticks(q1,q2)
xlabel(r'$ x \rightarrow $', fontsize=20)
ax=plt.title(r'$tan^{-1}(x) $ plot',fontsize=20)
plt.show()
```

 For a function of fourier
 $$ f(x) = \frac{1}{2}a_o + \sum_{n=1}^\infty a_n \cos nx + \sum_{n=1}^\infty b_n \sin nx $$
where,
 $ \displaystyle\qquad a_o = \frac{1}{2\pi} \int_{0}^{2\pi} f(x) \text{ d}x \\ $ 
 $ \displaystyle\qquad a_n = \frac{1}{\pi} \int_{0}^{2\pi} f(x) \cos nx \text{ d}x \\ $ 
 $ \displaystyle\qquad b_n = \frac{1}{\pi} \int_{0}^{2\pi} f(x) \sin nx \text{ d}x \text{ for n = 1,2,3,4.....}$

```python
n=0
maxn=20  # intervals ; ** just like a fine knob beside a core knob
Ao=(1/(2*np.pi))*si.quad(lambda x:np.cos(n*x)*f(x),0,2*np.pi,limit=200)[0]
A,B=[0],[0]

for n in range(1,maxn):
    An=1/(np.pi)*si.quad(lambda x:np.cos(n*x)*f(x),0,2*np.pi,limit=200)[0]
    A.append(An)
    Bn=1/(np.pi)*si.quad(lambda x:np.sin(n*x)*f(x),0,2*np.pi,limit=200)[0]
    B.append(Bn)    
#print('Ao :',Ao,'\nAn :',A,'\nBn :',B)
F=Ao*np.ones(len(x))    
for n in range(maxn):
    F += A[n]*np.cos(n*x) + B[n]*np.sin(n*x)
    print(F)
fig = plt.figure(figsize=(20,6))
a = fig.add_subplot(111)
ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))    
ax.plot(x,y,'c-',linewidth=4)
ax.plot(x,F,'k-.',linewidth=2)
xlim(1.1*x.min(), 1.1*x.max())
xticks(q1,q2)
xlabel(r'$ x \rightarrow $', fontsize=20)
ax=plt.legend(['$tan^{-1} x$','fourier'],fontsize=15)
ax=plt.title(r'Fourier components of $tan^{-1} x $  plot for $%0.f$ itterations'%(maxn),fontsize=20)
plt.show()
```

3. Plot the Fourier integral of the function:
$$ f(x)=x\exp{(-\frac{x^2}{2\sigma^2})}$$

Plot the imaginary part as a function of $k$ for three values of $\sigma$.

Show that as increases the momentum uncertainty decreases. Also show that the real parts are zero in all cases.

```python
import numpy as np
import scipy.integrate as si
from scipy.linalg import norm
import matplotlib.pyplot as plt

def f(x,k,s):
    return x*np.exp(-(x**2/(2*s**2)))*np.cos(k*x)
def g(x,k,s):
    return x*np.exp(-(x**2/(2*s**2)))*np.sin(k*x)

k=np.arange(-1,1,0.001)
sigma=[3*np.pi,4*np.pi,5*np.pi]

Real=[[si.quad(f,-np.inf,np.inf,limit=1000,args=(i,j))[0] for i in k] for j in sigma]
Imag=[[si.quad(g,-np.inf,np.inf,limit=1000,args=(i,j))[0] for i in k] for j in sigma] 

plt.subplots(figsize=(16,12))    
for i in range(len(sigma)):
    plt.plot(k,Imag[i])
    plt.xlabel("X")
    plt.ylabel("|Imaginary F(X)|")
    plt.legend(['$\sigma_1$','$\sigma_2$','$\sigma_3$'],fontsize=20)
    print(r'Real part norm for sigma_{%i} is :'%(i),norm(Real[i]))
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import numpy as np
plt.style.use("seaborn-poster")
%matplotlib inline
```

```python
def DFT(x):
    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    X = np.dot(e, x)
    return X
```

```python
sr = 96     # sampling rate
ts = 1.0/sr # sampling interval
t = np.arange(0,1,ts)
x=0
a = 1.
x += np.sin(2*np.pi*a*t)

a = 10.
x += np.sin(2*np.pi*a*t)

a = 100.
x += np.sin(2*np.pi*a*t)

X = abs(DFT(x))
# calculate the frequency
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T
fig,(ax1,ax2)=plt.subplots(1,2,figsize = (18, 6))
ax1.plot(t, x, "r")
ax1.set_ylabel("Amplitude")
ax1.set_title('Sample data')
ax2.stem(freq, X, "b", markerfmt=" ", basefmt="-b")
ax2.set_title('Power Spectrum')
ax2.set_xlabel("Freq (Hz)")
ax2.set_ylabel("DFT Amplitude |X(freq)|")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fft
w=1           # signal frequency
Tp=2*np.pi/w  # signal time period
T=1000*Tp     # total time period
k=10          # intevals
N=2**k        # total no. of data points
Nh=int(N/2)   # sampling the spectrum into equal two parts
X,h=np.linspace(0,T,N, retstep=True)
wo=2*np.pi/T  # frequency resolution
wmax=wo*Nh    # max frequency 
print ('sampling :',Nh,'\nfreq res. :', wo,'\nmaxn. frequency :', wmax)
Y=np.sin(w*X) + np.sin(np.pi*X)
YP=np.abs(fft.fft(Y))
WP=np.arange(0,wmax,wo)  # taking the positive side
WN=np.arange(-wmax,0,wo) # taking the negative side
fig,(ax,ax1,ax2)=plt.subplots(1,3,figsize=(16,6))
plt.suptitle('Power Spectrum',y=0.99,fontsize=30)
ax.plot(X[:100],Y[:100],linewidth=3)
ax.set_title('Sample data',fontsize=20)
ax1.plot(WP,YP[:Nh])
ax1.set_title('+ve frequencies',fontsize=20)
ax2.plot(WN,YP[Nh:])
ax2.set_title('-ve frequencies',fontsize=20)
plt.show()
```
