 1. Generate 1 M uniform variate. Use transformation method $ y = −\frac{log(x)}{β}$. Generate the histogram.                        Fit the histogram with $f(y, A, λ) = A exp (−λy)$. Compare the values of β and λ.

```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
from scipy.optimize import curve_fit

def f(x,A,l):
    y=A*np.exp(-(l*x))
    return (y)

x= np.random.random(10**6)
beta=5
z=-(np.log(x)/beta)
z

c,b,p=plt.hist(z,50,ec='red')
print('count:\n',c,'\n\n binborder:\n',b)
plt.show()
bcenter=(b[:-1]+b[1:])/2
plt.plot(bcenter,c, 'o')
p,s=curve_fit(f,bcenter,c)
print('\tβ: \t\t λ:\n',p)
plt.plot(bcenter, f(bcenter, *p))
plt.show()

bcenter.size, c.size
```

2. Generate 1 M Gaussian variate by averaging over N trials. Fit the histogram of the gaussian random numbers thus created by
 $$ f(y, A, μ, σ) = A exp(−\frac{(y−μ)^2}{σ^2})$$
    Study the variation of σ with N.

```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
from scipy.optimize import curve_fit

N=[10,20,30]

def f(x,A,mu,sig):
    y=A*np.exp(-(x-mu)**2/(2*sig**2))
    return(y)
for i in N:
    R= np.random.rand(i,10**6)
    B=np.mean(R,0)
    B
    c,b,p=plt.hist(B,50,ec='red')
    #print('count',c,'\n\n binborder',b)
    plt.show()
    bcenter=(b[:-1]+b[1:])/2
    plt.plot(bcenter,c, 'o')

    p,s=curve_fit(f,bcenter,c,maxfev=10000)
    A,mu,sig=p
    print('μ =',mu ,'\nσ =',sig )
    print('\tf:\t\tbcenter:\tc:\n',p)
    plt.plot(bcenter, f(bcenter, *p))
    plt.show()
```
