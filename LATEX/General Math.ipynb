## Statistics 

Mean $ \displaystyle\bar{x}=\frac{\sum{x_{\tiny{i}}}}{n}=\frac{ \sum{ f_{\tiny{i}} x_{\tiny{i}} } } {\sum f_{\tiny{i}}}$

$\displaystyle \begin{equation*} \text{Median} = \left\{\begin{array}{rl} \frac{(n+1)}{2}th\text{ term} & \text{if } n \text{ is odd}, \\  \left[\frac{\frac{n}{2}th\text{ term} + \frac{(n+1)}{2}th\text{ term}}{2} \right] & \text{if } n \text{ is even}. \end{array} \right. \end{equation*} = l+\frac{\frac{n}{2}-C}{f} \times d$ 

Mode $ = \displaystyle l+\frac{f_m-f_{m-1}}{2f_m-f_{m-1}-f_{m+1}}\times d $

Mode = 3Median - 2Mean

Mean - Mode = 3 ( Mean - Median )

variance $=\begin{equation*} var(x)=\sigma^2 = \displaystyle\frac{\displaystyle \sum_{i=1}^n (x_{\tiny{i}} - \bar{x})^2}{n} = \displaystyle\frac{\displaystyle \sum_{i=1}^n f_{\tiny{i}}(x_{\tiny{i}} - \bar{x})^2}{\displaystyle \sum_{i=1}^n x_{\tiny{i}}} \end{equation*}$

Standard Deviation = $ \displaystyle\sigma = \sqrt{\text{variance}} = \sqrt{\frac{\sum(x_{\tiny{i}} - \bar{x})^2}{n}}$
