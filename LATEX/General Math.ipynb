## Profit and Loss

$Profit \ =S.P.-C.P.$

$Profit \ \%$ on $C.P.= \frac{S.P.-C.P.}{C.P.}\times 100=(\frac{S.P.}{C.P.}-1)\times 100$ 

$Profit \ \%$ on $S.P.= \frac{S.P.-C.P.}{S.P.}\times 100=(1-\frac{C.P.}{S.P.})\times 100$ 

$Loss \ = C.P.-S.P.$

$Loss \ \%$ on $C.P.= \frac{C.P.-S.P.}{C.P.}\times 100=(1-\frac{S.P.}{C.P.})\times 100$

$Loss \ \%$ on $S.P.= \frac{C.P.-S.P.}{S.P.}\times 100=(\frac{C.P.}{S.P.}-1)\times 100$

$C.P.=\frac{100}{100+\text{ profit }\% }\times S.P. \ or, \ S.P.=\frac{100+\text{ profit }\%}{100 }\times C.P. \qquad\because \ S.P.>C.P.$

$C.P.=\frac{100}{100-\text{ loss }\% }\times S.P. \ or, \  S.P.=\frac{100-\text{ loss }\%}{100 }\times C.P. \qquad\because \ S.P.<C.P.$

$Discount=M.P.-S.P.$

$Discount \ \% = \frac{M.P.-S.P.}{M.P.}\times 100$

$C.P. = M.P.-(Discount + Profit)=M.P.-(Discount - Loss)$

$\displaystyle \begin{equation*} \text{Median} = \left\{\begin{array}{rl} \frac{(n+1)}{2}th\text{ term} & \text{if } n \text{ is odd}, \\  \left[\frac{\frac{n}{2}th\text{ term} + \frac{(n+1)}{2}th\text{ term}}{2} \right] & \text{if } n \text{ is even}. \end{array} \right. \end{equation*} = l+\frac{\frac{n}{2}-C}{f} \times d$ 

Mode $ = \displaystyle l+\frac{f_m-f_{m-1}}{2f_m-f_{m-1}-f_{m+1}}\times d $

Mode = 3Median - 2Mean

Mean - Mode = 3 ( Mean - Median )

variance $=\begin{equation*} var(x)=\sigma^2 = \displaystyle\frac{\displaystyle \sum_{i=1}^n (x_{\tiny{i}} - \bar{x})^2}{n} = \displaystyle\frac{\displaystyle \sum_{i=1}^n f_{\tiny{i}}(x_{\tiny{i}} - \bar{x})^2}{\displaystyle \sum_{i=1}^n x_{\tiny{i}}} \end{equation*}$

## Logarithms

$ \qquad\displaystyle\begin{aligned} 
        &\textbf{Basic Definition : } && a^x = N \quad \Leftrightarrow \quad \log_a N = x, \quad (a>0,\; a \neq 1,\; N>0) \\ 
        &\textbf{Laws of Logarithms :} \\ 
 \qquad & 1. \quad  \log_a (MN) = \log_a M + \log_a N & &\quad \text{(Product rule)} \\
 \qquad & 2. \quad  \log_a \!\left(\frac{M}{N}\right) = \log_a M - \log_a N & &\quad \text{(Quotient rule)} \\
 \qquad & 3. \quad  \log_a (M^k) = k \cdot \log_a M & &\quad \text{(Power rule)} \\
 \qquad & 4. \quad  \log_{a^k} M = \tfrac{1}{k} \log_a M & &\quad \text{(Base power rule)} \\
 \qquad & 5. \quad  \log_a M = \frac{\log_b M}{\log_b a} & &\quad \text{(Change of base)} \\ 
\\ &\textbf{Special Values :} \\ & \log_a 1 = 0, \quad \log_a a = 1 \\
\end{aligned} $

* $\displaystyle a^m \times a^n = a^{m+n} \;\;\Rightarrow\;\; \log_a(MN) = \log_a M + \log_a N$
* $\displaystyle\frac{a^m}{a^n} = a^{m-n} \qquad\; \ \Rightarrow\;\; \log_a \left(\frac{M}{N}\right) = \log_a M - \log_a N$
* $\displaystyle\; (a^m)^n = a^{mn} \qquad\Rightarrow\;\; \log_a(M^k) = k \log_a M$
