<span style="color:red">Q. given $A$ and $B$ are two hermitian matrix. show that                                                                               
<span style="color:red"> 1. $A=0$ if $A^2=0$                                                                                                                  
<span style="color:red"> 2. $AB + BA$  is hermitian                                                                                                            
<span style="color:red"> 3. $AB-BA$ is also hermitian                                                                                                     
<span style="color:red"> 4. $AB$ is hermitian only if $AB=BA$

---

Let $A$ and $B$ be **Hermitian matrices**, i.e.,

$$
A^\dagger = A \quad \text{and} \quad B^\dagger = B
$$

We’ll analyze each statement:

---

#### **1. If $A^2 = 0$, then $A = 0$**

**Proof:**                                                                                                                               
Let $x$ be any complex vector. Consider : $ x^\dagger A^2 x = \|A x\|^2 \geq 0 $                                                   
If $A^2 = 0$, then : $x^\dagger A^2 x = 0 \Rightarrow \|A x\|^2 = 0 \Rightarrow A x = 0$                                    
This is true for all $x$, so : $A = 0$                                                                                                
 If $A$ is Hermitian and $A^2 = 0$, then $A = 0$.

---

#### **2. $AB + BA$ is Hermitian**

**Proof:**                                                                                                                      
$(AB + BA)^\dagger = (AB)^\dagger + (BA)^\dagger = B^\dagger A^\dagger + A^\dagger B^\dagger = BA + AB = AB + BA$             
The sum is Hermitian.

---

#### **3. $AB - BA$ is also Hermitian**

**Proof:**                                                                                                                      
$ (AB - BA)^\dagger = (AB)^\dagger - (BA)^\dagger = B^\dagger A^\dagger - A^\dagger B^\dagger = BA - AB = -(AB - BA)$           
So, $(AB - BA)^\dagger = - (AB - BA)$                                                                                           
$AB - BA$ is **skew-Hermitian**, not Hermitian.

---

#### **4. $AB$ is Hermitian only if $AB = BA$**

**Proof:**                                                                                                                      
$ (AB)^\dagger = B^\dagger A^\dagger = BA$                                                                                         
So for $AB$ to be Hermitian : $(AB)^\dagger = AB \Rightarrow BA = AB \Rightarrow AB = BA$                                    
$AB$ is Hermitian **only if** $AB = BA$ (i.e., A and B commute).

<span style="color:red">Q. prove the following :                                                                                                                
<span style="color:red"> 1. $(A^\dagger)^\dagger = A$                                                                                                                                                                                                                                
<span style="color:red"> 2. $(A+B)^\dagger = A^\dagger + B^\dagger$                                                                                                                                                                                                    
<span style="color:red"> 3. $(AB)^\dagger = B^\dagger A^\dagger$

---

We will prove the following identities involving **Hermitian conjugation** (also called **adjoint** or **dagger**) for complex matrices.                 
The Hermitian conjugate of a matrix $A$ is defined as:

$A^\dagger = (\bar{A})^T$

---

#### **1. $(A^\dagger)^\dagger = A$**

**Proof:**                                                                                                                    
$(A^\dagger)^\dagger = \left( (\bar{A})^T \right)^\dagger = \left( \overline{ (\bar{A})^T } \right)^T = \left( (\bar{A})^T \right)^\dagger = \left( \bar{(\bar{A})} \right)^T = A^T$                                                                                                                
But $A^\dagger = (\bar{A})^T \Rightarrow \bar{A} = (A^\dagger)^T \Rightarrow A = \overline{ (A^\dagger)^T } = (A^\dagger)^\dagger$

 **Hence**, $(A^\dagger)^\dagger = A$

---

#### **2. $(A + B)^\dagger = A^\dagger + B^\dagger$**

**Proof:**                                                                                                                       
$(A + B)^\dagger = \left( \overline{A + B} \right)^T = (\bar{A} + \bar{B})^T = (\bar{A})^T + (\bar{B})^T = A^\dagger + B^\dagger$                                                                                                                 
**Hence**, $ (A + B)^\dagger = A^\dagger + B^\dagger$

---

#### **3. $(AB)^\dagger = B^\dagger A^\dagger$**

**Proof:**                                                                                                                       
Let’s take the conjugate transpose of the product:                                                                                                                
$(AB)^\dagger = \left( \overline{AB} \right)^T = (\bar{A} \bar{B})^T = (\bar{B})^T (\bar{A})^T = B^\dagger A^\dagger$                                                                                                                
**Hence**, $ (AB)^\dagger = B^\dagger A^\dagger$

<span style="color:red">Q. show that the matrix multiplication follows the law of distribution , that is , $A(B+C)=AB+AC$

---

Let $A \in \mathbb{C}^{m \times n}$, and $B, C \in \mathbb{C}^{n \times p}$. Then:

$\begin{aligned} \left[A(B + C)\right]_{ij}  &= \sum_{k=1}^{n} A_{ik} \cdot (B_{kj} + C_{kj}) \\ &= \sum_{k=1}^{n} \left( A_{ik} B_{kj} + A_{ik} C_{kj} \right) \\ &= \sum_{k=1}^{n} A_{ik} B_{kj} + \sum_{k=1}^{n} A_{ik} C_{kj} \quad\text{(Using scalar distributivity)} \\ &= [AB]_{ij} + [AC]_{ij} \end{aligned} $

Hence, $A(B + C) = AB + AC$

<span style="color:red">Q. show that the diagonal elements of a skew hermitian matrix are either zero or pure imaginary

---

To show that the **diagonal elements** of a **skew-Hermitian matrix** are either **zero** or **pure imaginary**, we proceed with a direct proof using the definition.

---

 **Proof for Diagonal Elements**

Let $A$ be skew-Hermitian. Then for each diagonal element $A_{ii}$, we have:

$$
A_{ii} = -\overline{A_{ii}}
$$

Now let $A_{ii} = a + ib$, where $a, b \in \mathbb{R}$.

Then $\overline{A_{ii}} = a - ib$, and we substitute:

$$
a + ib = - (a - ib) = -a + ib
$$

Comparing both sides:

* Real parts: $a = -a$ → $a = 0$
* Imaginary parts: $ib = ib$ (always true)

Thus,

$$
A_{ii} = 0 + ib
$$

i.e., **pure imaginary**.

<span style="color:red">Q. show that the matrix addition is both commutative and associative , what is the scalar multiplication of a matrix ?

--- 

#### **Matrix Addition is Commutative**

Let

$$
A = [a_{ij}], \quad B = [b_{ij}]
$$

be two matrices of the same size (say $m \times n$).

Then:

$$
A + B = [a_{ij} + b_{ij}] = [b_{ij} + a_{ij}] = B + A
$$

 So, **matrix addition is commutative**:

$$
A + B = B + A
$$

---

#### **Matrix Addition is Associative**

Let

$$
A = [a_{ij}], \quad B = [b_{ij}], \quad C = [c_{ij}]
$$

be three matrices of the same size.

Then:

$$
(A + B) + C = [(a_{ij} + b_{ij}) + c_{ij}] = [a_{ij} + (b_{ij} + c_{ij})] = A + (B + C)
$$

 So, **matrix addition is associative**:

$$
(A + B) + C = A + (B + C)
$$

---

#### **Scalar Multiplication of a Matrix**

Let $k \in \mathbb{R}$ (or $\mathbb{C}$) be a scalar, and let $A = [a_{ij}]$ be a matrix.

Then:

$$
kA = [k \cdot a_{ij}]
$$

Each element of the matrix is **multiplied by the scalar**.

For example:

$$
A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \quad k = 3 \quad \Rightarrow \quad 3A = \begin{bmatrix} 3 & 6 \\ 9 & 12 \end{bmatrix}
$$

---

#### Scalar Multiplication is Also Compatible with Addition:

$$
k(A + B) = kA + kB
$$

$$

Start with the fact that:

$$
U^\dagger = U^{-1}

$$
(U^\dagger)^\dagger = (U^{-1})^\dagger

$$

$$
((U^{-1})^\dagger)^\dagger = U^\dagger = U^{-1}
$$

So:

$$
(U^{-1})^\dagger = (U^\dagger)^{-1} = U
\Rightarrow (U^{-1})^\dagger = (U^{-1})^{-1}
$$

Thus, $U^{-1}$ satisfies:

$$
(U^{-1})^\dagger = (U^{-1})^{-1}
\Rightarrow U^{-1} \text{ is unitary}
$$

 **Hence proved.**

<span style="color:red">Q. show that                                                                                                                  
i. the eigenvalues of a Hermitian matrix are real and                                                                                
ii. the eigen matrices corresponding to distinct eigenvalues are mutually orthogonal

--- 

#### Given:

A matrix $A \in \mathbb{C}^{n \times n}$ is **Hermitian**, i.e.,

$A^\dagger = A \quad \text{(i.e., } A = \overline{A}^T \text{)}$

---

#### Part (i): *Eigenvalues of a Hermitian matrix are real*

 Let :

* $A \mathbf{x} = \lambda \mathbf{x}$, where $\mathbf{x} \neq 0$ is an eigenvector.
* $\lambda \in \mathbb{C}$ is the corresponding eigenvalue.

##### Take the Hermitian inner product:

$\mathbf{x}^\dagger A \mathbf{x} = \lambda \mathbf{x}^\dagger \mathbf{x} \tag{1}$

Now take the **complex conjugate** of both sides:

But since $A$ is Hermitian, we also know:

$\mathbf{x}^\dagger A \mathbf{x} = (\mathbf{x}^\dagger A \mathbf{x})^* = (\lambda \mathbf{x}^\dagger \mathbf{x})^* = \bar{\lambda} \mathbf{x}^\dagger \mathbf{x} \tag{2}$

From (1) and (2), we equate:

$\lambda \mathbf{x}^\dagger \mathbf{x} = \bar{\lambda} \mathbf{x}^\dagger \mathbf{x}$

Since $\mathbf{x}^\dagger \mathbf{x} > 0$ (because $\mathbf{x} \neq 0$):

$\lambda = \bar{\lambda} \Rightarrow \lambda \in \mathbb{R}$

Hence, **The eigenvalues of a Hermitian matrix are always real.**

---

#### Part (ii): *Eigenvectors corresponding to distinct eigenvalues are orthogonal*

Let :

* $A \mathbf{x}_1 = \lambda_1 \mathbf{x}_1$
* $A \mathbf{x}_2 = \lambda_2 \mathbf{x}_2$, where $\lambda_1 \neq \lambda_2$
* $\mathbf{x}_1$ and $\mathbf{x}_2$ are eigenvectors of $A$


$\mathbf{x}_1^\dagger A \mathbf{x}_2 = \lambda_2 \mathbf{x}_1^\dagger \mathbf{x}_2 \tag{3}$

But since $A$ is Hermitian:

$\mathbf{x}_1^\dagger A \mathbf{x}_2 = (A \mathbf{x}_1)^\dagger \mathbf{x}_2 = (\lambda_1 \mathbf{x}_1)^\dagger \mathbf{x}_2 = \bar{\lambda}_1 \mathbf{x}_1^\dagger \mathbf{x}_2 = \lambda_1 \mathbf{x}_1^\dagger \mathbf{x}_2 \tag{4}$

From (3) and (4):

$\lambda_1 \mathbf{x}_1^\dagger \mathbf{x}_2 = \lambda_2 \mathbf{x}_1^\dagger \mathbf{x}_2
\Rightarrow (\lambda_1 - \lambda_2)\mathbf{x}_1^\dagger \mathbf{x}_2 = 0$

Since $\lambda_1 \neq \lambda_2$, we get:

$\mathbf{x}_1^\dagger \mathbf{x}_2 = 0 \Rightarrow \mathbf{x}_1 \perp \mathbf{x}_2$

Hence, **Eigenvectors corresponding to distinct eigenvalues of a Hermitian matrix are orthogonal.**

<span style="color:red">Q. $[A,B]=iC$ . if $A$ and $B$ are Hermitian matrices, then show that $C$ is also Hermitian

---

We are given:

$$
[A, B] = iC
\quad \text{where } A^\dagger = A \text{ and } B^\dagger = B \text{ (i.e., A and B are Hermitian)}
$$

We are to **show that** $C^\dagger = C$, i.e., **$C$ is Hermitian**.

Recall the definition of the commutator:

$[A, B] = AB - BA$

So the given equation becomes:

$AB - BA = iC \tag{1}$

Taking the Hermitian conjugate (†) of both sides of (1):

$(AB - BA)^\dagger = (iC)^\dagger$

$(AB)^\dagger - (BA)^\dagger = -i C^\dagger \Rightarrow B^\dagger A^\dagger - A^\dagger B^\dagger = -i C^\dagger$

<span style="color:red">Q. Show that :

<span style="color:red"> 1. a matrix and its transpose have the same characteristic root or eigenvalues.<span style="color:red">

<span style="color:red"> 2. $0$ is a characteristic root of a matrix, if and only if, the matrix is singular.<span style="color:red">

<span style="color:red"> 3. the characteristic roots of a Hermitian matrix are all real and those of a skew-Hermitian is either zero or pure imaginary.<span style="color:red">

<span style="color:red"> 4. the modulus of each characteristic root of a unitary matrix is unitary, i.e., characteristic roots are unimodular.<span style="color:red">

#### **1. A matrix and its transpose have the same characteristic roots (eigenvalues)**

Let $A$ be an $n \times n$ matrix.

The **characteristic polynomial** of $A$ is:

$ p_A(\lambda) = \det(A - \lambda I) $

We want to show:

$ \det(A - \lambda I) = \det(A^T - \lambda I) $

This follows from the property:

 $\det(M^T) = \det(M)$ for any square matrix $M$

So, $ \det(A^T - \lambda I) = \det((A - \lambda I)^T) = \det(A - \lambda I) $

Hence, the characteristic polynomials of $A$ and $A^T$ are equal ⇒ they have the same **eigenvalues**.

Hence,  **Conclusion:** A matrix and its transpose have the same characteristic roots.

---

#### **2. $0$ is a characteristic root (eigenvalue) of a matrix if and only if the matrix is singular**

Let $A$ be an $n \times n$ matrix.

* $0$ is an eigenvalue ⇔ $\det(A - 0 \cdot I) = \det(A) = 0$
* A matrix is **singular** ⇔ $\det(A) = 0$

So,  $ 0 \text{ is an eigenvalue of } A \iff A \text{ is singular} $

Hence, **Conclusion:** $0$ is a characteristic root if and only if the matrix is singular.

---

#### **3. The characteristic roots of a Hermitian matrix are real, and those of a skew-Hermitian matrix are zero or pure imaginary**

#### **(i) Hermitian matrix ⇒ eigenvalues are real**

Let $A$ be Hermitian: $A^\dagger = A$

Let $A v = \lambda v$, where $v \ne 0$

Take inner product:

$ \langle A v, v \rangle = \langle \lambda v, v \rangle = \lambda \langle v, v \rangle $

But since $A$ is Hermitian:

$ \langle A v, v \rangle = \langle v, A v \rangle^* = \langle v, \lambda v \rangle^* = \lambda^* \langle v, v \rangle $

Equating both:

$ \lambda \langle v, v \rangle = \lambda^* \langle v, v \rangle \Rightarrow \lambda = \lambda^* \Rightarrow \lambda \in \mathbb{R} $

Hence, **Eigenvalues of Hermitian matrices are real**

#### **(ii) Skew-Hermitian matrix ⇒ eigenvalues are zero or pure imaginary**

Let $A$ be skew-Hermitian: $A^\dagger = -A$

Suppose $A v = \lambda v$

Then, $ \langle A v, v \rangle = \lambda \langle v, v \rangle $

Also, $ \langle A v, v \rangle = \langle v, A v \rangle^* = \langle v, \lambda v \rangle^* = \lambda^* \langle v, v \rangle $

But using $A^\dagger = -A$, we also have, $ \langle A v, v \rangle = - \langle v, A v \rangle^* = - \lambda^* \langle v, v \rangle $

So, $ \lambda \langle v, v \rangle = - \lambda^* \langle v, v \rangle \Rightarrow \lambda = -\lambda^* \Rightarrow \lambda \text{ is purely imaginary or zero} $

Hence, **Eigenvalues of skew-Hermitian matrices are pure imaginary or zero**

---

#### **4. The modulus of each characteristic root (eigenvalue) of a unitary matrix is 1 (i.e., they are unimodular)**

Let $U$ be a **unitary matrix**, i.e., $ U^\dagger U = I $

Let $U v = \lambda v$

Then,  $ \| U v \|^2 = \| \lambda v \|^2 = |\lambda|^2 \| v \|^2 $

But also, $ \| U v \|^2 = \langle U v, U v \rangle = \langle v, U^\dagger U v \rangle = \langle v, v \rangle = \| v \|^2 $

Thus, $|\lambda|^2 \| v \|^2 = \| v \|^2 \Rightarrow |\lambda|^2 = 1 \Rightarrow |\lambda| = 1 $

Hence, **Eigenvalues of a unitary matrix lie on the unit circle: they are unimodular**

<span style="color:red">Q. State and explain, with an example, the Caley-Hamilton theorem. Hence find the inverse of a matrix. Can you prove the theorem ? 

---

**Cayley-Hamilton Theorem**
**Every square matrix satisfies its own characteristic equation.**

Let $A$ be an $n \times n$ matrix with characteristic polynomial:

$p(\lambda) = \det(A - \lambda I) = \lambda^n + a_1 \lambda^{n-1} + \cdots + a_{n-1} \lambda + a_n$

Then **Cayley-Hamilton theorem** says $p(A) = A^n + a_1 A^{n-1} + \cdots + a_{n-1} A + a_n I = 0$

---

Let $A = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}$

**Step 1: Characteristic polynomial**

$ \det(A - \lambda I) = 0$

$  \begin{vmatrix} 2-\lambda & 1 \\ 1 & 2-\lambda \end{vmatrix} = 0$

$ (2 - \lambda)^2 - 1 = \lambda^2 - 4\lambda + 3 = 0$

So, the characteristic equation is:

$\lambda^2 - 4\lambda + 3 = 0$

**Step 2: Apply Cayley-Hamilton theorem:**

$A^2 - 4A + 3I = 0$

* $A = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}$
* $A^2 = \begin{bmatrix} 5 & 4 \\ 4 & 5 \end{bmatrix}$
* $4A = \begin{bmatrix} 8 & 4 \\ 4 & 8 \end{bmatrix}$
* $3I = \begin{bmatrix} 3 & 0 \\ 0 & 3 \end{bmatrix}$

$A^2 - 4A + 3I = \begin{bmatrix} 5 & 4 \\ 4 & 5 \end{bmatrix} - \begin{bmatrix} 8 & 4 \\ 4 & 8 \end{bmatrix} + \begin{bmatrix} 3 & 0 \\ 0 & 3 \end{bmatrix} = \begin{bmatrix} 0 & 0 \\ 0 & 0 \end{bmatrix}$

**Hence Verified!**

---

**Finding the Inverse using Cayley-Hamilton**

From the above:

$A^2 - 4A + 3I = 0 \Rightarrow A^2 = 4A - 3I$

Multiply both sides by $A^{-1}$:

$A = 4I - 3A^{-1} \Rightarrow 3A^{-1} = 4I - A \Rightarrow A^{-1} = \frac{1}{3}(4I - A)$

So, $A^{-1} = \frac{1}{3} \left( 4 \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} - \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix} \right) = \frac{1}{3} \begin{bmatrix} 2 & -1 \\ -1 & 2 \end{bmatrix}$

---

**Proof of Cayley-Hamilton Theorem (Sketch)**

We can outline the proof using **adjugate matrix identity**.

Let $p(\lambda) = \det(A - \lambda I)$

We know,

$(A - \lambda I) \cdot \text{adj}(A - \lambda I) = \det(A - \lambda I) I = p(\lambda) I$

Now, the adjugate $\text{adj}(A - \lambda I)$ is a matrix with polynomial entries in $\lambda$.

So, let,

$\text{adj}(A - \lambda I) = B_0 + B_1 \lambda + \cdots + B_{n-1} \lambda^{n-1}$

Multiply both sides:

$(A - \lambda I)(B_0 + B_1 \lambda + \cdots + B_{n-1} \lambda^{n-1}) = p(\lambda) I$

Now compare coefficients of powers of $\lambda$ on both sides — you get:

$A^n + a_1 A^{n-1} + \cdots + a_n I = 0$

That proves Cayley-Hamilton theorem.

<span style="color:red">Q. Prove the theorem for the diagonalisation of a matrix. What is meant by similarity transformation ?

---

#### **Theorem: Diagonalization of a Matrix**

**Statement:**
A square matrix $A \in \mathbb{C}^{n \times n}$ is **diagonalizable** if there exists an invertible matrix $P$ and a diagonal matrix $D$ such that:

$A = P D P^{-1}$

This is equivalent to saying that:

* $A$ has $n$ **linearly independent eigenvectors**,
* and the diagonal entries of $D$ are the **eigenvalues** of $A$.


#### **Proof of Diagonalization Theorem:**

Let $A$ be an $n \times n$ matrix with $n$ **linearly independent** eigenvectors $\vec{v}_1, \vec{v}_2, \dots, \vec{v}_n$, and let the corresponding eigenvalues be $\lambda_1, \lambda_2, \dots, \lambda_n$.

Form a matrix $P$ using these eigenvectors as columns:

$P = \begin{bmatrix} \vec{v}_1 & \vec{v}_2 & \cdots & \vec{v}_n \end{bmatrix}$

Since the eigenvectors are linearly independent, $P$ is invertible.

Now define the diagonal matrix:

$ \qquad D = \begin{bmatrix} \lambda_1 & 0 & \cdots & 0 \\ 0 & \lambda_2 & \cdots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \cdots & \lambda_n \end{bmatrix} $

Then:

$\qquad A P = A \begin{bmatrix} \vec{v}_1 & \vec{v}_2 & \cdots & \vec{v}_n \end{bmatrix} = \begin{bmatrix} A \vec{v}_1 & A \vec{v}_2 & \cdots & A \vec{v}_n \end{bmatrix} = \begin{bmatrix} \lambda_1 \vec{v}_1 & \lambda_2 \vec{v}_2 & \cdots & \lambda_n \vec{v}_n \end{bmatrix} = P D $

So, $A P = P D \quad \Rightarrow \quad A = P D P^{-1}$

Hence, $A$ is **diagonalizable**.

---

A **similarity transformation** is a transformation of the form:

$\qquad B = P^{-1} A P $

Where:

* $A$ and $B$ are $n \times n$ matrices,
* $P$ is an invertible matrix.

We say that **$A$ and $B$ are similar** matrices.

**Why it's useful:**

* Similar matrices represent the same linear transformation under different bases.
* They share the **same eigenvalues**, **determinant**, **trace**, and **characteristic polynomial**.
* Diagonalizing a matrix is a similarity transformation that simplifies computations.

**Example:**

Let $A = \begin{bmatrix} 4 & 1 \\ 0 & 2 \end{bmatrix}$

Eigenvalues: $\lambda = 4, 2$

Eigenvectors:

* For $\lambda = 4$: $\vec{v}_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$
* For $\lambda = 2$: $\vec{v}_2 = \begin{bmatrix} -1 \\ 2 \end{bmatrix}$

Then:

* $P = \begin{bmatrix} 1 & -1 \\ 0 & 2 \end{bmatrix}$
* $D = \begin{bmatrix} 4 & 0 \\ 0 & 2 \end{bmatrix}$

So $A = P D P^{-1}$

<span style="color:red">Q. Prove that the inverse of an orthogonal matrix is an orthogonal matrix and the inverse of a unitary matrix is a unitary matrix.

---


#### **1. Inverse of an Orthogonal Matrix is Orthogonal** ($Q^{-1} \text{ is orthogonal}$)

Let $Q \in \mathbb{R}^{n \times n}$ be an **orthogonal matrix**.
By definition:

$$
Q^T Q = QQ^T = I
$$

##### Proof:

From $Q^T Q = I$, we see that:

$Q^T = Q^{-1}$

Now consider:

$(Q^{-1})^T Q^{-1} = (Q^T)^T Q^T = Q Q^T = I$

So,

$ (Q^{-1})^T Q^{-1} = I \Rightarrow Q^{-1} \text{ is orthogonal}$

 **Hence, the inverse of an orthogonal matrix is orthogonal.**

---

#### **2. Inverse of a Unitary Matrix is Unitary**  ($ U^{-1} \text{ is unitary} $)

Let $U \in \mathbb{C}^{n \times n}$ be a **unitary matrix**.
By definition:

$U^\dagger U = UU^\dagger = I$

where $U^\dagger = \overline{U}^T$ is the **conjugate transpose**.

##### Proof:

From the definition:

$ U^\dagger = U^{-1}$

Now compute:

$ (U^{-1})^\dagger U^{-1} = (U^\dagger)^\dagger U^\dagger = U U^\dagger = I $

So,

$ (U^{-1})^\dagger U^{-1} = I \Rightarrow U^{-1} \text{ is unitary} $

**Hence, the inverse of a unitary matrix is unitary.**

<span style="color:red">Q. Show that the product of two orthogonal matrices is an orthogonal matrix and the product of two unitary matrices is also a unitary matrix.

---

#### 1. **Product of Two Orthogonal Matrices is Orthogonal**  (The product $Q = Q_1 Q_2$ is also orthogonal.)

Let $Q_1$ and $Q_2$ be orthogonal matrices.

$(Q_1 Q_2)^T (Q_1 Q_2) = I \Rightarrow Q_1 Q_2 \text{ is orthogonal}.$

**Therefore, the product of two orthogonal matrices is orthogonal.**

$$U_1^\dagger U_1 = I, \quad U_2^\dagger U_2 = I$$

$$U^\dagger U = (U_1 U_2)^\dagger (U_1 U_2) = U_2^\dagger U_1^\dagger U_1 U_2 = U_2^\dagger I U_2 = U_2^\dagger U_2 = I$$

$$(U_1 U_2)^\dagger (U_1 U_2) = I \Rightarrow U_1 U_2 \text{ is unitary}.$$

**Therefore, the product of two unitary matrices is unitary.**
