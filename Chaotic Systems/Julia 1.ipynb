```python
from matplotlib import pyplot as plt
from random import random
from sympy import sqrt, re, im, I
plt.rcParams['text.usetex'] = True
plt.rcParams["figure.figsize"] = (20,12)
plt.rcParams['font.size'] = 20
plt.rcParams['legend.fontsize'] = 20
```

```python
def map_julia(a,b,k=15):
    num_iterations = 2**k
    
    def julia(X):
        x, y = X
        x1, y1 = x, y
        u = sqrt((x1-a)**2 + (y1-b)**2) / 2
        v = (x - a) / 2
        u1, v1 = sqrt(u + v), sqrt(u - v)
        xn, yn = u1, v1
        if y1 < b:
            yn = -yn
        if random() < 0.5:
            xn, yn = -u1, -yn
        return (xn, yn)
    
    x1 = (re(0.5 + sqrt(0.25 - (a + b*I)))).expand(complex=True)
    y1 = (im(0.5 + sqrt(0.25 - (a + b*I)))).expand(complex=True)
    is_unstable = 2 * abs(x1 + y1*I)
    print(is_unstable)
    X0 = [x1, y1]
    X, Y = [], []
    for i in range(num_iterations):
        xn, yn = julia(X0)
        X, Y = X + [xn], Y + [yn]
        X0 = [xn, yn]

    fig, ax = plt.subplots()
    ax.scatter(X, Y, color='blue', s=0.15)
    ax.axis('off')
    plt.show()
    return None
```

```python
map_julia(0.,0.)
```
